```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('test_picture/1yuan_1.jpg').convert('RGB')  # (RGB)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans0 = transforms.RandomAffine(degrees=0, shear=(45, 45), fillcolor=(255, 0, 9))
trans1 = transforms.RandomGrayscale(p=1)
# Apply randomly a list of transformations with a given probability
trans_apply = transforms.RandomApply([trans0, trans1], p=0.6)  # transformations
```

```python
trans_apply(img_resize)  # trans0trans1
```

```python
trans_apply(img_resize)  # trans0trnas1
```

```python
trans_apply(img_resize)  # trans0trans1
```
