```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1.5 * np.pi, 500)
y1 = np.sin(x)
y2 = np.sin(3 * x)

fig, ax = plt.subplots()

ax.fill(x, y1, 'g', alpha=0.8)  # 绘制曲线y1和y1首尾坐标连线构成的直线之间的区域
ax.fill(x, y2, 'r', alpha=0.8)
# Outline of the region we've filled in
ax.plot(x, y1, c='b', alpha=0.8)
ax.plot(x, y2, c='g', alpha=0.8)
ax.plot([x[0], x[-1]], [y1[0], y1[-1]], c='g', alpha=1)
ax.plot([x[0], x[-1]], [y2[0], y2[-1]], c='r', alpha=1)

plt.show()
```
