```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan.jpg').convert('RGB')  # (RGB)
img
```

```python
trans = transforms.CenterCrop(size=(224, 200))  # Crops the given PIL Image at the center.
'''
size (sequence or int) 
    Desired output size of the crop. If size is an int instead of sequence like (h, w), a square crop (size, size) is made.
'''
trans_img = trans(img)
trans_img
```
