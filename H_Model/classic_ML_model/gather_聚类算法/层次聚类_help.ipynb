&emsp;&emsp;AGNES是一种采用自下向上聚合的层次聚类算法.它先将数据集中的每个样本看作一个初始聚类簇,
然后在算法运行的每一步中找出距离最近的两个聚类簇进行合并,该过程不断重复,直到达到预设的聚类簇个数.这里的关键是如何计算聚类簇之间的距离.
实际上,每个簇是一个样本集合,因此,只需采用关于集合的某种距离即可.例如,给定簇$C_i$与$C_j$,可通过下面的式子来计算距离          

$$
\begin{aligned}
& \text{最小距离}: d_{min}(C_i,C_j) = \min_{\mathbf{x} \in C_i,\mathbf{z} \in C_j} dist(\mathbf{x}, \mathbf{z}) \\
& \text{最大距离}: d_{max}(C_i,C_j) = \max_{\mathbf{x} \in C_i,\mathbf{z} \in C_j} dist(\mathbf{x}, \mathbf{z}) \\   
& \text{平均距离}: d_{avg}(C_i,C_j) = \frac{1}{|C_i| |C_j|} \sum_{\mathbf{x} \in C_i} \sum_{\mathbf{z} \in C_j} dist(\mathbf{x}, \mathbf{z}) \\  
\end{aligned} 
$$        

显然,最小距离由两个簇的最近样本决定,最大距离由两个簇的最远距离决定,而平均距离则由两个簇的所有样本共同决定.
当样本簇距离由$ d_{min},d_{max} $或$ d_{avg} $计算时,AGNES相应地称为"单链接"(single-linkage),"全链接"(complete-linkage)或"均链接"(average-linkage)算法.   



<font color='red' size=4>算法:</font>

***

输入:样本集$ D=\{ \mathbf{x}_1, \mathbf{x}_2, \cdots, \mathbf{x}_m \} $;           
&emsp;&emsp;聚类簇距离度量函数$d$;      
&emsp;&emsp;聚类簇数$k$;     

过程:     
1. 初始化单样本聚类簇
       
**for** $j=1,2,\cdots,m$ **do**

&emsp;&emsp;$C_j=\{ \mathbf{x}_j \}$

**end for**

2.初始化聚类簇距离矩阵 

**for** $i=1,2,\cdots,m$ **do**

&emsp;&emsp;**for** $j=i+1, \cdots,m$ **do**

&emsp;&emsp;&emsp;&emsp;$M(i,j)=d(C_i, C_j)$  

&emsp;&emsp;&emsp;&emsp;$M(i,j)=M(j,i))$  

&emsp;&emsp;**end for**

**end for**

3. 迭代合并距离最小的两个类

**while** $q>k$ **do**

&emsp;&emsp;找出距离最近的两个聚类簇$C_{i^*}$和$C_{j^*}$;

&emsp;&emsp;合并$C_{i^*}$和$C_{j^*}$:$ C_{i^*} = C_{i^*} \bigcup C_{j^*} $

&emsp;&emsp;**for** $ j=j^* +1,j^*+2,\cdots,q $  **do**

&emsp;&emsp;&emsp;&emsp;将聚类簇$C_j$重编号为$C_{j-1}$

&emsp;&emsp;**end for**

&emsp;&emsp;删除距离矩阵$M$的第$j^*$行与第$j^*$列

&emsp;&emsp;**for** $j=1,2,\cdots,q-1$ **do**

&emsp;&emsp;&emsp;&emsp;$M(i^*,j)=d(C_{i^*}, C_j)$  

&emsp;&emsp;&emsp;&emsp;$M(i^*,j)=M(j,i^*))$  

&emsp;&emsp;**end for**

&emsp;&emsp;$q=q-1$

**end while**

输出:簇划分$\mathcal{C} = \{ C_1, C_2, \cdots, C_K \}$
