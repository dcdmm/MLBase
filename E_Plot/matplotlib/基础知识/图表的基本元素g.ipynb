1. Axes.legend

```python
import numpy as np
import matplotlib.pyplot as plt

a = b = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1]
fig, ax = plt.subplots(2, 1, figsize=(10, 16))

line1 = ax[0].plot(a, c, 'k--', label='Model length')
line2 = ax[0].plot(a, d, 'k:', label='Data length')
line3 = ax[0].plot(a, c + d, 'k', label='Total message length')
ax[0].legend(edgecolor='white')
'''
The elements to be added to the legend are automatically determined, 
when you do not pass in any extra arguments.In this case, the labels
are taken from the artist.
'''

ax[1].plot(a, c, 'k--', color='red')
ax[1].plot(a, d, 'k:', color='green')
ax[1].plot(a, c + d, 'k', 'blue')
legends = ax[1].legend(labels=('Model length', 
                     'Data length',
                     'Total message length'), # 每条图例的名称
             loc='upper left', # 图例位置的基准点
             bbox_to_anchor=(0.5, 0.85), # 图例的位置(左上坐标为0.5, 0.85)
             prop={'size': 10, 'weight':'bold'}, # 图例的字体属性设置
             ncol=3, # 图例的列数
             facecolor='yellow', # 图例的背景颜色 
             edgecolor='red', # 图例边框的颜色
             title="legend", # 图例标题
             title_fontsize=20, # # 图例标题大小
             framealpha=0.4, # 图例facecolor,edgecolor的透明度
             )
```
