<font color='red' size=4>定义:</font>   
&emsp;&emsp;设随机变量$X\sim \chi_{m}^2, Y\sim \chi_{n}^2,$且$X$与$Y$相互独立,则随机变量$F=\frac{X/m}{Y/n}$ 
的分布称为自由度为$m$和$n$的$F$分布,记为$F \sim F_{m,n}$  

<font color='red' size=4>性质:</font>
1. 设$F \sim F_{m,n}$对于给定的$\alpha,0<\alpha<1$称满足条件  
$$ P\{F>F_{m,n}(\alpha)\} =\int_{F_{m,n}(\alpha)}^{+\infty} f(x)dx=\alpha $$   
的点$F_{m,n}(\alpha)$为$F_{m,n}$上的$\alpha$分位点  
2. 设$X \sim F_{m,n},记Y=1/X,则Y \sim F_{n,m}$

```python
# IMPORTS
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style
import warnings
warnings.filterwarnings('ignore') #忽略F分布分母不能为0的警告

style.use('classic')
fig = plt.figure(figsize=(14, 7), dpi=100)
ax  = plt.subplot(111)

# PDF
ax.plot(np.linspace(0, 10, 1000), 
         stats.f.pdf(np.linspace(0, 10, 1000), dfn=1, dfd=1) ,)
ax.fill_between(np.linspace(0, 10, 1000),
                 stats.f.pdf(np.linspace(0, 10, 1000), dfn=1, dfd=1) ,
                 alpha=.15)

# CDF
ax.plot(np.linspace(0, 10, 1000), 
         stats.f.cdf(np.linspace(0, 10, 1000), dfn=1, dfd=1))

ax.text(x=1, y=.25, s="pdf (normed)", alpha=.95, weight="bold", color="#008fd5")
ax.text(x=3, y=.85, s="cdf", alpha=.95, weight="bold", color="#fc4f30")

ax.set_xticks(np.arange(0, 11, 2))
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = 0, y = 4.0, s = r"$F$ Distribution - Overview",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = 0, y = 3.53, 
         s = 'Depicted below are the normed probability density function (pdf) and the cumulative density\nfunction (cdf) of a Fdistributed random variable $ F \sim F_{m,n} $, given dfn=1,dfd=1',
         fontsize = 19, alpha = .85)
```
