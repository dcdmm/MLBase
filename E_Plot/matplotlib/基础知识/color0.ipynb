```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(4, 1, figsize=(7, 40))

mu = 100  
sigma = 15  
x = mu + sigma * np.random.randn(437)
num_bins = 50
ax[0].hist(x, num_bins, density=True,
                           facecolor='g', edgecolor='r')
s = pd.Series(np.random.randint(0, 10, 16), index=list('abcdefghijklmnop'))
ax[1].bar(s.index, s, facecolor='g', edgecolor='r')

q = np.linspace(0, 1, 500)
y1 = np.sin(4 * np.pi * q) * np.exp(-5*q)
y2 = -np.sin(4 * np.pi * q) * np.exp(-5*q)
ax[2].fill(q, y1, facecolor='g', edgecolor='r')

f = ax[3].boxplot(x, patch_artist=True)
for box in f['boxes']:
    box.set( edgecolor='r', linewidth=1) 
    box.set( facecolor = 'g' ,alpha=1) 
    
'''
color:Set both the edgecolor and the facecolor
facecolor:Set the patch face color
edgecolor:Set the patch edge color
alpha:透明度(0-1)
可用于:hist, bar, barh, fill, fill_between, boxplot
'''

# 常见的颜色设置格式有
# 1.十六进制RGB字符串,如:#eafff5(★★★一般使用该格式) 
# 2. 颜色的具体名称,如:green
    # 一般使用:blue,green,red,cyan/青色,red,magenta/品红,yellow,black,white
    #         purple/紫色,pink/粉红色,brown/棕色,orange/橙色,grey/灰色
# 3. 颜色具体名称的缩写,如,g
# 4. RGB tuple,如:(255, 255, 255)
# 5. RGBA tuple,如(1.0, 0.0, 0.0, 1.0)
'''
以下颜色支持缩写：
字符	        颜色
'b'"blue	    蓝色
'g':green	    绿色
'r':red	        红色
'c':cyan	    青色
'm':magenta	    品红
'y':yellow	    黄色
'k':black	    黑色
'w':white       白色
'''

```
