```python
import pandas as pd
import numpy as np

np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 2] = np.nan
```

```python
df.style
```

```python
df.style.format("{:.2%}")
```

```python
df.style.format({'B': "{:0<4.1f}", 'D': '{:+.2f}'}) # 使用字典格式化特定列
```

```python
df.style.format({"B": lambda x: "±{:.2f}".format(abs(x))}) # 灵活的格式化操作(B列中的每个元素)
```
