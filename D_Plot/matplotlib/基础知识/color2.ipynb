### matplotlib颜色相关设置大全

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0.0, 2.0, 201)
s = np.sin(2 * np.pi * t)

fig, ax = plt.subplots(facecolor='green') # 整个图片背景颜色
ax.set_facecolor('#eafff5') # axes背景颜色
ax.set_title('Voltage vs. time chart', color='red') # 图表标题颜色设置
ax.set_xlabel('time (s)', color='r') # x轴标题颜色设置(y轴同理)
ax.plot(t, s, color='brown', label='brown')
ax.plot(t, .7*s, color='#0165fc', linestyle='--')
ax.set_xticklabels(list('ABCDEFGFHJ'), color='#ff00ff') # xticklabels颜色设置(y轴同理)
ax.grid(True, axis='y', color='b') # 网格线颜色
ax.tick_params(axis='x', color='red', width=7) # 刻度线颜色
ax.legend(facecolor='yellow', edgecolor='red')
ax.text(1, -0.6, r'$\sum_{i=0}^\infty sin(x)$', fontsize=20, # 注释点坐标指的是该点的(xticks, yticks)
         color='green') # text颜色设置
# 箭头颜色设置
ax.annotate('sin(x)', xy=(1.2, 1), xytext=(1, -0.2), color='pink', 
            arrowprops=dict(edgecolor='r', facecolor='g',
                            arrowstyle='<|-')) #注释点坐标指的是该点的(xticks, yticks)
```
