```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
fig, axes = plt.subplots(3, 5, figsize=(40, 15))
df = pd.DataFrame(np.random.rand(50, 15))
colors = plt.cm.tab10(np.linspace(0, 1, len(df.columns)))  # tab10(Colormaps)的区间为[0,1],取将区间分为15等分处点的颜色
print(colors)  # RGBA颜色列表
print(plt.cm.tab10)  # ListedColormap object
for i in df.columns:
    j = i // 5  # 地板除法
    k = i % 5
    axes[j][k].plot(df[i], color=colors[i], linewidth=4)
```

```python
fig = plt.figure(figsize=(12, 7))
ax1 = plt.subplot(121)
x = np.random.randn(1000)
y = np.random.randn(1000)
color_list = ListedColormap(['red', 'green', 'blue', 'yellow'])  # Colormap object generated from a list of colors.

s = ax1.scatter(x, y, marker='_', alpha=0.9,
                s=np.abs(np.random.randn(1000) * 100),
                c=x, cmap='Set1',  # 内置colormaps
                lw=10, edgecolor='red')
fig.colorbar(s)

ax2 = plt.subplot(122)
s1 = ax2.scatter(x, y, marker='_', alpha=0.9,
                 s=np.abs(np.random.randn(1000) * 100),
                 c=x, cmap=color_list,  # 自定义colormaps
                 lw=10, edgecolor='red')
fig.colorbar(s1)
```
