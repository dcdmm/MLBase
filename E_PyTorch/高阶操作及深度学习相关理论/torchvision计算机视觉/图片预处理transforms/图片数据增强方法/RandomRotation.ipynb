```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # 读取图片(RGB模式)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
# 新增部分默认填充为0(即为黑色)
trans = transforms.RandomRotation(degrees=45) # 逆时针旋转
'''
degrees (sequence or float or int) –
    Range of degrees to select from. If degrees is a number instead of sequence like (min, max), the range of degrees will be (-degrees, +degrees).
'''
trans(img_resize)
```

```python
trans1 = transforms.RandomRotation(degrees=(45, 90), expand=True)
'''
expand (bool, optional) –
    Optional expansion flag. If true, expands the output to make it large enough to hold the entire rotated image.
    If false or omitted, make the output image the same size as the input image.
    ★★★★★Note that the expand flag assumes rotation around the center and no translation.
'''
last = trans1(img_resize)
last
```

```python
last.size # 设置expand=True后,图片大小发生改变.训练时若需图片size相等,则使用Resize进行缩放成同大小
```

```python
trans2 = transforms.RandomRotation(degrees=(45, 90), fill=(178, 56, 21))
'''
fill (n-tuple or int or float) –
    Pixel fill value for area outside the rotated image. If int or float, the value is used for all bands respectively. Defaults to 0 for all bands.
'''
trans2(img_resize)
```

```python
trans3 = transforms.RandomRotation(degrees=(35, 35),
                                   fill=(178, 56, 21),
                                   center=(0, 0)) # 旋转中心为左上角
'''
center (2-tuple, optional) –
    Optional center of rotation. Origin is the upper left corner. Default is the center of the image.
'''
trans3(img_resize)
```
