<font color='red' size=4>定理:</font>设$ X $是$ m $维随机变量,$\Sigma$是$X$的协方差矩阵,
$\Sigma$的特征值分别是$ \lambda_1 \geq \lambda_2 \geq \cdots,\lambda_m \geq 0 $(因为协方差矩阵是半正定矩阵,故特征均大于等于0),
特征值对应的单位特征向量分别是$ \boldsymbol{\alpha}_1, \boldsymbol{\alpha}_2, \cdots,\boldsymbol{\alpha}_m $,
则$X$的第$k$主成分是

$$ Y_k = \boldsymbol{\alpha}_k^T X $$

$ X $的第$ k $主成分的方差是

$$ \mathrm{var}(Y_k) = \boldsymbol{\alpha}_k^T \Sigma \boldsymbol{\alpha}_k = \lambda_k, \quad k=1,2,\cdots,m $$

<font color='red' size=4>证明:</font>采用拉格朗日乘子法求出主成分

&emsp;&emsp;首先求$X$的第一主成分$Y_1 = \boldsymbol{\alpha}_1^T X$,即求稀疏向量$\boldsymbol{\alpha}_1^T $.易知,第一主成分的$\boldsymbol{\alpha}_1$是在$\boldsymbol{\alpha}_1^T \boldsymbol{\alpha}_1$条件下,$X$的所有线性变换中使方差

$$ \mathrm{var}(Y_1) = \boldsymbol{\alpha}_1^T \Sigma \boldsymbol{\alpha}_1 $$

达到最大的.

&emsp;&emsp;求第一主成分就是求解约束最优化问题

$$
\begin{aligned}
& \max_{ \boldsymbol{\alpha}_1}  \quad \boldsymbol{\alpha}_1^T \Sigma  \boldsymbol{\alpha}_1 \\
& s.t. \quad  \boldsymbol{\alpha}_1^T \boldsymbol{\alpha}_1 =1
\end{aligned}
$$

定义拉格朗日函数

$$ \boldsymbol{\alpha}_1^T \Sigma  \boldsymbol{\alpha}_1 - \lambda( \boldsymbol{\alpha}_1^T \boldsymbol{\alpha}_1 -1 ) $$

其中$\lambda$是拉格朗日乘子.将拉格朗日函数对$\boldsymbol{\alpha}_1$求导,并令其为0,得

$$ \Sigma \boldsymbol{\alpha}_1 - \lambda \boldsymbol{\alpha}_1 =0 $$

因此,$\lambda$是$\Sigma$得特征值,$\boldsymbol{\alpha}_1$是对应的单位特征向量.于是,目标函数

$$ \boldsymbol{\alpha}_1^T \Sigma \boldsymbol{\alpha}_1 = \boldsymbol{\alpha}_1^T \lambda \boldsymbol{\alpha}_1 = \lambda \boldsymbol{\alpha}_1^T \boldsymbol{\alpha}_1 = \lambda $$

&emsp;&emsp;假设$\boldsymbol{\alpha}_1$是$\Sigma$的最大特征值$\lambda_1$对应的单位特征向量,显然$\boldsymbol{\alpha}_1$与$\lambda_1$是最优化问题的解.所以$\boldsymbol{\alpha}_1^T X$构成第一主成分,其方差等于协方差矩阵的最大特征值.

$$ \mathrm{var}(\boldsymbol{\alpha}_1^T X) = \boldsymbol{\alpha}_1^T \Sigma \boldsymbol{\alpha}_1 = \lambda_1 $$

&emsp;&emsp;接着求$X$的第二主成分$Y_2 = \boldsymbol{\alpha}_2^T X $.第二主成分的$\boldsymbol{\alpha}_2$是在$\boldsymbol{\alpha}_2^T \boldsymbol{\alpha}_2=1$,且$\boldsymbol{\alpha}_2^T X$与$ \boldsymbol{\alpha}_1^T X $不相关条件下,$X$的所有线性变换中使方差

$$ \mathrm{var}(Y_2) = \boldsymbol{\alpha}_2^T \Sigma \boldsymbol{\alpha}_2 $$

达到最大的.

&emsp;&emsp;求第二主成分需要求解约束最优化问题

$$
\begin{aligned}
& \max_{ \boldsymbol{\alpha}_2}  \quad \boldsymbol{\alpha}_2^T \Sigma  \boldsymbol{\alpha}_2 \\
& s.t. \quad  \boldsymbol{\alpha}_2^T \boldsymbol{\alpha}_2 =1  \\
& \quad \quad \quad  \boldsymbol{\alpha}_1^T \Sigma \boldsymbol{\alpha}_2 =0, \quad  \boldsymbol{\alpha}_2^T \Sigma \boldsymbol{\alpha}_1=0
\end{aligned}
$$

注意到

$$ \boldsymbol{\alpha}_1^T \Sigma \boldsymbol{\alpha}_2 =  \boldsymbol{\alpha}_2^T \Sigma \boldsymbol{\alpha}_1 =  \boldsymbol{\alpha}_2^T \lambda_1 \boldsymbol{\alpha}_1 =  \lambda_1 \boldsymbol{\alpha}_2^T  \boldsymbol{\alpha}_1 = \lambda_1 \boldsymbol{\alpha}_1^T  \boldsymbol{\alpha}_2 $$

以及

$$ \boldsymbol{\alpha}_1^T  \boldsymbol{\alpha}_2 =0,\quad  \boldsymbol{\alpha}_2^T  \boldsymbol{\alpha}_1 = 0 $$

定义拉格朗日函数

$$ \boldsymbol{\alpha}_2^T \Sigma  \boldsymbol{\alpha}_2 - \lambda( \boldsymbol{\alpha}_2^T \boldsymbol{\alpha}_2 -1 ) - \phi  \boldsymbol{\alpha}_2^T  \boldsymbol{\alpha}_1 $$

其中$\lambda, \phi$是拉格朗日乘子.对$\boldsymbol{\alpha}_2$求导,并令其为0,得

$$ 2 \Sigma \boldsymbol{\alpha}_2 - 2 \lambda \boldsymbol{\alpha}_2 - \phi \boldsymbol{\alpha}_1 =0 \tag{2} $$

将方差左乘以$\boldsymbol{\alpha}_1^T$有

$$ 2 \boldsymbol{\alpha}_1^T \Sigma \boldsymbol{\alpha}_2 - 2 \lambda \boldsymbol{\alpha}_1^T \boldsymbol{\alpha}_2 - \phi \boldsymbol{\alpha}_1^T  \boldsymbol{\alpha}_1 =0 $$

此式前两项为0,且$\boldsymbol{\alpha}_1^T\boldsymbol{\alpha}_1 = 1$,导出$\phi=0$,因此式(2)可化简为

$$ \Sigma \boldsymbol{\alpha}_2 -  \lambda \boldsymbol{\alpha}_2 =0 $$
由此,$\lambda$是$\Sigma$的特征值,$\boldsymbol{\alpha}_2$是对应的单位特征向量.于是,目标函数

$$ \boldsymbol{\alpha}_2^T \Sigma \boldsymbol{\alpha}_2 = \boldsymbol{\alpha}_2^T \lambda \boldsymbol{\alpha}_2 = \lambda \boldsymbol{\alpha}_2^T \boldsymbol{\alpha}_2 = \lambda $$

&emsp;&emsp;假设$\boldsymbol{\alpha}_2$是$\Sigma$第二大特征值$\lambda_2$对应的单位特征向量,显然$\boldsymbol{\alpha}_2$与$\lambda_2$是以上最优化问题的解.于是$ \boldsymbol{\alpha}_2^T X$构成第二主成分,其方差等于协方差矩阵的第二大特征值,

$$ \mathrm{var}(\boldsymbol{\alpha}_2^T X) = \boldsymbol{\alpha}_2^T \Sigma \boldsymbol{\alpha}_2 = \lambda_2 $$

&emsp;&emsp;一般地,$X$的第$k$主成分是$\boldsymbol{\alpha}_k^T X$,并且$\mathrm{var}(\boldsymbol{\alpha}_2^T X) = \lambda_k$,
这里$\lambda_k$是$\Sigma$的第$k$个特征值并且$\boldsymbol{\alpha}_k$是对应的单位特征向量.

&emsp;&emsp;按照上述方法求得第一,第二,直到第$m$个主成分,其系数向量$\boldsymbol{\alpha}_1,\boldsymbol{\alpha}_2,\cdots,\boldsymbol{\alpha}_m$分别是$\Sigma$的第一个,
第二个,直到第$m$个单位特征向量,$\lambda_1, \lambda_2,\cdots,\lambda_m$分别是对应的特征值,并且,
第$k$个主成分的方差等于$\Sigma$的第$k$个特征值,

$$ \mathrm{var}(Y_k) = \boldsymbol{\alpha}_k^T \Sigma \boldsymbol{\alpha}_k = \lambda_k, \quad k=1,2,\cdots,m $$


&emsp;&emsp;<font color='red' size=4>推论:</font>$m$维随机向量$Y=(Y_1,Y_2,\cdots,Y_m)^T$的分量依次是$X$的第一主成分到第$m$主成分的充要条件是:

1. $Y = A^T X$,$A$为正交矩阵

$$
\begin{aligned}
A =
\begin{bmatrix}
\alpha_{11} & \alpha_{12} & \dots  & \alpha_{1m} \\
\alpha_{21} & \alpha_{22} & \dots  & \alpha_{2m} \\
\vdots      &  \vdots     &        &  \vdots     \\
\alpha_{m1} & \alpha_{m2} & \dots  & \alpha_{mm} \\
\end{bmatrix}
\end{aligned}
$$

2. $Y$的协方差矩阵为对角矩阵

$$
\begin{aligned}
& cov(Y) = \mathrm{diag}(\lambda_1, \lambda_2, \cdots, \lambda_m)  \\
& \lambda_1 \geq \lambda_2 \geq \cdots, \lambda_m
\end{aligned}
$$

其中$\lambda_k$是$\Sigma$的第$k$个特征值,$\boldsymbol{\alpha}_k$是对应的单位特征向量,$k=1,2,\cdots,m$.

&emsp;&emsp;以上证明中,$\lambda_k$是$\Sigma$的第$k$个特征值,$\boldsymbol{\alpha}_k$(其中$\boldsymbol{\alpha}_k^T = (\alpha_{1k}, \alpha_{2k}, \cdots, \alpha_{mk})$,即矩阵$A$的第$k$列)是对应的单位特征向量,即

$$ \Sigma_k \boldsymbol{\alpha}_k = \lambda_k \boldsymbol{\alpha}_k, \quad k=1,2,\cdots,m \tag{3} $$

用矩阵表示即为

$$ \Sigma A = A \Lambda \tag{4} $$

这里$A= [\alpha_{ij}]_{m \times m}$,$\Lambda$是对角矩阵,其第$k$个对角元素是$\lambda_k$.因为$A$是正交矩阵,即$A^T A = A A^T = I$,由式(4)得到两个公式

$$ A^T \Sigma A = \Lambda \tag{5} $$
和

$$ \Sigma = A \Lambda A^T \tag{6} $$

下面叙述总体主成分的性质:

1. 总体主成分$Y$的协方差矩阵是对角矩阵

$$ cov(Y) = \Lambda =  \mathrm{diag}(\lambda_1, \lambda_2, \cdots, \lambda_m) \tag{7} $$

2. 总体主成分$Y$的方差之和等于随机变量$X$的方差之和,即

$$ \sum_{i=1}^m \lambda_i = \sum_{i=1}^m \sigma_{ii} \tag{8} $$

其中$\sigma_{ii}$是随机变量$X_i$的方差,即协方差矩阵$\Sigma$的对角元素.事实上,利用式(6)及矩阵的迹(trace)的性质,可知

$$
\begin{aligned}
\sum_{i=1}^m \mathrm{var}(x_i) &= \mathrm{tr}(\Sigma^T) = \mathrm{tr}(A \Lambda A^T) = \mathrm{tr}(A^T \Lambda A) \\
					  &= \mathrm{tr}(\Lambda) = \sum_{i=1}^m \lambda_i = \sum_{i=1}^m \mathrm{var}(Y_i)
\end{aligned}
$$

3. 第$k$个主成分$Y_k$与变量$X_i$的相关系数$\rho(Y_k, X_i)$称为因子载荷量(factor loading),它表示第$k$个主成分$Y_k$与变量$X_i$的相关系数,计算公式是

$$ \rho(Y_k, X_i) = \frac{\sqrt{\lambda_k} \alpha_{ik}}{\sqrt{\sigma_{ii}}},\quad k,i=1,2,\cdots,m \tag{9} $$

因为

$$ \rho(Y_k, X_i) = \frac{cov(Y_k, X_i)}{\sqrt{\mathrm{var}(Y_k)\mathrm{var}(X_i)}} = \frac{cov(\boldsymbol{\alpha}_k^T X, \mathbf{e}_i^T X)}{\sqrt{\lambda_k}\sqrt{\sigma_{ii}}} $$

其中$ \mathbf{e}_i$为基本单位向量,其第$i$个分量为1,其余为0.再由协方差的性质

$$ cov(\boldsymbol{\alpha}_k^T X, \mathbf{e}_i^T X) = \boldsymbol{\alpha}_k^T \Sigma \mathbf{e}_i = \mathbf{e}_i^T \Sigma \boldsymbol{\alpha}_k = \lambda_k  \mathbf{e}_i^T \boldsymbol{\alpha}_k  = \lambda_k \alpha_{ik} $$

4. 第$k$个主成分$Y_k$与$m$个变量的因子载荷量满足

$$ \sum_{i=1}^m \sigma_{ii} \rho^2(Y_k, X_i) = \lambda_k $$

由式(9)有

$$ \sum_{i=1}^{m} \sigma_{ii}  \rho^2(Y_k, X_i) = \lambda_k = \sum_{i=1}^m \lambda_k \sigma_{ik}^2 = \lambda_k \boldsymbol{\alpha}_k^T \boldsymbol{\alpha}_k = \lambda_k $$

5. $m$个主成分与第$i$个变量$X_i$的因子载荷量满足

$$ \sum_{k=1}^m \rho^2(Y_k, X_i) = 1 $$

&emsp;&emsp;由于$Y_1,Y_2,\cdots,Y_m$互不相关,故

$$ \rho^2(X_i, (Y_1,Y_2,\cdots,Y_m)) = \sum_{k=1}^m \rho^2(Y_k, X_i) $$

又因$X_i$可以表示为$Y_1,Y_2,\cdots,Y_m$的线性组合(由$Y=A^TX$得$X=AY$),所以$X_i$与$Y_1,Y_2,\cdots,Y_m$的相关系数的平方为1,即

$$ \rho^2(X_i,(Y_1,Y_2,\cdots,Y_m)) = 1 $$
