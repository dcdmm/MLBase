```python
import tensorflow as tf
```

### sigmoid激活函数
### 参考torch.nn.Sigmoid

```python
x = tf.linspace(-5, 5, 200)
```

```python
# 函数
tf.nn.sigmoid(x)
```

```python
# 类
layer = tf.keras.layers.Activation(tf.nn.sigmoid, name='sigmoid')  # 可简称为'sigmoid'
output = layer(x)

print(layer.name)
print(output)
```

### 激活函数tanh
### 参考torch.nn.Tanh

```python
x = tf.linspace(-20, 20, 200)
```

```python
tf.nn.tanh(x)
```

```python
layer = tf.keras.layers.Activation(tf.nn.tanh)  # 可简称为'tanh'
output = layer(x)
output
```

### 激活函数relu
### 参考torch.nn.ReLU

```python
x = tf.linspace(-5, 5, 200)
```

```python
tf.nn.relu(x)
```

```python
layer = tf.keras.layers.Activation(tf.nn.relu)  # 可简称为'relu'
output = layer(x)
output
```
