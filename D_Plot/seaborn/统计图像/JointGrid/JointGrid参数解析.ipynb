```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
g = sns.JointGrid(marginal_ticks=True,  # f False, suppress ticks on the count/density axis of the marginal plots.
                  # Size of each side of the figure in inches (it will be square).
                  height=6,
                  # Ratio of joint axes height to marginal axes height.
                  ratio=3,
                  # Space between the joint and marginal axes
                  space=.2)

x, y = penguins["bill_length_mm"], penguins["bill_depth_mm"]

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species",
                ax=g.ax_joint)  # 
sns.histplot(data=penguins, x="bill_length_mm",
             ax=g.ax_marg_x)  # 
sns.kdeplot(data=penguins, y="bill_depth_mm",
            ax=g.ax_marg_y)  # 
```
