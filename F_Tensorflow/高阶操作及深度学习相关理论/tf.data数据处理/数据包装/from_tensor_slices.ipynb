```python
import tensorflow as tf
```

### 参考torch.utils.data.TensorDataset

```python
x = tf.linspace(0.0, 0.1, 100)
y = tf.reshape(tf.linspace(1, 400, 400), (100, 2, 2))
z = tf.range(1000, 1100)
```

```python
# Creates a Dataset whose elements are slices of the given tensors.
for i in tf.data.Dataset.from_tensor_slices(x).as_numpy_iterator():
    print(i)
```

```python
# The given tensors are sliced along their first dimension.
# This operation preserves the structure of the input tensors, removing the first dimension of each tensor and using it as the dataset dimension.
# All input tensors must have the same size in their first dimensions.
dataset = tf.data.Dataset.from_tensor_slices((x, y, z))  # 使用元组打包x,y,z(★★★★★x, y, z维度必须>=1)

print(dataset)
print(len(dataset))  # 数据集元素的第一个维度
print(dataset._tensors)
print(len(dataset._tensors))
```

```python
for i in dataset:
    print(i)  # 输出数据集(x[i, ...], y[i, ...], z[i, ...])
```

```python
dataset = tf.data.Dataset.from_tensor_slices({"x": x, "y": y, "z": z})  # 使用字典打包x,y
dataset
```

```python
dataset._tensors  # 数据集[x, y, z]
```

```python
for i in dataset:
    print(i)  # 输出{"x": x[i, ...], "y": y[i, ...], "z": z[i, ...]}
```
