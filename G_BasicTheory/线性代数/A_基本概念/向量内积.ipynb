设有$n$维向量

$$
\begin{equation}
\mathbf{x}  =
\begin{bmatrix}
x_{1} \\
x_{2} \\
\vdots \\
x_{n} \\
\end{bmatrix}
\end{equation}
$$

$$
\begin{equation}
\mathbf{y} =
\begin{bmatrix}
y_{1} \\
y_{2} \\
\vdots \\
y_{n} \\
\end{bmatrix}
\end{equation}
$$

令$[\mathbf{x}, \mathbf{y}]=\mathbf{x}^T \mathbf{y} = x_1y_1+x_2y_2 + \cdots, +x_ny_n$称为向量$\mathbf{x}$与$\mathbf{y}$的内积.

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([2, 3, 4, 5, 6])
arr1 @ arr2 # 向量的内积(点积)
```

```python
np.dot(arr1, arr2)
```
