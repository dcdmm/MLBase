## 概率密度函数（Probability Density Function）
根据密度分布函数来计算

```python
from scipy.stats import f

# additional imports for plotting purpose
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import warnings
warnings.filterwarnings('ignore')

style.use('classic')
fig = plt.figure(figsize=(10, 5), dpi=100)
ax  = plt.subplot(111)

# continuous pdf for the plot
x_s = np.arange(0, 3, 0.01)
y_s = f.pdf(x=x_s, dfn=5, dfd=2)
plt.scatter(x_s, y_s, s=3);
```

## Cumulative Probability Density Function¶

```python
from scipy.stats import f

# probability of x less or equal 0.3
print("P(X <=3) = {}".format(f.cdf(x=3, dfn=5, dfd=2)))

# probability of x in [1, 8]
print("P(1 < X <= 8) = {}".format(f.cdf(x=8, dfn=5, dfd=2) - f.cdf(x=1, dfn=5, dfd=2)))
```
