```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### matplotlib实现

```python
groupby = tips["total_bill"].groupby(tips["day"]).mean()
groupby
```

```python
plt.bar(groupby.index, groupby.values)
```

### seaborn参数解析

```python
# Draw a set of vertical bar plots grouped by a categorical variable:
sns.barplot(x="day",  # 分组列为"day"(自动识别)
            y="total_bill",
            data=tips)
```

```python
# Draw a set of horizontal bars
sns.barplot(x="total_bill",
            y="day",
            data=tips)
```

### 参数理解

```python
groupby = tips["total_bill"].groupby([tips["day"], tips["sex"]]).mean()
groupby
```

```python
# Draw a set of vertical bars with nested grouping by a two variables:
sns.barplot(x="day",
            y="total_bill",
            hue="sex",  # 指定第二个分组列为"sex"
            data=tips)
```

```python
# Control bar order by passing an explicit order:
ax = sns.barplot(x="time", y="tip", data=tips,
                 order=["Dinner", "Lunch"])  # 自定义分组顺序
```

```python
# Use median as the estimate of central tendency:
from numpy import median

# 默认estimator=np.mean
sns.barplot(x="day", y="tip", data=tips,
            estimator=median)  # 自定义聚类函数
```

```python
# Show the standard error of the mean with the error bars:
sns.barplot(x="day", y="tip", data=tips,
            ci=31)  # 误差线长度大小(置信区间)
```

```python
# Show standard deviation of observations instead of a confidence interval:
sns.barplot(x="day", y="tip", data=tips,
            ci="sd")
```

```python
# Add “caps” to the error bars:
sns.barplot(x="day", y="tip", data=tips,
            errcolor='green',  # 误差线的颜色
            capsize=.2)  # 误差线的宽度
```

```python
# Use a different color palette for the bars:
sns.barplot(x="day", y="total_bill", data=tips,
            palette="Set1")  # 通过调色盘为不同分组绘制不同颜色
```

```python
# 单一颜色绘制所有列
sns.barplot(x="size", y="total_bill", data=tips,
            color="red")
```

```python
iris = sns.load_dataset("iris")

# 绘制DataFrame的所有列(未指定x,y)
sns.barplot(data=iris, orient="v")
```

```python
sns.barplot(data=iris, orient="h")  # 水平方向

```
