1. Axes.grid

```python
import  numpy as np
import  matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.sin(x), np.cos(x)
fig = plt.figure(figsize=(11, 6))
ax = plt.subplot(2, 1, 1)

ax.plot(c)
ax.plot(s)
ax.grid(b=True, linestyle='--', which='major', color='r', axis='both', linewidth=2) #配置网格线
'''
b:是否显示网格

which : {'major', 'minor', 'both'}
    The grid lines to apply the changes on.grid lines to apply the changes on.

linestyle:网格线型{'-', '--', '-.', ':'}
color:网格线颜色
linewidth:网格线宽度

axis:{'both', 'x', 'y'}
    The axis to apply the changes on
'''

ax.tick_params(axis='y', left=True,right=True,
                color='r', width=4,
                direction='in', length=10) 
ax.tick_params(axis='x', bottom=True,top=True,
                color='g', width=7,
                direction='out', length=14) 
```
