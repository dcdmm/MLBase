```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # 读取图片(RGB模式)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
# 先随机裁取原始图片scale比例的面积
# 再随机将新裁取的图片的长宽比缩放为ratio
# 再将其转化为size大小输出
trans = transforms.RandomResizedCrop(size=300, scale=(0.05, 0.08),
                                     ratio=(1/2, 5/4))
trans(img_resize)
```
