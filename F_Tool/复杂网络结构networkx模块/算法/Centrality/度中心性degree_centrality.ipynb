```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

nx.draw(G)
```

```python
G.degree
```

```python
# The degree centrality for a node v is the fraction of nodes it is connected to.
nx.algorithms.centrality.degree_centrality(G)
```
