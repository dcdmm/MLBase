<font color='red' size=4>定义:</font>      
&emsp;&emsp;设$ X $是连续随机变量,$ X $服从逻辑斯蒂分布是指$X$具有下列分布函数和密度函数:      
$$  F(x) = P(X \leq x) = \frac{1}{1+ e^{-(x-\mu)/ \gamma}} $$     
$$ f(x) = F'(x) = \frac{e^{-(x-\mu)/ \gamma}}{\gamma (1 + e^{-(x-\mu) / \gamma})^2} $$   
式中,$ \mu  $为未知参数,$\gamma > 0 $为形状参数.   
&emsp;&emsp;其分布函数图像是一条$ S $形曲线(sigmoid curve).该曲线以点$ (\mu, \frac{1}{2}) $为中心对称,即满足     
$$ F(-x+u) -\frac{1}{2} = -F(x+ \mu) + \frac{1}{2} $$      
曲线在中心附近增长速度较快,在两端增长速度较慢.形状参数$ \gamma $ 越小,曲线在中心附近增长速度得越快.    


```python
from scipy.stats import logistic
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.style as style
style.use('fivethirtyeight')
```

## 参数$ \mu $对逻辑斯蒂分布的影响

```python
plt.figure(figsize=(20, 6))
x = np.linspace(-10, 10, 1000)
colors = plt.cm.Accent(np.linspace(0, 1, 5))

loc_list = [-6, -3, 0, 3, 6]

ax = plt.subplot(121)
for i in range(len(loc_list)):
    ax.plot(x, logistic.pdf(x, loc=loc_list[i], scale=1), lw=2,
            color=colors[i], label="loc="+str(i))
    ax.axvline(loc_list[i], lw=3, color=colors[i])
    ax.legend()

ax1 = plt.subplot(122)
for j in range(len(loc_list)):
    ax1.plot(x, logistic.cdf(x, loc=loc_list[j], scale=1), color=colors[j],
             lw=2, label="loc="+str(j))
    ax1.axvline(loc_list[j], color=colors[j], lw=3)
    ax1.legend()
```

## 参数 $ \gamma $对逻辑斯蒂分布的影响

```python
plt.figure(figsize=(20, 6))

scale_list = [0.25, 0.5, 1, 2, 4]

ax = plt.subplot(121)
for i in scale_list:
    ax.plot(x, logistic.pdf(x, loc=0, scale=i), lw=2, label="scale="+str(i))
ax.axvline(0, color='r', lw=2)
ax.legend()

ax1 = plt.subplot(122)
for j in scale_list:
    ax1.plot(x, logistic.cdf(x, loc=0, scale=j), lw=2, label="scale="+str(j))
ax1.axvline(0, color='r', lw=2)
ax1.legend()
```
