```python
import  numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import matplotlib.patches
```

```python
fig = plt.figure()
ax = plt.subplot(111)
s = pd.Series(np.random.rand(4)*3, index=list('ABCD'), name='Series')
ax.pie(s, explode=[0.1, 0, 0, 0.3], labels=[None, None, None, None], 
        colors=['r', 'y', 'g', 'pink'], autopct='%.4f%%',
        pctdistance=0.5, shadow=False, startangle=0,
        radius=2, frame=True, labeldistance=1.1, textprops={'color': 'b', 'fontsize':15})
ax.axis('equal') 
```

```python
fig = plt.figure()
ax = plt.subplot(111)
s = pd.Series(np.random.rand(4)*3, index=list('ABCD'), name='Series')
ax.pie(s, explode=[0.1, 0, 0, 0.3], labels=s.index, 
        colors=['r', 'y', 'g', 'pink'], #color, sequence, or sequence of color
        autopct='%.4f%%',
        pctdistance=0.5, shadow=False, startangle=0,
        radius=2, frame=True, labeldistance=1.1, textprops={'color': 'b', 'fontsize':15})
ax.axis('equal') 
my_circle = matplotlib.patches.Circle( (0,0), 1.3, color='white')
'''
classmethod circle(center=(0.0, 0.0), radius=1.0, readonly=False)[source]
Return a Path representing a circle of a given radius and center.

Parameters:	
    center : pair of floats
        The center of the circle. Default (0, 0).

    radius : float
        The radius of the circle. Default is 1.
'''
ax.add_artist(my_circle) #Add any Artist to the axes.
```
