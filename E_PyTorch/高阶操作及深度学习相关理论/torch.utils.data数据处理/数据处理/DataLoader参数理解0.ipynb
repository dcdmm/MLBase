### 基本参数理解

```python
import torch
import torch.utils.data as Data

BATCH_SIZE = 4

x = torch.linspace(0.01, 0.1, 300).reshape(25, 3, 4)
y = torch.linspace(1, 10, 25)

torch_dataset = Data.TensorDataset(x, y)  # 包装数据集
```

```python
def show_batch():
    loader = Data.DataLoader(
        # dataset (Dataset): dataset from which to load the data.
        # dataset类型参考:https://pytorch.org/docs/stable/data.html?highlight=iterabledataset#iterable-style-datasets
        dataset=torch_dataset,
        batch_size=BATCH_SIZE,  # how many samples per batch to load (default: 1)
        drop_last=False,  # 数据集大小不能被batch_size大小整除时,是否舍弃最后一批数据(default:False)
        # 若shuffle=True,则使用的是RandomSampler采样
        shuffle=True,
        #  set to True to have the data reshuffled at every epoch (default: False);内部实现为RandomSampler(dataset, generator=generator)
        generator=torch.Generator().manual_seed(3)  # 功能类似numpy中的random_state,作用于shuffle参数
    )
    print(len(loader))  # 批次数量
    print(loader.dataset)  # 即实参torch_dataset
    print(len(loader.dataset))  # torch_dataset的大小
    print()

    for epoch in range(3):
        for step, batch_data in enumerate(loader):
            print(type(batch_data),  # 列表
                  len(batch_data))
            print(batch_data[0].shape, batch_data[1].shape)  # 第一个维度为批次信息
            print("batch:{}, batch_data:\n{}".format(step, batch_data))  # tensor长度为batch_size
            print(end='\n\n')  # 换行


show_batch()


```
