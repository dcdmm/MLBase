&emsp;&emsp;假设给定训练数据只包含$S$个长度为$T$的观测序列$\{ O_1,O_2,\cdots,O_s\}$而没有对应的状态序列,目标是学习因马可夫模型$\boldsymbol{\lambda}=(A,B,\pi)$的参数.我们将观测序列数据看作观测数据$O$,状态序列数据看作不可观测的隐数据$I$,那么马尔可夫模型事实上是包含一个含有隐变量的概率模型

$$ P(O \mid \boldsymbol{\lambda})=\sum_{I} P(O \mid I, \boldsymbol{\lambda}) P(I \mid \boldsymbol{\lambda}) \tag{1} $$

它的参数学习可以由EM算法实现.

1. 确定完全数据的对数似然函数

   所有的观测数据写成$O=\{ o_1,o_2,\cdots,o_T \}$,所有的隐数据写成$I=\{i_1,i_2,\cdots,i_T \}$,完全数据是$(O,I)=(o_1,o_2,\cdots,o_T,i_1,i_2,\cdots,i_T)$.完全数据的对数似然函数是$\log P(O,I|\boldsymbol{\lambda})$

2. EM算法的E步:求$Q$函数$Q(\boldsymbol{\lambda},\boldsymbol{\overline{\lambda}}^t)$

    $$ Q(\boldsymbol{\lambda},\boldsymbol{\overline{\boldsymbol{\lambda}}}^t)=\sum_{I} \log P(O, I \mid \boldsymbol{\lambda}) P(O, I \mid \bar{\boldsymbol{\lambda}}^t) \qquad \text{注:由EM算法Q函数的定义}   $$

其中,$\bar{\boldsymbol{\lambda}}^t$是隐马尔可夫模型参数的当前估计值,$\boldsymbol{\lambda}$是要极大化的隐马尔科夫模型参数.

又有,

$$
\begin{aligned}
P(O, I \mid \boldsymbol{\lambda})&=P(O|I,\boldsymbol{\lambda})P(I|\boldsymbol{\lambda})
\\
&=b_{i_{1}}\left(o_{1}\right)b_{i_{2}}\left(o_{2}\right) \cdots,b_{i_{T}}\left(o_{T}\right) P(I|\boldsymbol{\lambda})
\\
&=\pi_{i_{1}} b_{i_{1}}\left(o_{1}\right) a_{i_{1} i_{2}} b_{i_{2}}\left(o_{2}\right) \cdots a_{i_{T-1} i_{T}} b_{i_{T}}\left(o_{T}\right)
\end{aligned}
$$

于是$Q(\boldsymbol{\lambda},\boldsymbol{\overline{\boldsymbol{\lambda}}}^t)$可以写成:

$$ Q(\boldsymbol{\lambda},\boldsymbol{\overline{\boldsymbol{\lambda}}}^t)=\sum_{I} \log \pi_{i_{1}} P(O, I \mid \overline{\boldsymbol{\lambda}}^t)+\sum_{I}\left(\sum_{t=1}^{T-1} \log a_{i_{t} i_{t+1}}\right) P(O, I \mid \overline{\boldsymbol{\lambda}}^t) + \sum_{I} \left( \sum_{t=1}^T \log b_{i_t} \right)  P(O, I \mid \overline{\boldsymbol{\lambda}}^t) \tag{2} $$

式中求和都是对所有数据的序列总长度$T$进行的.

3. EM算法的M步:极大化$Q$函数$Q(\boldsymbol{\lambda},\boldsymbol{\overline{\boldsymbol{\lambda}}}^t) $求模型参数$\boldsymbol{\bar{\lambda}}^{t+1}=(A^{t+1},B^{t+1},\boldsymbol{\pi}^{t+1})$

   由于要极大化的参数在式(2)中单独地出现在3个项中,所以只需对各项分别极大化.

   1. 式(2)的第一项可以写成
      $$ \sum_{I} \log \pi_{i_{1}} P(O, I \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t)=\sum_{i=1}^{N} \log \pi_{i} P\left(O, i_{1}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right) $$

    易知$\pi_i$满足约束条件$\sum_{i=1}^N \pi_i = 1$,利用拉格朗日乘子法,写出拉格朗日函数:

    $$ \sum_{i=1}^{N} \log \pi_{i} P\left(O, i_{1}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t \right)+\gamma\left(\sum_{i=1}^{N} \pi_{i}-1\right) $$

    对其求偏导数并令结果为0

    $$ \frac{\partial}{\partial \pi_{i}}\left[\sum_{i=1}^{N} \log \pi_{i} P\left(O, i_{1}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)+\gamma\left(\sum_{i=1}^{N} \pi_{i}-1\right)\right]=0 $$

    得
    $$  P\left(O, i_{1}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)+\gamma \pi_{i}=0 \tag{3} $$

    对$i$求和得到$\gamma$
    $$ \gamma = -P(O|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t) $$

    代入式(3)可得

    $$ \pi_{i}=\frac{P\left(O, i_{1}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)}{P(O \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t)} $$

    即
    $$ \pi_{i}^{t+1}=\frac{P\left(O, i_{1}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)}{P(O \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t)} \tag{4} $$

 2. 式(2)的第二项可以写成
    $$ \sum_{I}\left(\sum_{t=1}^{T-1} \log a_{i_{t} i_{t+1}}\right) P(O, I \mid  \boldsymbol{\overline{\boldsymbol{\lambda}}}^t)=\sum_{i=1}^{N} \sum_{j=1}^{N} \sum_{t=1}^{T-1} \log a_{i j} P\left(O, i_{t}=i, i_{t+1}=j \mid  \boldsymbol{\overline{\boldsymbol{\lambda}}}^t \right) $$

    与上类似,应用具有约束条件$\sum_{j=1}^N a_{ij} =1$的拉格朗日乘子法可以求出

    $$ a_{i j}^{t+1}=\frac{\sum_{t=1}^{T-1} P\left(O, i_{t}=i, i_{t+1}=j \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)}{\sum_{t=1}^{T-1} P\left(O, i_{t}=i \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)} \tag{5} $$

 3. 对式(2)的第三项为
    $$ \sum_{I}\left(\sum_{t=1}^{T} \log b_{i_{t}}\left(o_{t}\right)\right) P(O, I \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t)=\sum_{j=1}^{N} \sum_{t=1}^{T} \log b_{j}\left(o_{t}\right) P\left(O, i_{t}=j \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t \right) $$

    同样用拉格朗日乘子法,约束条件是$\sum_{k=1}^M b_j(k) = 1$.注意,只有在$o_t=v_k$时,$b_j(o_t)$对$b_j(k)$的偏导数才不为0,以$\mathbb{I}(o_t = v_k)$表示.求得

    $$ b_{j}(k)^{(t+1)}=\frac{\sum_{t=1}^{T} P\left(O, i_{t}=j \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right) I\left(o_{t}=v_{k}\right)}{\sum_{t=1}^{T} P\left(O, i_{t}=j \mid \boldsymbol{\overline{\boldsymbol{\lambda}}}^t\right)} \tag{6} $$


&emsp;&emsp;将式(4)式(5)式(6)中的概率分别用$\gamma_t(i),\xi_t(i,j)$表示,则可将相应的公式写成:

$$ \pi_i^{t+1} = \gamma_{1|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(i) \tag{7} $$

$$ a_{i j}^{t+1}=\frac{\sum_{t=1}^{T-1} \xi_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(i, j)}{\sum_{t=1}^{T-1} \gamma_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(i)} \tag{8} $$

$$ b_{j}(k)^{(t+1)}=\frac{\sum_{t=1, o_{t}=v_{k}}^{T} \gamma_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(j)}{\sum_{t=1}^{T} \gamma_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(j)} \tag{9} $$

式(7)式(8)式(9)就是Baum-Welch算法,它是EM算法在隐马尔科夫模型学习中的具体实现,由Baum和Welch提出.


<font color='red' size=4>算法(Baum-Welch算法)</font>

输入:观测数据$O=(o_1,o_2,\cdots,o_T)$

输出:因马可夫模型参数

1. 初始化.对$t=0$,选取$a_{ij}^0,b_j(k)^0$,得到模型$\boldsymbol{\bar{\lambda}}^0 = (A^0, B^0, \boldsymbol{\pi}^0)$

2. 递推,对$t=1,2,\cdots$
    $$ \pi_i^{t+1} = \gamma_{1|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(i) \tag{7} $$

    $$ a_{i j}^{t+1}=\frac{\sum_{t=1}^{T-1} \xi_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(i, j)}{\sum_{t=1}^{T-1} \gamma_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(i)} \tag{8} $$

    $$ b_{j}(k)^{(t+1)}=\frac{\sum_{t=1, o_{t}=v_{k}}^{T} \gamma_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(j)}{\sum_{t=1}^{T} \gamma_{t|\boldsymbol{\overline{\boldsymbol{\lambda}}}^t}(j)} \tag{9} $$

    右端各值按观测$O=(o_1,o_2,\cdots,oT)$和模型$\boldsymbol{\bar{\lambda}}^{t}=(A^{t},B^{t},\boldsymbol{\pi}^{t})$计算.

3. 终止,得到模型参数$\boldsymbol{\bar{\lambda}}^{t+1}=(A^{t+1},B^{t+1},\boldsymbol{\pi}^{t+1})$
