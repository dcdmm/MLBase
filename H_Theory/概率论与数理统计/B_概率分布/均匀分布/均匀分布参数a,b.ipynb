```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('fivethirtyeight')
fig = plt.figure(figsize=(8, 4), dpi=100)
ax  = plt.subplot(111)

# PDF loc=0, scale=1, :a=0, b-a=1
ax.plot(np.linspace(-8, 8, 100), 
         stats.uniform.pdf(np.linspace(-8, 8, 100), loc=0, scale=1),)
ax.fill_between(np.linspace(-8, 8, 100),
                 stats.uniform.pdf(np.linspace(-8, 8, 100), loc=0, scale=1),
                 alpha=.15)

# PDF loc=0, scale=3, :a=0, b-a=3
ax.plot(np.linspace(-8, 8, 100), 
         stats.uniform.pdf(np.linspace(-8, 8, 100), loc=0, scale=3))
ax.fill_between(np.linspace(-8, 8, 100),
                 stats.uniform.pdf(np.linspace(-8, 8, 100), loc=0, scale=3),
                 alpha=.15)

# PDF loc=-3, scale=3, :a=-3,b-a=3
ax.plot(np.linspace(-8, 8, 100), 
         stats.uniform.pdf(np.linspace(-8, 8, 100), loc=-3, scale=3))
ax.fill_between(np.linspace(-8, 8, 100),
                 stats.uniform.pdf(np.linspace(-8, 8, 100),loc=-3, scale=3),
                 alpha=.15)

ax.text(x=-1, y=.65, s="a=0, b=1", rotation=65, alpha=.75, weight="bold", color="#008fd5")
ax.text(x=1, y=.65, s="a=0, b=2", rotation=65, alpha=.75, weight="bold", color="red")
ax.text(x=-3, y=.65, s="a=-3, b=0", rotation=65, alpha=.75, weight="bold", color="#e5ae38")


ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

plt.text(x = -5, y = 1.1, s = "Nniform Distribution - loc and scale",
               fontsize = 26, weight = 'bold', alpha = .75)
```
