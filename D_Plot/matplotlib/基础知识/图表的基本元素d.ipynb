1. Axes.tick_params

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 100, 1)
s = np.sin(0.1 * np.pi * t) * np.exp(-t * 0.01)
fig = plt.figure(figsize=(10, 7))
ax = plt.subplot(111)
ax.plot(t, s, marker='*', linestyle='--')

ax.tick_params(axis='y', left=True, right=True,
               color='r', width=4, length=10,
               direction='in', which='both',
               labelsize=17, labelcolor='red', labelrotation=30)
ax.tick_params(axis='x', bottom=True, top=True,
               color='g', width=7, length=5,
               direction='out', which='both',  # length=0(即刻度点长度为0),则刻度点消失
               labelsize=25, labelcolor='yellow', labelrotation=90)
'''
axis : {'x', 'y', 'both'}, optional
    Which axis to apply the parameters to.
    
which : {'major', 'minor', 'both'}
    Default is 'major'; apply arguments to which ticks.
    
direction : {'in', 'out', 'inout'}
    Puts ticks inside the axes, outside the axes, or both.

labelsize : float or str
    Tick label font size in points or as a string (e.g., 'large').

labelcolor : color
    Tick label color; mpl color spec.

labelrotation : float
    Tick label rotation

left:左刻度是否显示
right:右刻度是否显示
top:上刻度是否显示
bottom:下刻度是否显示
color:刻度颜色
width:刻度宽度
length:刻度长度
'''
```
