```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Plot a univariate distribution along the x axis:
sns.kdeplot(data=tips, x="total_bill")
```

```python
# Flip the plot by assigning the data variable to the y axis:
sns.kdeplot(data=tips, y="total_bill")
```

```python
# If True, fill in the area under univariate density curves or between bivariate contours.
sns.kdeplot(data=tips, x="total_bill", fill=True)  # 默认fill=False
```

```python
# Use less smoothing:
sns.kdeplot(data=tips, x="total_bill",
            bw_adjust=.2)  # 越大曲线越平滑
```

```python
# Plot conditional distributions with hue mapping of a second variable:
sns.kdeplot(data=tips, x="total_bill",
            hue="time")  # 使用列"time"进行分组
```

```python
# 累计概率分布
sns.kdeplot(
    data=tips, x="total_bill", hue="time",
    # If True, estimate a cumulative distribution function.
    cumulative=True,
    # If True, scale each conditional density by the number of observations such that the total area under all densities sums to 1. Otherwise, normalize each density independently.
    common_norm=False)
```

```python
geyser = sns.load_dataset("geyser")

# Plot a bivariate distribution:
sns.kdeplot(data=geyser, x="waiting", y="duration")
```

```python
sns.kdeplot(data=geyser, x="waiting", y="duration",
            hue="kind")
```

```python
sns.kdeplot(data=geyser, x="waiting", y="duration",
            hue="kind",
            fill=True)
```

```python
sns.kdeplot(data=geyser, x="waiting", y="duration",
            hue="kind",
            fill=True,
            palette='Set1')  # 通过调色盘为不同分组绘制不同颜色
```

```python
iris = sns.load_dataset("iris")

# 绘制DataFrame的所有列(未指定x,y)
sns.kdeplot(data=iris)

```
