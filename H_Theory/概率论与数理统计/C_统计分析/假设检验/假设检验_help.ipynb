## 似然比检验
<font color='red' size=4>定义:</font>   
&emsp;&emsp;关于检验$ H_0 : \theta \in \Theta_0 $对$ H_1 : \theta \in \Theta_1 $的似然比检验统计量是:  
$$ \lambda(x) = \frac{\sup \limits _{\Theta_0} L(\theta|x) }{\sup \limits _{\Theta} L(\theta|x) }  $$   
任何一个拒绝区域的形式为$ \{ x: \lambda(x) \leq c \} $的检验都叫做似然比检验(简记LRT).这里c是任意一个满足$ 0\leq c \leq 1 $的数    
&emsp;&emsp;$ \lambda(x) $ 的分子就是观测样本出现的最大概率,这里最大是指对参数取遍原假设范围的计算而言,而$ \lambda(x) $ 的分母则是
取遍所有可能的参数时观测样本出现的最大概率.如果存在备择假设中的参数值,使得样本出现的可能性比原假设下的参数对应的可能性更大的话,这两个最大
值之比就小.在这种情况下LRT准则就说$ H_0 $应该被拒绝而接受 $ H_1 $ 为真.      
&emsp;&emsp;设$ \theta $存在一个MLE$ \hat{\theta}, \hat{\theta} $ 是通过做$ L(\theta|x) $的无限制最大化得到的.我们也可以假定
$ \Theta_0$ 是参数空间而考虑$ \theta $的MLE,记为$ \hat{\theta_0}$,通过做有限制的最大化得到它.即$ \hat{\theta_0} $是$ \theta \in \Theta_0 $ 
中最大化$ L(\theta|x)$的 $ \theta $值.于是,LRT统计量就是:    
$$ \lambda(x) = \frac{L(\hat{\theta_0}|x) }{L(\hat{\theta}|x) }  $$   


## 功效函数,检验水平
&emsp;&emsp;一个拒绝区域为$ R$的假设检验的功效函数是由$ \beta(\theta) = P_{\theta}(X \in R ) $所定义的函数    
\begin{equation}
P_{\theta}(X \in R)=\begin{cases}
		犯第一类错误的概率, & \text{if} \quad \theta \in  \Theta_{0} \\
        1 - 犯第二类错误的概率, & \text{if} \quad  \theta \in  \Theta_{0}^C
     \end{cases}
\end{equation}  
<font color='red' size=4>定义:</font>   
&emsp;&emsp;设$ \varphi$是假设检验问题的一个检验,而$ 0  \leq \alpha  \leq 1$.            
&emsp;&emsp;若$ \varphi $犯第一类错误的概率总不超过$ \alpha$(不论$ \theta $ 在 $ \Theta$ 内取何值,总不超过$ \alpha$),则称
$ \alpha $是检验$ \varphi $ 的一个水平,而 $ \varphi $ 称为水平$ \alpha $ 检验.   
&emsp;&emsp;易知$ \varphi $ 为水平$ \alpha$ 检验的充要条件是   
$$  \beta_{\varphi}(\theta) \leq \alpha  (\theta \in \Theta_0)$$    
&emsp;&emsp;按这个定义,一个检验的水平不唯一:若$ \alpha $ 是检验 $ \varphi $ 的水平,而$ \alpha < \alpha ' \leq 1 $,则
$ \alpha' $也是检验 $ \varphi$的水平,为免除这种不方便,有时称一个检验的最小水平为其真实水平.显然      
$$ \varphi $$的真实水平 $$ =\sup\{ \beta_{\varphi}(\theta) | \theta \in \Theta_{0} \}  $$


## 常见检验统计量
 ###  总体均值的检验(大样本)
* 假设条件
	* 大样本
* 使用Z检验统计量
	* $ \sigma^2 已知:  z =  \frac{ \bar{x} - u_{0} }{ \sigma / \sqrt{n} } \sim N(0,1) $
	* $ \sigma^2 未知:  z =  \frac{ \bar{x} - u_{0} }{ s  / \sqrt{n} } \sim N(0,1) $

### 总体均值的检验(正态,小样本)
* 假设条件
	* 总体服从正态分布
	* 小样本
* 检验统计量
	* $ \sigma^2 已知:  z =  \frac{ \bar{x} - u_{0} }{ \sigma / \sqrt{n} } \sim N(0,1) $
	* $ \sigma^2 未知:  t =  \frac{ \bar{x} - u_{0} }{ s  / \sqrt{n} } \sim t(n-1) $

### 总体方差的检验
* 假设条件
	* 总体随机变量服从正态分布,并且方差等于$ \sigma^2 $
* 检验统计量
	* ${\chi ^2} = \frac{{(n - 1){s^2}}}{{\sigma^2}} \sim  {\chi ^2}(n - 1)$&emsp;$\sigma^2为假设的总体方差$

### 总体比例检测
* 假设条件
	* 总体服从二项分布
	* 可用正态分布来近似$(np,n(1-p)都大于等于5)$
* 使用 Z 检验统计量 
	* $z = \frac{{\hat{p} - {p}}}{{\sqrt {\frac{{{p}(1 - {p})}}{n}} }} \sim N(0,1)$&emsp;$p为假设的总体比例$

### 两总体均值差的推断(大样本)
* 假设条件
	* 每组观察样本必须是从相同总体中独立,随机地抽取(独立性假定)
	* 被比较的两个观察样本相互之间必须保持独立性(样本间的独立性假定)
	* $ n_{1}, n_{2}均为$大样本
* 检验统计量
	* $ \sigma_{1}^2, \sigma_{2}^2 已知: z = \frac{{({{\bar x}_1} - {{\bar x}_2}) - ({\mu _1} - {\mu _2})}}{{\sqrt {\frac{{\sigma _1^2}}{{{n_1}}} + \frac{{\sigma _2^2}}{{{n_2}}}} }} \sim N(0,1) $ 
	* $ \sigma_{1}^2, \sigma_{2}^2 未知: z = \frac{{({{\bar x}_1} - {{\bar x}_2}) - ({\mu _1} - {\mu _2})}}{{\sqrt {\frac{{s _1^2}}{{{n_1}}} + \frac{{s _2^2}}{{{n_2}}}} }} \sim N(0,1) $ 

### 两总体均值差的推断(正态,两方差已知)
* 假设条件
	* 每组观察样本必须是从相同总体中独立,随机地抽取(独立性假定)
	* 被比较的两个观察样本相互之间必须保持独立性(样本间的独立性假定)
	* 每个样本来自的总体呈正态分布(正态性假定)
	* $ \sigma_{1}^2, \sigma_{2}^2 $ 已知
	
* 检验统计量
	* $ z = \frac{{({{\bar x}_1} - {{\bar x}_2}) - ({\mu _1} - {\mu _2})}}{{\sqrt {\frac{{\sigma _1^2}}{{{n_1}}} + \frac{{\sigma _2^2}}{{{n_2}}}} }} \sim N(0,1) $ 

### 两总体均值差的推断(正态,小样本,两方差未知且相等)
* 假设条件
	* 每组观察样本必须是从相同总体中独立,随机地抽取(独立性假定)
	* 被比较的两个观察样本相互之间必须保持独立性(样本间的独立性假定)
	* 每个样本来自的总体呈正态分布(正态性假定)
	* $ \sigma_{1}^2, \sigma_{2}^2 未知但相等,即 \sigma_{1}^2 = \sigma_{2}^2$
* 检验统计量
	* $t = \frac{{({{\bar x}_1} - {{\bar x}_2}) - ({\mu _1} - {\mu _2})}}{{{s_p}\sqrt {\frac{1}{{{n_1}}} + \frac{1}{{{n_2}}}} }}$
&emsp;其中:$s_p^2 = \frac{{({n_1} - 1)s_1^2 + ({n_2} - 1)s_2^2}}{{{n_1} + {n_2} - 2}}$
&emsp;自由度:${n_1} + {n_2} - 2$

### 两总体均值差的推断(正态,小样本,两方差未知且不等)
* 假设条件
	* 每组观察样本必须是从相同总体中独立,随机地抽取(独立性假定)
	* 被比较的两个观察样本相互之间必须保持独立性(样本间的独立性假定)
	* 每个样本来自的总体呈正态分布(正态性假定)
	* $ \sigma_{1}^2, \sigma_{2}^2 未知且不相等,即 \sigma_{1}^2 \neq \sigma_{2}^2$
* 检验统计量
	* $t = \frac{{({{\bar x}_1} - {{\bar x}_2}) - ({\mu _1} - {\mu _2})}}{{\sqrt {\frac{{s_1^2}}{{{n_1}}} + \frac{{s_2^2}}{{{n_2}}}} }}$
&emsp;自由度:$ \frac{ \Big (  \frac{s_{1}^2}{n_{1}}  +  \frac{s_{2}^2}{n_{2}}  \Big )^2   } {  \frac{ (s_{1}^2 / n_{1} )^2}{n_{1} } + \frac{ (s_{2}^2 / n_{2} )^2}{n_{2}}  }  $

### 两总体均值差的推断(配对试验,基本上和单样本均值t检验相对应)
* 假设:
	* 观察数据成对
	* 配对资料差的总体服从正态分布 
	* 成对数据之差要求相互独立
* 检验统计量
	* $ t= \frac{ \bar{d} - d_{0} }{ s_{d} / \sqrt{ n_{d} }}  \sim t(n-1)$
&emsp;样本差值均值:$\bar d = \frac{{\sum\limits_{i = 1}^n {{d_i}} }}{{{n_d}}}$
&emsp;样本差值标准差:${s_d} = \sqrt {\frac{{\sum\limits_{i = 1}^n {{{({d_i} - \bar d)}^2}} }}{{{n_d} - 1}}} $

### 两个总体比例之差的检验
* 假设条件
    * 每组观察样本必须是从相同总体中独立,随机地抽取(独立性假定)
	* 两个总体都服从二项分布
	* 可以用正态分布来近似($ n_{1}p_{1},n_{1}(1-p_{1}),n_{2}p_{2},n_{2}(1-p_{2})都大于等于5 $)
* 检验统计量
	* 检测$ H_{0} : p_{1} - p_{0} = 0$ &emsp; $z = \frac{{{\hat{p}_1} - {\hat{p}_2}}}{{\sqrt {\hat{p}(1 - \hat{p})\left( {\frac{1}{{{n_1}}} + \frac{1}{{{n_2}}}} \right)} }}$
&emsp;$ \hat{p} = \frac{{{x_1} + {x_2}}}{{{n_1} + {n_2}}} = \frac{{{p_1}{n_1} + {p_2}{n_2}}}{{{n_{_1}} + {n_2}}}$
	* 检测$ H_{0} : p_{1} - p_{0} = D(D \neq 0) $ &emsp; $ z = \frac{{({\hat{p}_1} - {\hat{p}_2}) - {D}}}{{\sqrt {\frac{{{\hat{p}_1}(1 - {\hat{p}_1})}}{{{n_1}}} + \frac{{{\hat{p}_2}(1 - {\hat{p}_2})}}{{{n_2}}}} }}$

### 两个总体方差比的检验
* 假设条件
	* 两个总体都服从正态分布，且方差相等
	* 两个独立的随机样本
* 检验统计量
	* $F = \frac{{s_1^2}}{{s_2^2}} \sim F({n_1} - 1,{n_2} - 1)\quad $
