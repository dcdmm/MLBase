```python
# sin
import  numpy as np
import  matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 201)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```

```python
# cos
x = np.linspace(-np.pi, np.pi, 201)
plt.plot(x, np.cos(x))
plt.xlabel('Angle [rad]')
plt.ylabel('cos(x)')
plt.axis('tight')
plt.show()
```

```python
# tan
x = np.linspace(-np.pi/2+0.2 , np.pi/2-0.2, 201)
plt.plot(x, np.tan(x))
plt.xlabel('Angle [rad]')
plt.ylabel('tan(x)')
plt.axis('tight')
plt.show()
```

```python
# numpy中的数学函数既可可作用与标量,也可以作用于向量或矩阵
np.sin(1) # 标量
```

```python
np.sin([1, 2, 3, 4]) # 向量
```

```python
np.sin([[1, 2, 3, 4],
        [5, 6, 7, 8]]) # 矩阵
```
