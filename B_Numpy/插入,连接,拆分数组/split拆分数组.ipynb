```python
import numpy as np
```

```python
arr1 = np.arange(40).reshape((4, 10))
arr1
```

```python
a, b, c, d = np.split(arr1, [1, 3, 7], axis=1)
print(a)  # 即arr1[:, 0:1]
print(b)  # 即arr1[:, 1:3]
print(c)  # 即arr1[:, 3:7]
print(d)  # 即arr1[:, 7:]
```

```python
m, n, z = np.split(arr1, [2, 3], axis=0)
print(m)  # 即arr1[0:2, :]
print(n)  # 即arr1[2:3, :]
print(z)  # 即arr1[3:, :]
```

```python
np.split(arr1, 4, axis=0)  # 拆分成大小均等的快
```
