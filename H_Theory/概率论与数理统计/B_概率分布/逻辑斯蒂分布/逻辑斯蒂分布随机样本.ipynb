```python
import numpy as np
from scipy.stats import logistic
import matplotlib.pyplot as plt

# draw a single sample
np.random.seed(42)
print(logistic.rvs(loc=0, scale=1), end="\n\n")

# draw 10 samples
print(logistic.rvs(loc=3, scale=0.1, size=10), end="\n\n")
```

```python
r = logistic.rvs(size=1000)
plt.hist(r, bins=20, density=True, histtype='stepfilled',
         alpha=0.9, label='logistic rvs')
x = np.linspace(logistic.ppf(0.001),
                logistic.ppf(0.999), 1000)
plt.plot(x, logistic.pdf(x),
       'r-', lw=3, alpha=0.8, label='logistic pdf')
plt.legend(loc='best')
```
