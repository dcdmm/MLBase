&emsp;&emsp;主成分分析的主要目的是降维,所以一般选择$k(k  \ll m)$个主成分(线性无关变量)来代替$m$个原有变量(线性相关变量),使问题得以简化,为能保证保留原有变量的大部分信息,这里所说的信息是指原有变量的方差.为此,先给出一个定理,说明选择$k$个主成分是最优选择.

<font color='red' size=4>定理:</font>对任意正整数$q,1 \leq q \leq m$,考虑正交线性变换

$$ Y = B^T X $$

其中$Y$是$q$维随机向量,$B^T$是$q \times m$矩阵,则$Y$的协方差矩阵为

$$ Y= B^T \Sigma B $$

则$\Sigma_{Y}$的迹$\mathrm{tr}(\Sigma_{Y})$在$B=A_q$时取得最大值,其中矩阵$A_q$由正交矩阵$A$的前$q$列组成.

&emsp;&emsp;<font color='red' size=4>证明:</font>令$\boldsymbol{\beta}_k$是$B$的第$k$列,由于正交矩阵$A$的列构成$m$维空间的基,所以$\boldsymbol{\beta}_k$可以由$A$的列表示,即

$$ \boldsymbol{\beta}_k = \sum_{j=1}^m c_{jk} \boldsymbol{\alpha}_j, \quad k=1,2,\cdots,q $$

等价地

$$ B = AC \tag{1} $$

其中$C$是$m \times q$矩阵,其第$j$行第$k$个元素为$c_{jk}$.

&emsp;&emsp;首先,

$$ B^T \Sigma B = C^T A^T \Sigma AC = C^T \Lambda C = \sum_{j=1}^m \lambda_j \boldsymbol{c}_j \boldsymbol{c}_j^T $$

其中$\boldsymbol{c}_j^T $是$C$的第$j$行.因此

\begin{align}
\mathrm{tr}(B^t \Sigma B) &= \sum_{j=1}^{m} \lambda_j \mathrm{tr}(\boldsymbol{c}_j \boldsymbol{c}_j^T)  \\
						  &= \sum_{j=1}^m \lambda_j \mathrm{tr}(\boldsymbol{c}_j^T \boldsymbol{c}_j)  \quad \text{由矩阵迹的性质}\\
						  &= \sum_{j=1}^m \lambda_j \boldsymbol{c}_j^T \boldsymbol{c}_j \\
						  &= \sum_{j=1}^m \sum_{k=1}^q \lambda_j c_{jk}^2 \\
						  \tag{2}
\end{align}

&emsp;&emsp;其次,由式(1)及$A$的正交性知

$$ C = A^T B $$
由于$A$是正交的,$B$的列也是正交的,所以

$$ C^T C = B^T AA^T B = B^T B = I_q $$

即$C$的列也是正交的.于是

\begin{align}
\mathrm{tr}(C^T C) = \mathrm{tr}(I_q) \\
\sum_{j=1}^m \sum_{k=1}^q c_{jk}^2 = q \tag{3}
\end{align}

&emsp;&emsp;这样,矩阵$C$可以认为是某个$m$阶正交矩阵$D$的前$q$列.正交矩阵$D$的行也正交,所以满足

$$ \boldsymbol{d}_j^T \boldsymbol{d}_j=1,\quad j=1,2,\cdots,m $$

其中$\boldsymbol{d}_j^T$是$D$的第$j$行.由于矩阵$D$包含矩阵$C$的行的前$q$个元素,所以

$$ \boldsymbol{c}_j^T \boldsymbol{c}_j \leq 1,\quad j=1,2,\cdots,m $$

即

$$ \sum_{k=1}^q c_{jk}^2 \leq1, j=1,2,\cdots,m \tag{4} $$

&emsp;&emsp;注意到在式(2)中$\sum_{k=1}^q c_{jk}^2 $是$\lambda_j$的系数,由式(3)这些系数之和是$q$,且由式(4)知这些系数小于等于1.因为$\lambda_1 \geq \lambda_2 \geq \cdots \geq \lambda_q \cdots \lambda_m$,显然,当能找到$c_{jk}$使得

\begin{equation}
\sum_{k=1}^q c_{jk}^2=\begin{cases}
		1, & j=1,\cdots,q \\
        0, & j = q+1,\cdots,m
     \end{cases}
     \tag{5}
\end{equation}

时,$\sum_{j=1}^m \left(  \sum_{k=1}^q c_{jk}^2 \right) \lambda_j$最大.而当$B=A_q$时,满足式(5).所以,当$B=A_q$,$\mathrm{\Sigma_Y}$达到最大值.

&emsp;&emsp;上述定理表明,当$X$的线性变换$Y$在$B=A_q$时,其协方差矩阵$\Sigma_Y$的迹$\mathrm{tr}(\Sigma_Y)$取得最大值,这就是说,当取$A$的前$q$列取$X$的前$q$个主成分时,能够最大限度地保留原有变量方差的信息.



&emsp;&emsp;<font color='red' size=4>定义:</font>第$k$主成分$Y_k$的方差贡献率定义为$Y_k$的方差与所有方差之和的比,记作$\eta_k$

$$ \eta_k = \frac{\lambda_k}{ \sum_{i=1}^m \lambda_i} $$

$k$个主成分$Y_1,Y_2,\cdots,Y_k$的累计方差贡献率定义为$k$个方差之和与所有方差之和的比

$$ \sum_{i=1}^k \eta_i = \frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^m \lambda_i} $$

&emsp;&emsp;通常取$k$使得累计方差贡献率达到规定的百分比以上,例如$70\% \sim 80\%$.累计方差贡献率反映了主成分保留信息的比例,但它不能反映对某个原有变量$X_i$保留信息的比例,这时通常利用$k$个主成分$Y_1,Y_2,\cdots,Y_k$对原有变量$X_i$的贡献率.



&emsp;&emsp;<font color='red' size=4>定义:</font>$k$个主成分$Y_1,Y_2,\cdots,Y_k$对原有变量$X_i$的贡献率定义为$X_i$与$(Y_1,Y_2,\cdots,Y_k)$的相关系数的平方,记作$v_i$

$$ v_i = \rho^2(X_i,(Y_1,Y_2,\cdots,Y_k)) $$

计算公式如下:

$$ v_i = \rho^2(X_i,(Y_1,Y_2,\cdots,Y_k)) = \sum_{j=1}^k \rho^2(X_i, Y_i)=\sum_{j=1}^k \frac{\lambda_j \alpha_{ij}^2}{\sigma_{ii}} $$
