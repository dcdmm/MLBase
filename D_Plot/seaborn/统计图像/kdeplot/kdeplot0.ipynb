```python
import numpy as np
import seaborn as sns

np.random.seed(10)
sns.set(color_codes=True)
```

```python
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov,
                                     size=50).T  # Draw random samples from a multivariate normal distribution.
```

```python
ax = sns.kdeplot(x)  # pandas.DataFrame or numpy.ndarray
```

```python
# Shade under the density curve and use a different color:
ax = sns.kdeplot(x, shade=True, color="r")
```

```python
# Plot a bivariate density:
ax = sns.kdeplot(x, y)
```

```python
# Use filled contours:
ax = sns.kdeplot(x, y, shade=True)
```

```python
# Use more contour levels and a different color palette:
ax = sns.kdeplot(x, y, n_levels=30, cmap="Set1")
```

```python
# Use a narrower bandwith:
ax = sns.kdeplot(x, bw=.15)
```

```python
#Plot the density on the vertical axis:
ax = sns.kdeplot(y, vertical=True)
```

```python
#Add a colorbar for the contours:
ax = sns.kdeplot(x, y, shade=True, cbar=True)
```

```python
# Plot two shaded bivariate densities:
iris = sns.load_dataset("iris")

iris['sepal_length'].name
```

```python
# 坐标轴名等于iris['sepal_length'].name
sns.kdeplot(iris['sepal_length'])
```

```python
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
ax = sns.kdeplot(x=setosa.sepal_width, y=setosa.sepal_length,
                 cmap="Reds", cbar=True, shade=True)
ax = sns.kdeplot(x=virginica.sepal_width, y=virginica.sepal_length,
                 cmap="Blues", cbar=True, shade=True,
                 cbar_kws={'orientation': 'horizontal'})
sns.rugplot(x=setosa.sepal_width)
sns.rugplot(y=setosa.sepal_length)
```
