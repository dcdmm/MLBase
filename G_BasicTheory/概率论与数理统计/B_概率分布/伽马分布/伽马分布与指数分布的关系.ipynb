<font color='red' size=4>定义:</font>  
&emsp;&emsp;已知伽马分布的概率密度函数为: 
$$ f(x|\alpha ,\beta) = \frac{1}{\Gamma(\alpha)\beta^{\alpha}} x^{\alpha -1} e^{-x/\beta}, 0<x<+\infty,\alpha > 0, \beta >0$$  
若令$ \alpha = 1 $ ,此时概率密度函数为:  
$$ f(x|\beta) = \frac{1}{\beta} e^{-x/\beta}, 0<x<\infty, \beta > 0 $$  
此即尺度参数为$ \beta $的指数概率密度函数     


* 期望和方差:$$ EX=\beta, VarX=\beta^2 $$  
* 矩母函数:$$ M_X(t)  = \frac{1}{1 - \beta t}, t<1/ \beta $$  

<font color='red' size=5>性质:</font>  
&emsp;&emsp;指数函数也具有几何分布的"无记忆性"------若$ x $ 服从参数为$ \beta$ 的指数分布,则对任意$ s>t \geq 0 $,有  
$$ P(X>s| X>t) = P(X>s-t) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
from scipy.stats import expon

# PLOTTING CONFIG
%matplotlib inline
style.use('fivethirtyeight')


class Exponential:
    """手工实现指数分布"""
    
    def __init__(self, x, b):
        self.x = x
        self.b = b
        
    def pdf(self):
        part0 = 1 / self.b
        part1 = np.exp(-self.x / self.b)
        return part0 * part1 
```

## 参数$ \beta $对指数分布的影响

```python
x = np.linspace(0, 10, 100)
b = np.linspace(1, 5, 5)

plt.figure(figsize=(10, 7), dpi=100)
ax1 = plt.subplot(111)
for i in b:
    gam = Exponential(x,  i)
    show = r'$ \beta  $=' + str(i)
    ax1.plot(x, gam.pdf(), label=show)
    ax1.legend()
```
