```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('white')

fig = plt.figure(figsize=(10, 15))
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
plt.subplots_adjust(hspace=0.3)


def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)


ax1 = fig.add_subplot(311)
sinplot()

ax2 = fig.add_subplot(312)
sns.violinplot(data=data)

ax3 = fig.add_subplot(313)
sns.boxplot(data=data, palette='deep')

fig1 = plt.figure(figsize=(10, 5))
ax11 = plt.subplot(111)
sinplot()

# Remove the top and right spines from plot(s)
sns.despine(fig=fig, bottom=True, top=True)  # ,
sns.despine(fig=fig1, left=True, right=False, offset=-30)  # ,(right)
'''
fig : matplotlib figure, optional
    Figure to despine all axes of, default uses current figure.
    
top, right, left, bottom : boolean, optional
    If True, remove that spine.
    
offset : int or dict, optional
    Absolute distance, in points, spines should be moved away from the axes
    (negative values move spines inward). A single value applies to all spines;
    a dict can be used to set offset values per side.
'''
```
