## 卡方分布（Chi-Squared Distribution）
<font color='red' size=4>定义:</font>  
&emsp;&emsp;$设X_{1},X_{2},...,X_{n}$为独立同分布的随机变量,且都服从$N(0,1).$记$Y=X_{1}^2+X_{2}^2+...+X_{n}^2$.则$ Y $的分布服从自由度为$ n $ 
的$ \chi^2 $分布,记为$ Y\sim \chi_{n}^2 $  

<font color='red' size=4>性质:</font> 
1. 总体服从正态分布并且方差等于$ \sigma^2 $时,统计量${\chi ^2} = \frac{{(n - 1){s^2}}}{{\sigma^2}} $服从自由度为$n-1$的卡方分布   
2. 设$ Z_1  \sim \chi_{n1}^2,  Z_2  \sim \chi_{n2}^2 $,且$ Z_1,Z_2 $相互独立,则$ Z_1 + Z_2 \sim \chi_{n1 + n2}^2 $  
对于给定的$\alpha,0<\alpha<1$称满足条件  
$$ P\{\chi_{n}^{2}>\chi_{n}^{2}(\alpha)\} =\int_{\chi_{n}^{2}(\alpha)}^{+\infty} f(x)dx=\alpha $$   
的点$\chi_{n}^{2}(\alpha)$为$\chi_{n}^{2}$上的$\alpha$分位点


* 均值和方差:$$E(\chi^2)=n, Var(\chi^2)=2n$$   

```python
# IMPORTS
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('classic')
fig = plt.figure(figsize=(14, 7), dpi=100)
ax  = plt.subplot(111)

# PDF
ax.plot(np.linspace(0, 20, 100), 
         stats.chi2.pdf(np.linspace(0, 20, 100), df=4))
ax.fill_between(np.linspace(0, 20, 100),
                 stats.chi2.pdf(np.linspace(0, 20, 100), df=4) ,
                 alpha=.15)

# CDF
ax.plot(np.linspace(0, 20, 100), 
         stats.chi2.cdf(np.linspace(0, 20, 100), df=4))

ax.text(x=11, y=.25, s="pdf (normed)", alpha=.75, weight="bold", color="#008fd5")
ax.text(x=11, y=.85, s="cdf", alpha=.75, weight="bold", color="#fc4f30")

ax.set_xticks(np.arange(0, 21, 2))
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = -2, y = 1.25, s = r"Chi-Squared $(\chi^{2})$ Distribution - Overview",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = -2, y = 1.1, 
         s = 'Depicted below are the normed probability density function (pdf) and the cumulative density\nfunction (cdf) of a Chi-Squared distributed random variable $ y \sim \chi^{2}(k) $, given $k$=4.',
         fontsize = 19, alpha = .85)
```
