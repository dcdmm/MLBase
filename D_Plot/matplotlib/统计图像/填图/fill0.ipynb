```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 1, figsize=(10, 7))
x = np.linspace(0, 1, 500)
y1 = np.sin(4 * np.pi * x) * np.exp(-5 * x)
y2 = -np.sin(4 * np.pi * x) * np.exp(-5 * x)
axes.fill(x, y1, color='red',  # color
          alpha=0.5, label='y1')  # 绘制曲线y1和y1首尾坐标连线构成的直线之间的区域
axes.fill(x, y2, color='g', alpha=0.5, label='y2')
axes.plot([x[0], x[-1]], [y1[0], y1[-1]], c='b', alpha=1, linewidth=3)
axes.legend()
```
