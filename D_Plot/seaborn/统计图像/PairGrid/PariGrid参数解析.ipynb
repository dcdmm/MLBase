```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
x_vars = ["body_mass_g", "bill_length_mm", "bill_depth_mm", "flipper_length_mm"]
y_vars = ["body_mass_g"]
penguins.head()
```

```python
# Calling the constructor sets up a blank grid of subplots with each row and one column corresponding to a variable in the dataset:
sns.PairGrid(penguins,
             diag_sharey=False,
             x_vars=x_vars,  # x轴方向使用的列
             y_vars=y_vars)  # y轴方向使用的列
```

```python
g = sns.PairGrid(penguins, diag_sharey=False, x_vars=x_vars, y_vars=x_vars)

# Plot with a bivariate function on the upper diagonal subplots.
g.map_upper(sns.scatterplot)

# Plot with a univariate function on each diagonal subplot.
g.map_diag(sns.kdeplot)

# Plot with a bivariate function on the lower diagonal subplots.
g.map_lower(sns.kdeplot)
```

```python
# Or to avoid the redundancy altogether:
g = sns.PairGrid(penguins, diag_sharey=False, x_vars=x_vars, y_vars=x_vars,
                 # If True, don’t add axes to the upper (off-diagonal) triangle of the grid, making this a “corner” plot.
                 corner=True)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
```

```python
# The PairGrid constructor accepts a hue variable. This variable is passed directly to functions that understand it:
g = sns.PairGrid(penguins, diag_sharey=False, x_vars=x_vars, y_vars=y_vars,
                 hue="species")  # 使用列"species"进行分组(对每个子图)
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot)
g.add_legend()  # Variables within data to use, otherwise use every column with a numeric datatype.

print(g.fig, len(g.fig.get_axes()), end='\n\n')
for i in g.fig.get_axes():
    """访问matplotlib底层"""
    print('title:', i.get_title())
    print('xlabel:', i.get_xlabel())
    print('ylabel:', i.get_ylabel())
    print('xlim:', i.get_xlim())
    print('ylim:', i.get_ylim())
    print('xticks:', i.get_xticks())
    print('xticklabels:', i.get_xticklabels(), end='\n\n')
```
