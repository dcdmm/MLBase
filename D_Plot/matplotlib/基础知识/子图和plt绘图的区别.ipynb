```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_axies(i):  # 每次调用生成一张图片(推荐此方式)
    t = np.arange(0.0, 2.0, 0.01)
    s = np.sin(i * np.pi * t)
    fig1, axes = plt.subplots()
    axes.plot(t, s)
    axes.set_title(i, color='red', size=23)


def plot_plt(i):  #  所有图片融合成一张图片
    t = np.arange(0.0, 2.0, 0.01)
    s = np.sin(i * np.pi * t)
    plt.plot(t, s)
    plt.title(i, color='red', size=23)


def plot_plt_show(i):  # 每次调用生成一张图片
    t = np.arange(0.0, 2.0, 0.01)
    s = np.sin(i * np.pi * t)
    plt.plot(t, s)
    plt.title(i, color='red', size=23)
    plt.show()  # Display all open figures.
```

```python
lst = [1, 2, 3]
```

```python
for i in lst:
    plot_axies(i)
```

```python
for j in lst:
    plot_plt(j)
```

```python
for k in lst:
    plot_plt_show(k)
```
