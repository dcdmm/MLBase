```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB')  # (RGB)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans = transforms.FiveCrop(size=100)
'''
size (sequence or int) 
    Desired output size of the crop. If size is an int instead of sequence like (h, w), a square crop of size (size, size) is made.
'''
five_result = trans(img_resize)  # 
five_result
```
