```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # (RGB)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans_rand0 = transforms.RandomCrop(size=(100, 200)) # Crop the given PIL Image at a random location.
'''
size (sequence or int) 
    Desired output size of the crop. If size is an int instead of sequence like (h, w), a square crop (size, size) is made.
'''
trans_rand0(img_resize)
```

```python
trans_rand0(img_resize)
```

```python
# 0()
trans_rand1 = transforms.RandomCrop(size=300, pad_if_needed=True) # size,pad_if_needed=True;pad_if_needed=False
trans_rand1(img_resize)
```

```python
trans_rand2 = transforms.RandomCrop(size=300, pad_if_needed=True, fill=(34, 123, 43))
'''
fill 
    Pixel fill value for constant fill. Default is 0. If a tuple of length 3, it is used to fill R, G, B channels respectively.
    This value is only used when the padding_mode is constant
'''
trans_rand2(img_resize)
```

```python
trans_rand30 = transforms.RandomCrop(224, padding=200) # /200,size
trans_rand30(img_resize)
```

```python
trans_rand31 = transforms.RandomCrop(224, padding=(30, 50)) # 20,50
trans_rand31(img_resize)
```

```python
# 30,50,60,80
trans_rand32 = transforms.RandomCrop(224, padding=(30, 50, 60, 80), fill=(24, 51, 151))
trans_rand32(img_resize)
```

```python
# pads with the last value on the edge of the image
trans_rand40 = transforms.RandomCrop(224 , padding=60, padding_mode='edge') # fillpadding_mode='constant'()
trans_rand40(img_resize)
```

```python
trans_rand41 = transforms.RandomCrop(224 , padding=60, padding_mode='reflect')
'''
reflect: pads with reflection of image (without repeating the last value on the edge)
    padding [1, 2, 3, 4] with 2 elements on both sides in reflect mode will result in [3, 2, 1, 2, 3, 4, 3, 2]
'''
trans_rand41(img_resize)
```

```python
trans_rand42 = transforms.RandomCrop(1024 , padding=500, padding_mode='symmetric')
'''
symmetric: pads with reflection of image (repeating the last value on the edge)
    padding [1, 2, 3, 4] with 2 elements on both sides in symmetric mode will result in [2, 1, 1, 2, 3, 4, 4, 3]
'''
trans_rand42(img_resize)
```
