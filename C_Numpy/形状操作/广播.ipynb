```python
import  numpy as np

x = np.random.randn(4, 32, 32, 3)
x.shape
```

```python
(x + 4).shape # 等价于x + np.broadcast_to(4, x.shape)
```

```python
(x * 4).shape # 等价于x * np.broadcast_to(4, x.shape)
```

```python
(x ** 3).shape
```

```python
(x == 4).shape
```

```python
# it starts with the trailing dimensions, and works its way forward
(x + np.random.randn(3)).shape # [4, 32, 32, 3] broadcasting [3, ] --> [4, 32, 32, 3]
```

```python
# if current dim==1,expand to same
# if either has no dim,insert one dim and expand to same
(x + np.random.randn(1, 3)).shape # [4, 32, 32, 3] boradcasting [1, 3] --> [4, 32, 32, 3]
```

```python
(x + np.random.randn(32, 32, 1)).shape # [4, 32, 32, 3] boradcasting [32, 32, 1] --> [4, 32, 32, 3]
```

```python
(x + np.random.randn(4, 1, 1, 1)).shape # [4, 32, 32, 3] boradcasting [4, 1, 1, 1] --> [4, 32, 32, 3]
```

```python
a = np.ones([1, 12, 8, 1, 4, 1])
b = np.zeros([5, 4, 1, 1, 3, 1, 4]) 
# [1, 12, 8, 1, 4, 1] boradcasting [5, 4, 1, 1, 3, 1, 4] --> [5, 4, 12, 8, 3, 4, 4]

(a + b).shape
```

```python
# 显式的使用broadcast_to函数完成此操作(不推荐)
a1 = np.broadcast_to(a, (5, 4, 12, 8, 3, 4, 4))
a1.shape
```

```python
b1 = np.broadcast_to(b, (5, 4, 12, 8, 3, 4, 4))
b1.shape
```

```python
(a + b) - (a1 + b1)
```
