<font color='red' size='4'>定义:</font>  
&emsp;&emsp;$ Poisson $ 分布的一个基本假设是:在较短的时间段内,事情出现的概率与等待时间成正比如果:
$ f(X=x|\lambda) = \frac{\lambda^{x}e^{-\lambda}}{x!},x=0,1...$,记该分布为$Poisson(\lambda)$  

* $$ E(X)=Var(Y)= \lambda$$  
* $$ Poisson分布的矩母函数M_{X}(t)=e^{\lambda (e^\lambda-1)} $$       

<font color='red' size='4'>性质:</font>   
&emsp;&emsp;若令$X\sim$ 参数为$(n,p)$的二项分布,$Y \sim$参数为 $\lambda$的Poisson分布,且$\lambda = np$,则该近似分布可以表示为:当n很大,np很小时,有   
$$ P(X=x) \approx P(Y=x) $$  

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

# PLOTTING CONFIG
%matplotlib inline
style.use('fivethirtyeight')
plt.figure(figsize=(14, 7), dpi=100)
ax = plt.subplot(111)

# PDF
ax.bar(x=np.arange(20), 
        height=(stats.poisson.pmf(np.arange(20), mu=5)), 
        width=.75,
        alpha=0.75
       )

# CDF
ax.plot(np.arange(20), 
         stats.poisson.cdf(np.arange(20), mu=5),
         color="#fc4f30",
        )

# LEGEND
ax.text(x=8, y=.25, s="pmf (normed)", alpha=.75, weight="bold", color="#008fd5")
ax.text(x=8.5, y=.9, s="cdf", alpha=.75, weight="bold", color="#fc4f30")

# TICKS
ax.set_xticks(range(21)[::2])
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0.005, color = 'black', linewidth = 1.3, alpha = .7)

# TITLE, SUBTITLE & FOOTER
ax.text(x = -2.5, y = 1.25, s = "Poisson Distribution - Overview",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = -2.5, y = 1.1, 
         s = 'Depicted below are the normed probability mass function (pmf) and the cumulative density\nfunction (cdf) of a Poisson distributed random variable $ y \sim Poi(\lambda) $, given $ \lambda = 5 $.',
         fontsize = 19, alpha = .85)
```
