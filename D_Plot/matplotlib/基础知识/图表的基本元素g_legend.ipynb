1. Axes.legend

```python
import numpy as np
import matplotlib.pyplot as plt

a = b = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1]
fig, ax = plt.subplots(2, 1, figsize=(10, 16))

'''
The elements to be added to the legend are automatically determined, when you do not pass in any extra arguments.
In this case, the labels are taken from the artist.
'''
ax[0].plot(a, c, '--', label='Model length')
ax[0].plot(a, d, ':', label='Data length')
ax[0].plot(a, c + d, label='Total message length')
ax[0].legend(edgecolor='white')

# For full control of which artists have a legend entry, it is possible to pass an iterable of legend artists followed by an iterable of legend labels respectively:
line0 = ax[1].plot(a, c, '--', color='red')
line1 = ax[1].plot(a, d, ':', color='green')
line2 = ax[1].plot(a, c + d)
legends = ax[1].legend(handles=[line0[0], line1[0], line2[0]],
                       labels=('Model length',  # line0
                               'Data length',  # line1
                               'Total message length'),  # line2
                       loc='upper left',  # 
                       bbox_to_anchor=(0.5, 0.85),  # (0.5, 0.85)
                       prop={'size': 10, 'weight': 'bold'},  # 
                       ncol=3,  # 
                       facecolor='yellow',  # 
                       edgecolor='red',  # 
                       title="legend",  # 
                       title_fontsize=20,  # 
                       framealpha=0.4)  # facecolor,edgecolor
```
