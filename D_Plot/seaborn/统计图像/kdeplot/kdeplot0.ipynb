```python
import numpy as np
import seaborn as sns

np.random.seed(10)
sns.set(color_codes=True)
```

```python
# Draw random samples from a multivariate normal distribution.
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov, size=50).T
```

```python
sns.kdeplot(x=x, ax=None)  # pandas.DataFrame or numpy.ndarray
```

```python
# Single color specification for when hue mapping is not used. Otherwise, the plot will try to hook into the matplotlib property cycle.
sns.kdeplot(x=x, color='red')  # 默认color=None
```

```python
# If True, fill in the area under univariate density curves or between bivariate contours. If None, the default depends on multiple.
sns.kdeplot(x=x, fill=True, color="red")  # 默认fill=None
```

```python
sns.kdeplot(y=x, fill=True, color="red")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.kdeplot(data=tips, x="total_bill")
```

```python
'''
bw_methodstring, scalar, or callable, optional
    Method for determining the smoothing bandwidth to use; passed to scipy.stats.gaussian_kde(https://docs.scipy.org/doc/scipy/reference/reference/generated/scipy.stats.gaussian_kde.html#scipy.stats.gaussian_kde).

bw_adjustnumber, optional
    Factor that multiplicatively scales the value chosen using bw_method. Increasing will make the curve smoother. See Notes.
'''
# Use less smoothing:
sns.kdeplot(data=tips, x="total_bill", bw_method='scott', bw_adjust=0.1)
```

```python
sns.kdeplot(data=tips, x="total_bill", bw_method='scott', bw_adjust=5)
```

```python
# plot conditional distributions with hue mapping of a second variable:
sns.kdeplot(data=tips, x="total_bill", hue="time")
```

```python
# “Stack” the conditional distributions:
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="stack")
```

```python
# Normalize the stacked distribution at each value in the grid:
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="fill")
```

```python
# Plot distributions for each column of a wide-form dataset:
iris = sns.load_dataset("iris")
iris
```

```python
# Plot distributions for each column of a wide-form dataset:
sns.kdeplot(data=iris)
```

```python
# Use numeric hue mapping:
sns.kdeplot(data=tips, x="total_bill", hue="size")
```

```python
# Map the data variable with log scaling:
diamonds = sns.load_dataset("diamonds")
sns.kdeplot(data=diamonds, x="price", log_scale=True)
```

```python
# 坐标轴名等于iris['sepal_length'].name
sns.kdeplot(iris['sepal_length'])
```

```python
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
sns.kdeplot(x=setosa.sepal_width, y=setosa.sepal_length,
                 cmap="Reds", cbar=True, shade=True)
sns.kdeplot(x=virginica.sepal_width, y=virginica.sepal_length,
                 cmap="Blues", cbar=True, shade=True,
                 cbar_kws={'orientation': 'horizontal'})
sns.rugplot(x=setosa.sepal_width)
sns.rugplot(y=setosa.sepal_length)
```
