```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

# PLOTTING CONFIG
%matplotlib inline
style.use('fivethirtyeight')
plt.figure(figsize=(14, 7), dpi=100)
ax = plt.subplot(111)

# PDF LAM = 1
ax.scatter(np.arange(20),
            (stats.poisson.pmf(np.arange(20), mu=1)),
            alpha=0.75, s=100)
ax.plot(np.arange(20),
         (stats.poisson.pmf(np.arange(20), mu=1)), alpha=0.75)

# PDF LAM = 5
ax.scatter(np.arange(20),
            (stats.poisson.pmf(np.arange(20), mu=5)),
            alpha=0.75, s=100)
ax.plot(np.arange(20),
         (stats.poisson.pmf(np.arange(20), mu=5)), alpha=0.75)

# PDF LAM = 10
ax.scatter(np.arange(20),
            (stats.poisson.pmf(np.arange(20), mu=10)),
            alpha=0.75, s=100)
ax.plot(np.arange(20),
         (stats.poisson.pmf(np.arange(20), mu=10)),
         alpha=0.75)

# LEGEND
ax.text(x=3, y=.1, s="$\lambda = 1$", alpha=.75, rotation=-65, weight="bold", color="#008fd5")
ax.text(x=8.25, y=.075, s="$\lambda = 5$", alpha=.75, rotation=-35, weight="bold", color="#fc4f30")
ax.text(x=14.5, y=.06, s="$\lambda = 10$", alpha=.75, rotation=-20, weight="bold", color="#e5ae38")

# TICKS
ax.set_xticks(range(21)[::2])
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

# TITLE, SUBTITLE & FOOTER
ax.text(x = -2.5, y = .475, s = "Poisson Distribution - $\lambda$",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = -2.5, y = .425, 
         s = 'Depicted below are three Poisson distributed random variables with varying $\lambda $. As one can easily\nsee the parameter $\lambda$ shifts and flattens the distribution (the smaller $ \lambda $ the sharper the function).',
         fontsize = 19, alpha = .85)
```
