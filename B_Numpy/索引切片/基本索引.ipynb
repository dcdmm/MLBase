```python
import  numpy as np

arr_mul = np.random.randn(5, 7, 4)
arr_mul
```

```python
arr_mul[0, 1, 1]  # 索引标量数值
```

```python
arr_mul[0]  # 从第零个维度开始索引/切片(DataFrame从列开始进行索引切片)
```

```python
arr_mul[0, 0, -1]  # 索引,若为标量则返回具体数值(且支持负数索引)
```

```python
arr_mul[0:2]  # 切片,返回array
```

```python
arr_mul[0:2, 0:3, 0]  # 在多个维度上索引/切片
```

```python
arr_mul[0:2, 0:6:2, 2]  # 第一个维度上的切片步长为2(最后一步若小于步长,则丢弃),切片范围为0-6
```

```python
arr_mul[0:2, -1:0:-1, 2]  # 支持切片步长为负数
```

```python
arr_mul[:]  # 等价于arr_mul[:, :, :]
```

```python
arr_mul[::2]  # 第一个维度上切片步长为2,范围为整个维度;等价于arr_mul[::2, :, :]
```

```python
arr_mul[::-1]  # 第一个维度数据倒序
```

```python
arr_complex = np.random.randn(12, 7, 8, 10, 9, 15, 14)
arr_complex.shape
```

```python
print(arr_complex[1:5].shape)
print(arr_complex[1:5, :, :, :, :, :].shape)
print(arr_complex[1:5, ...].shape)  # ...:剩余维度上所有数据
```

```python
print(arr_complex[:, :, :, :, :, :, 2:4].shape)
print(arr_complex[..., 2:4].shape)
```

```python
print(arr_complex[2:7, :, :, :, :, :, 4].shape)
print(arr_complex[2:7, ..., 4].shape)
```
