```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB')  # (RGB)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans = transforms.TenCrop(100,
                           vertical_flip=True)  # ;vertical_flip=False,
# ,,,,,,,,,
ten_result = trans(img_resize)
```

```python
import matplotlib.pyplot as plt

for i in ten_result:
    plt.imshow(i)
    plt.show()
```
