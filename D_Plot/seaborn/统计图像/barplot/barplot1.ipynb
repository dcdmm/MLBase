```python
import  seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(x='sex', y='survived', data=titanic, palette='Set1')
titanic.groupby(['sex']).mean()
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class')
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class', palette='Set1')
titanic.groupby(['sex', 'class']).mean()
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class', palette='Set1',
            order=['female', 'male'])
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class', palette='Set1',
            order=['female', 'male'], capsize=0.05, saturation=0.5)
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class', palette='Set1',
            order=['female', 'male'], capsize=0.05, saturation=0.5, errcolor='red',
            errwidth=3)
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class', palette='Set1',
            order=['female', 'male'], capsize=0.05, saturation=0.5, errcolor='red',
            errwidth=3, ci='sd')
```
