```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170

X, y = make_blobs(n_samples=n_samples, random_state=random_state) # Generate isotropic Gaussian blobs for clustering
y_pred = AgglomerativeClustering(n_clusters=3, # 默认n_clusters=2 
                                 affinity='euclidean' # 默认affinity=’euclidean’(欧式距离)
                                                      # 'l2'(欧式距离)
                                                      # 'l1', manhattan(曼哈顿距离)
                                                      # 'cosine'(余弦距离) 
                                 ).fit_predict(X) 
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")

# Anisotropicly distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = AgglomerativeClustering(n_clusters=3).fit_predict(X_aniso)
plt.subplot(222)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Anisotropicly Distributed Blobs")

# Different variance
X_varied, y_varied = make_blobs(n_samples=n_samples,
                                cluster_std=[1.0, 2.5, 0.5],
                                random_state=random_state)
y_pred = AgglomerativeClustering(n_clusters=3).fit_predict(X_varied) # Fit the hierarchical clustering from features or distance matrix, and return cluster labels.
# 与上等价
# ward = AgglomerativeClustering(n_clusters=3).fit()
# y_pred = ward.labels_ 

plt.subplot(223)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Unequal Variance")

# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = AgglomerativeClustering(n_clusters=3).fit_predict(X_filtered)

plt.subplot(224)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs")
plt.show()
```
