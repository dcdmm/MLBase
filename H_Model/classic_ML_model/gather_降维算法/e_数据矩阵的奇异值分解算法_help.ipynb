&emsp;&emsp;由截断奇异值分解,对于$m \times n$实矩阵$A$假设其秩为$r,0 < k<r$,则可以将矩阵$A$进行截断奇异值分解

$$ A \approx U_k \Sigma _k V_k^T $$

其中$U_k$是$ m \times k$矩阵,$V_k$是$n \times k$矩阵,$\Sigma_k$是$k$阶对角矩阵;$U_k, V_k$分别由取$A$的完全奇异值分解的矩阵$U,
V$的前$k$列,$\Sigma_k$由取$A$的完全奇异值分解的矩阵$\Sigma$的前$k$个对角线元素得到.

&emsp;&emsp;定义一个新的$n \times m$矩阵$X'$

$$ X' = \frac{1}{\sqrt{n-1}} X^T $$

$X'$的每一列均值为零(即$X$的每一行均值为零),不难得知,

$$
\begin{aligned}
X'^T X' &= \left( \frac{1}{\sqrt{n-1}} X^T \right)^T \left( \frac{1}{\sqrt{n-1}} X^T \right) \\
    &= \frac{1}{n-1}XX^T
\end{aligned}
$$

***

&emsp;&emsp;假设对$m$维随机向量$X=(X_1, X_2, \dots,X_m)^T$进行$n$次独立观测,$\mathbf{x}_1, \mathbf{x}_2, \cdots, \mathbf{x}_n$表示观测样本,
其中$\mathbf{x}_j = (x_{1j}, x_{2j}, x_{mj})^T$表示第$j$个观测样本,$x_{ij}$表示第$j$个观测样本的第$i$个变量,
$j=1,2,\cdots,n$.观测数据样本矩阵用$\mathbf{X}$表示,记作

$$
\mathbf{X} =
\begin{bmatrix}
\mathbf{x}_{1} \mathbf{x}_{2} & \cdots & \mathbf{x}_{n}
\end{bmatrix}
=
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1 n} \\
x_{21} & x_{22} & \cdots & x_{2 n} \\
\vdots & \vdots & & \vdots \\
x_{m 1} & x_{m 2} & \cdots & x_{m n}
\end{bmatrix}
$$

&emsp;&emsp;给定样本矩阵$\mathbf{X}$,可以估计样本均值,以及样本方差.样本均值向量$\mathbf{\bar{x}}$为

$$ \mathbf{\bar{x}} = \left[ \frac{1}{n} \sum_{k=1}^n x_{i1},\frac{1}{n} \sum_{k=1}^n x_{i2},\cdots,\frac{1}{n} \sum_{k=1}^n x_{im} \right] $$

样本协方差矩阵$S$为

$$
\begin{aligned}
& S = [s_{ij}]_{m \times m} \\
& s_{ij} = \frac{1}{n-1} \sum_{k=1}^n (x_{ik} - \mathbf{\bar{x}}_i)(x_{jk} - \mathbf{\bar{x}}_j)
\end{aligned}
$$

其中$\mathbf{\bar{x}}_i = \frac{1}{n} \sum_{k=1}^n x_{ik}$为第$i$个变量的样本均值,$\mathbf{\bar{x}}_j = \frac{1}{n} \sum_{k=1}^n x_{jk}$.

&emsp;&emsp;样本相关矩阵$R$为

$$ R= [r_{ij}]_{m \times m}, \quad r_{ij} = \frac{s_{ij}}{\sqrt{s_{ii} s_{jj}}}, \quad i,j=1,2,\cdots,m $$

***

即$X'^T X'$等于$X$的协方差矩阵$S_X$.

$$ S_X =X'^T X' $$

主成分分析归结于求协方差矩阵$S_X$的特征值和对应的单位特征向量,所以问题转化为求$X'^T X'$的特征值和对应的单位特征向量.

&emsp;&emsp;假设$X'$的截断奇异值分解为$X' = U\Sigma V^T$,那么$V$的列向量就是$S_X =X'^T X'$的单位特征向量($S_X =X'^T X' = V\Sigma^T U^T U \Sigma V^T = V \Sigma^T \Sigma V^T$).因此,$V$的列向量就是$X$的主成分.于是,求$X$的主成分可以通过求$X'$的奇异值分解来实现.具体算法如下.


<font size=4 color='red'>算法:(主成分分析算法)</font>

输入:$m \times n$样本矩阵$X$,其每一行元素均值为零

输出:$k \times n$个样本主成分矩阵$Y$

参数:主成分个数$k$

1. 构造新的$n \times m$矩阵

$$ X' = \frac{1}{\sqrt{n-1}} X^T $$

$X'$的每一列均值为零.

2. 对矩阵$X'$进行截断奇异值分解,得到

$$ X' = U\Sigma V^T $$

有$k$个奇异值,奇异向量.矩阵$V$的前$k$列构成$k$个样本主成分.

3. 求$k \times n$样本主成分矩阵

$$ Y = V^T X $$
