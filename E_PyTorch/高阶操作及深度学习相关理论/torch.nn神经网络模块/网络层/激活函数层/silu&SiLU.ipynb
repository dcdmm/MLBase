$ \operatorname{silu}(x)=x∗\sigma(x)$,where $\sigma(x)$ is the logistic sigmoid.

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import torch.nn as nn
```

```python
x = torch.linspace(-5, 5, 200)
relu_f = F.relu(x)
silu_f = F.silu(x)

fig, ax = plt.subplots(1, 1)
ax.plot(x, relu_f, '--', color='red', label='rule')
ax.plot(x, silu_f, label='silu')  # 更加平滑
ax.legend(edgecolor='white')
```

```python
relu_c = nn.SiLU()
relu_c(x)
```
