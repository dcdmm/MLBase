```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = plt.subplot2grid((3, 3), # 共3 * 3个子图
                       (0, 0), # 该子图位于第1行第1列的位置
                       colspan=3) # 该子图跨度为3列
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3),
                       (1, 2), # 该子图位于第2行,第3列的位置 
                       rowspan=2) # 该子图跨度为2行
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax1.plot(x, y, label='ax1')
ax2.plot(x, y, label='ax2')
ax3.plot(x, y, label='ax3')
ax4.plot(x, y, label='ax4')
ax5.plot(x, y, label='ax5')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
ax5.legend()

plt.show()
```
