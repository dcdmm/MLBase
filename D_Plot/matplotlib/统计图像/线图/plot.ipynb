```python
import  matplotlib.pyplot as plt
import numpy as np
import  pandas as pd
```

```python
fig1 = plt.figure(figsize=(10, 7))
ax1 = plt.subplot(111)
a = [1, 5, 8, 2, 4, 5]
b = [1, 2, 9, 11, 5, 4]
line = ax1.plot(a, b, color='red', # color
                linewidth=3)

# 绘制坐标点(1,1), (5,2), (8,9), (2,11), (4,5), (5,4)依次连接所构成的直线
ax1.set_ylabel('some numbers')
for i, j in zip(a, b):
    ax1.text(i, j , (i, j), color='g') # 添加点坐标
print(line)
```

```python
fig1 = plt.figure(figsize=(10, 7))
ax2 = plt.subplot(512)
plt.plot([i**2 for i in range(100)], linestyle='-.')
'''
linestyle:线条样式
    ``'-'``          solid line style
    ``'--'``         dashed line style
    ``'-.'``         dash-dot line style
    ``':'``          dotted line style
'''
```

```python
fig3 = plt.figure(figsize=(10, 7))
ax3 = plt.subplot(111)
s = pd.Series(np.random.randn(100).cumsum())
ax3.plot(s, marker='*') # marker:Set the line marker.
```

```python
fig4 = plt.figure(figsize=(10, 7))
ax4 = plt.subplot(111)
ax4.plot(np.random.randn(50), color='b', alpha=0.6, linewidth=4)
'''
linewidth:Set the line width in points
color:Set the color of the line
'''
```

```python
fig5 = plt.figure(figsize=(10, 7))
ax5 = plt.subplot(111)
data = np.random.normal(size=(20, 6)) + np.arange(6)/2
df = pd.DataFrame(data) 
line_list = ax5.plot(df) # 直接对DataFrame进行绘图
for line in line_list: # plot函数的返回值:A list of Line2D objects representing the plotted data.
    print(line)
```

```python
fig6 = plt.figure(figsize=(10, 7))
ax6 = plt.subplot(111)
for i in range(df.shape[1]):
    ax6.plot(df.iloc[:, i], lable=i) # 与上等价
plt.legend()
```

```python
x = np.random.randint(1, 10, size=10)
y = np.random.randint(1, 10, size=10)
plt.plot(x, y, linestyle='-')
for i in range(len(x)):
    plt.scatter(x[i], y[i], cmap='Set1', s=100) # 线图和散点图结合
```
