### 原始问题       
&emsp;&emsp;线性不可分的线性支持向量机的学习问题为如下凸二次规划(convex quadratic programming)问题:

$$
\begin{aligned}
& \min_{\mathbf{w}, b, \xi} \frac{1}{2} ||\mathbf{w}||^2 + C \sum_{i=1}^N \xi_i \\
& s.t. y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1- \xi_i, \quad i=1,2,\cdots,N \\
& \xi_i \geq 0, \quad i=1,2,\cdots,N
\end{aligned}
$$     

这里,$C >0$为惩罚参数,一般由应用问题决定,$C$值大时对误分类的惩罚增大,$C$值小时对误分类的惩罚减少,最小化目
标函数包含二层含义:使得$\frac{1}{2} ||\mathbf{w}||^2$尽量小即间隔尽量大,同时使误分类点的个数尽量小,$C$是调和二者的系数.            

### 学习的对偶算法

&emsp;&emsp;原始问题的对偶问题是            

$$
\begin{aligned}
\min_{\alpha} \quad & \frac{1}{2} \sum_{i=1}^{N} \sum_{j=1}^{N} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j) - \sum_{i=1}^{N} \alpha_i  \\
s.t.          \quad & \sum_{i=1}^N \alpha_i y_i =0 \\
&0 \leq \alpha_i \leq C, \quad i=1,2,\cdots, N
\end{aligned} 
$$     

<font size=4 color='red'>证明:</font>原始问题的拉格朗日函数是       
$$ L(\mathbf{w}, b, \xi, \alpha, \mu) = \frac{1}{2} ||\mathbf{w}||^2 + C\sum_{i=1}^{N} \xi_i -\sum_{i=1}^{N}\alpha_i(y_i(\mathbf{w} \cdot \mathbf{x}_i + b)-1 + \xi_i) - \sum_{i=1}^{N} \mu_i \xi_i  \tag{1} $$     
其中,$\alpha_i \geq 0, \mu_i \geq 0$.         
&emsp;&emsp;对偶问题是拉格朗日函数的极大极小问题.首先求$L(\mathbf{w}, b, \xi, \alpha, \mu)$对$\mathbf{w}, b, \xi$的极小,由        

$$
\begin{aligned}
\nabla_\mathbf{w} L(\mathbf{w}, b, \xi, \alpha, \mu) &= \mathbf{w}- \sum_{i=1}^N \alpha_i y_i \mathbf{x}_i =0 \\
\nabla_b L(\mathbf{w}, b, \xi, \alpha, \mu) &= \sum_{i=1}^N \alpha_i y_i =0 \\
\nabla_{\xi} L(\mathbf{w}, b, \xi, \alpha, \mu) &= C - \alpha_i - \mu_i =0
\end{aligned}
$$       

得     

$$
\begin{align}
\mathbf{w} &= \sum_{i=1}^N \alpha_i y_i \mathbf{x}_i \tag{2} \\
\sum_{i=1}^N \alpha_i y_i &=0 \tag{3} \\
C - \alpha_i - \mu_i &=0 \tag{4}
\end{align}
$$     

&emsp;&emsp;将(2),(3),(4)代入(1)式,得       
$$  \min_{\mathbf{w}, b, \xi} L(\mathbf{w}, b, \xi, \alpha, \mu) = -\frac{1}{2} \sum_{i=1}^N \sum_{j=1}^N \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j) + \sum_{i=1}^{N} \alpha_i $$     
再对$ \min_{\mathbf{w}, b, \xi} L(\mathbf{w}, b, \xi, \alpha, \mu) $求$\alpha$的极大,即得对偶问题:              

$$
\begin{align}
\max_{\alpha} \quad & -\frac{1}{2} \sum_{i=1}^N \sum_{j=1}^N \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j) + \sum_{i=1}^{N} \alpha_i   \\
s.t. \quad & \sum_{i=1}^N \alpha_i y_i =0 \\
& C - \alpha_i - \mu_i =0 \tag{5} \\
& \alpha_i \geq 0 \tag{6} \\
& \mu \geq 0, \quad i=1,2,\cdots, N \tag{7}
\end{align}
$$      

&emsp;&emsp;利用等式约束(4)消去$\mu_i$,从而只留下变量$\alpha_i$,并将约束(5),(6),(7)写成        
$$ 0 \leq \alpha_i \leq C $$           
再将对目标函数求极大转换为求极小,得证.        

***

<font size=4 color='red'>定理:</font>设$\alpha^* = (\alpha^*_1,\alpha^*_2,\cdots, \alpha^*_N)^T$是线性支持向量
机对偶问题的一个解,若存在$\alpha^*$的一个分量$ 0<\alpha^*_j <C$,则原始问题的解$\mathbf{w}^*, b^*$可按下式求得:             

$$
\begin{align}
\mathbf{w}^* = \sum_{i=1}^N \alpha_i^* y_i \mathbf{x}_i  \tag{8}\\
b^* = y_j - \sum_{i=1}^N y_i\alpha_i^* (\mathbf{x}_i \cdot \mathbf{x}_j) \tag{9}
\end{align}
$$        

<font size=4 color='red'>证明:</font>原始问题是凸二次规划问题,解满足KKT条件.即得       

$$
\begin{align}
\nabla_\mathbf{w} L(\mathbf{w}^*, b^*, \xi^*, \alpha^*, \mu^*) &= \mathbf{w}^* - \sum_{i=1}^N \alpha_i^* y_i \mathbf{x}_i =0 \\
\nabla_b L(\mathbf{w}^*, b^*, \xi^*, \alpha^*, \mu^*) &= -\sum_{i=1}^N \alpha_i^* y_i =0 \\
\nabla_\mathbf{w} L(\mathbf{w}^*, b^*, \xi^*, \alpha^*, \mu^*) &= C -\alpha^* - \mu^*=0 \\
\alpha^*_i(y_i(\mathbf{w}^* \cdot \mathbf{x}_i + b^*)-1 + \xi_i^*) &= 0 \tag{10} \\
\mu^*_i \xi_i^* &=0  \tag{11} \\
y_i(\mathbf{w}^* \cdot \mathbf{x}_i + b^*) -1 + \xi^*_i &\geq 0 \\
\xi_i^* &\geq 0 \\
\alpha_i^* &\geq 0 \\
u_i^* & \geq 0, \quad i=1,2,\cdots, N
\end{align} 
$$         

易知(8)式显然成立.再由(10)可知,若存在$\alpha_j^*,0<\alpha_j^*<C$,此时由(10)可知$\mu_j^* \neq 0, \xi_j^* =0$,则$y_i(\mathbf{w}^* \cdot \mathbf{x}_i + b^*) -1 =0$.由此可得(9)式      
&emsp;&emsp;由此定理可知,分离超平面写成                     
$$ \sum_{i=1}^N \alpha_i^* y_i(\mathbf{x} \cdot \mathbf{x}_i) + b^* =0  $$      
分类决策函数可以写成          
$$  f(\mathbf{x}) = \mathrm{sign} \left(  \sum_{i=1}^N \alpha_i^* y_i(\mathbf{x} \cdot \mathbf{x}_i) + b^* =0  \right)   $$      

***

<font color='red' size=4>算法:</font>             
输入:训练数据集$T=\{ (\mathbf{x}_1,y_1), (\mathbf{x}_2, y_2), \cdots, (\mathbf{x}_N, y_N) \}$,其中,$\mathbf{x}_i \in \mathcal{X} = \mathbb{R}^n, y_i \in \{ -1, 1 \}, i=1,2,\cdots, N$;           
输出:分离超平面和分离决策函数.              

1. 选择惩罚参数$C >0$,构建并求解凸二次问题 

$$
\begin{aligned}
\min_{\alpha} \quad & \frac{1}{2} \sum_{i=1}^{N} \sum_{j=1}^{N} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j) - \sum_{i=1}^{N} \alpha_i  \\
s.t.          \quad & \sum_{i=1}^N \alpha_i y_i =0 \\
&0 \leq \alpha_i \leq C, \quad i=1,2,\cdots, N
\end{aligned}
$$

求得最优解$\alpha^* = (\alpha^*_1,\alpha^*_2,\cdots, \alpha^*_N)^T$.             

2. 计算$\mathbf{w}^* = \sum_{i=1}^N \alpha_i^* y_i \mathbf{x}_i $
选择$\alpha^*$的一个分量$\alpha^*_j$适合条件$0< \alpha_j^* <C$,计算            
$$ b^* = y_j - \sum_{i=1}^N y_i\alpha_i^* (\mathbf{x}_i \cdot \mathbf{x}_j) $$    

3. 求得分离超平面          
$$  \mathbf{w}^* \cdot \mathbf{x} + b^* =0 $$            
分类决策函数:           
$$ f(\mathbf{x}) = \mathrm{sign}(\mathbf{w}^* \cdot \mathbf{x} + b^*) $$                         
步骤(2)中,对任一适合条件$0< \alpha_j^* <C$,都可以计算求出$b^*$,故$b$的解可能不唯一,而是存在于一个区间.
