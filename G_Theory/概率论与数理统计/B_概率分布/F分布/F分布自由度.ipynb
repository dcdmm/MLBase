```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style
import warnings
warnings.filterwarnings('ignore')

style.use('classic')
fig = plt.figure(figsize=(12, 6), dpi=100)
ax  = plt.subplot(111)
plt.figure(dpi=100)

# PDF n1 = 1, n2=1
ax.plot(np.linspace(0, 5, 1000), 
         stats.f.pdf(np.linspace(0, 5, 1000), dfn=1, dfd=1), color='red' , label='F(1/1)')

# PDF n1=2, n2=1
ax.plot(np.linspace(0, 5, 1000), 
         stats.f.pdf(np.linspace(0, 5, 1000), dfn=2, dfd=1), color='green', label='F(2/1)')

# PDF n1=5,,n2=2
ax.plot(np.linspace(0, 5, 1000), 
         stats.f.pdf(np.linspace(0, 5, 1000), dfn=5, dfd=2), color='blue', label='F(5/2)')

# PDF n1=100, n2=100
ax.plot(np.linspace(0, 5, 1000), 
         stats.f.pdf(np.linspace(0, 5, 1000), dfn=100, dfd=100), color='black', label='F(100/100)')

ax.legend()

ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = 0, y = 5.21, s = "F Distribution - $ n_{1}, n_{2} $",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = 0, y = 4.61, 
         s = 'Depicted below are five F distributed random variables with varying $ n_{1}, n_{2} $. As one can\nsee the parameter $n_{1},n_{2}$ smoothens the distribution and softens the skewness.',
         fontsize = 19, alpha = .85)
```
