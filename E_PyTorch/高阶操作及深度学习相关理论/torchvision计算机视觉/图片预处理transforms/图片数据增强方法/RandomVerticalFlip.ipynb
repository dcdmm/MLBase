```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # 读取图片(RGB模式)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
# 垂直翻转
trans = transforms.RandomVerticalFlip(p=0.8) # 默认p=0.5(被翻转的可能性)
trans(img_resize)
```
