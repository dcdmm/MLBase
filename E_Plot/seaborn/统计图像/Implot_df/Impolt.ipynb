```python
import seaborn as sns
import  matplotlib.pyplot as plt
import  numpy as np
tips = sns.load_dataset("tips")
tips.head()
# g = sns.lmplot(x=tips["total_bill"], y=tips["tip"]) lmplot() missing 1 required positional argument: 'data'
```

```python
# Plot a simple linear relationship between two variables:
g = sns.lmplot(x="total_bill", y="tip", data=tips, 
               scatter_kws={'color':plt.cm.jet(np.linspace(0, 1, len(tips['total_bill'])))
                            ,'s':50}, markers='v')
```

```python
# Condition on a third variable and plot the levels in different colors:
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette='Set1', 
               legend_out=True)
```

```python
# Use different markers as well as colors so the plot will reproduce to black-and-white more easily:
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
                markers=["o", "x"])
```

```python
# Use a different color palette:
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
                palette="Set1")
```

```python
# Map hue levels to colors with a dictionary:
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
                palette=dict(Yes="y", No="m"))
```

```python
#Plot the levels of the third variable across different columns:
g = sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", hue='smoker', col="day",
               col_wrap=3, data=tips, aspect=0.5, palette='Set1') #col_warp应设为大于1的整数
```

```python
# Change the the height and aspect ratio of the facets:
g = sns.lmplot(x="size", y="total_bill", hue="day", col="day",
                data=tips, size=5, aspect=.4, x_jitter=0.1, palette='Set1')
```

```python
# Wrap the levels of the column variable into multiple rows:
g = sns.lmplot(x="total_bill", y="tip", col="day", data=tips, col_wrap=2)
```

```python
# Condition on two variables to make a full grid:
g = sns.lmplot(x="total_bill", y="tip", col="time", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", row="sex", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", row="sex", col="time",
               data=tips)
```

```python
# Use methods on the returned FacetGrid instance to further tweak the plot:
g = sns.lmplot(x="total_bill", y="tip", row="sex", col="time",
                data=tips, logx=True)
g = (g.set_axis_labels("Total bill (US Dollars)", "Tip")
       .set(xlim=(0, 60), ylim=(0, 12),
            xticks=[10, 30, 50], yticks=[2, 6, 10])
       .fig.subplots_adjust(wspace=.05))

```
