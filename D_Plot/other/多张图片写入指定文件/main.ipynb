```python
# Reads an image from the specified file.
img_1 = imageio.imread('../../../Other/img/rnn1N_0.jpg')
img_1
```

```python
image_list = []
image_list.append(img_1)
image_list.append(imageio.imread('../../../Other/img/rnn1N_1.jpg'))
image_list.append(imageio.imread('../../../Other/img/rnnN1.jpg'))
image_list.append(imageio.imread('../../../Other/img/rnnNN.jpg'))
image_list
```

```python
'''
duration : {float, list}
        The duration (in seconds) of each frame. Either specify one value
        that is used for all frames, or one value for each frame.
'''
# write multiple images to the specified file.
imageio.mimwrite(uri="test.gif", ims=image_list, format='GIF',
                duration=1)
```
