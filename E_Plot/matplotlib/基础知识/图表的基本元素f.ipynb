```python
import numpy as np
import  matplotlib.pyplot as plt

t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)
fig, ax = plt.subplots()
ax.plot(t,s)
ax.set_title(r'$\alpha_i > \beta_i$', fontsize=20)


ax.text(0.5, 0, # 注释文本坐标指的是该点的(xticks, yticks)
        r'$\mathcal{A}\mathrm{sin}(2 \omega t)$', 
        fontsize=20, # 设置注释文本大小
        fontweight='heavy', # 设置注释文本字体
        color='red', # 设置注释文本颜色
        alpha=0.9, # 设置注释字文本颜色透明度
        rotation='45', # 设置注释文本旋转角度
        # 下面将注释文本的左上点的坐标设置为(0.5, 0)
        ha='left', # 注释文本水平对齐方式
        va='top', # 注释文本垂直对齐方式
        rotation_mode='anchor') # None:先进行选择再进行对齐;anchor:先对齐再进行选择
ax.text(0.5, 0, # 注释点坐标指的是该点的(xticks, yticks)
        r'$\mathcal{A}\mathrm{sin}(2 \omega t)$', 
        fontsize=20, fontweight='heavy', color='g', 
        rotation='45', alpha=0.9, rotation_mode=None,
        ha='right', va='bottom')

'''
x, y : scalars
    The position to place the text. By default, this is in data coordinates. The coordinate system can be changed using the transform parameter.

s : str
    The text.

rotation: {angle in degrees, 'vertical', 'horizontal'}
    Set the rotation of the text.

color:Set the foreground color of the text
alpha:alpha:透明度(0-1)

fontsize:fontsize : {size in points, 'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}  
    Set the font size. May be either a size string, relative to the default font size, or an absolute font size in points.

fontweight:weight : {a numeric value in range 0-1000, 'ultralight', 'light', 'normal', 'regular', 'book', 'medium', 'roman', 'semibold', 'demibold', 'demi', 'bold', 'heavy', 'extra bold', 'black'}
    Set the font weight.
    
set_horizontalalignment/ha(self, align)
    Set the horizontal alignment to one of
    Parameters:	
    align:{'center', 'right', 'left'} 
    
set_verticalalignment/va(self, align)
    Set the vertical alignment
    Parameters:	
    align:{'center', 'top', 'bottom', 'baseline', 'center_baseline'}   
'''
ax.set_xlabel('time (s)')
ax.set_ylabel('volts (mV)')
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax1.plot(t, s, lw=2)

ax1.annotate('local max', xy=(2, 1), xytext=(3, 1.5), color='r',
            alpha=0.4, fontsize=18, rotation='vertical',
            arrowprops=dict(edgecolor='r', facecolor='g',
                            arrowstyle='<|-', alpha=0.3)) #注释点坐标指的是该点的(xticks, yticks)
'''
s : str
    The text of the annotation.

xy : (float, float)
    The point (x,y) to annotate.
    
color:Set the foreground color of the text
alpha:透明度(0-1)

fontsize : {size in points, 'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
    Set the font size. May be either a size string, relative to the default font size, or an absolute font size in points.

rotation: {angle in degrees, 'vertical', 'horizontal'}
    Set the rotation of the text.
    
xytext : (float, float), optional
    The position (x,y) to place the text at. If None, defaults to xy.
       
arrowprops : dict, optional
    The properties used to draw a FancyArrowPatch arrow between the positions xy and xytext.
    If arrowprops does not contain the key 'arrowstyle' the allowed keys are:
    Key	            Description
    width	            The width of the arrow in points
    headwidth	        The width of the base of the arrow head in points
    headlength	        The length of the arrow head in points
    shrink	            Fraction of total length to shrink from both ends
    ?	                Any key to matplotlib.patches.FancyArrowPatch
    If arrowprops contains the key 'arrowstyle' the above keys are forbidden. The allowed values of 'arrowstyle' are:
    Name	            Attrs
    '-'	            None
    '->'	            head_length=0.4,head_width=0.2
    '-['	            widthB=1.0,lengthB=0.2,angleB=None
    '|-|'	            widthA=1.0,widthB=1.0
    '-|>'	            head_length=0.4,head_width=0.2
    '<-'	            head_length=0.4,head_width=0.2
    '<->'	            head_length=0.4,head_width=0.2
    '<|-'	            head_length=0.4,head_width=0.2
    '<|-|>'	        head_length=0.4,head_width=0.2
    'fancy'	        head_length=0.4,head_width=0.4,tail_width=0.4
    'simple'	        head_length=0.5,head_width=0.5,tail_width=0.2
    'wedge'	        tail_width=0.3,shrink_factor=0.5
    color:Set both the edgecolor and the facecolor.
    facecolor:Set the patch face color.
    edgecolor:Set the patch edge color.
    alpha:透明度(0-1)

'''
ax1.set_ylim(-2, 2)
plt.show()
```
