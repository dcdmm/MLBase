```python
import  numpy as np
import  matplotlib.pyplot as plt
import  scipy.stats as stats
import matplotlib

matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False
fig = plt.figure(figsize=(12, 14))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212)
x1 = np.linspace(-4, 4, 100)
x2 = np.linspace(-0, 8, 100)
norm1 = stats.norm.pdf(x1, loc=0, scale=1)
norm2 = stats.norm.pdf(x2, loc=4, scale=1)

ax1.plot(x1, norm1, color='b', label='1')
ax1.plot(x2, norm2, color='r', label='2')

ax1.fill_between(x1, norm1, 0, where=np.any([x1<=-1.9, x1>=1.9], axis=0), color='blue')
ax1.fill_between(x2, norm2, 0, where=x2<=2, color='red')
ax1.set_ylim(0, 0.45)
ax1.set_yticks(np.linspace(0, 0.45, 10))

ax1.annotate('I\n', xy=(-2.5, 0.016), xytext=(-4, 0.07), color='black',
            fontsize=11,
            arrowprops=dict(edgecolor='b', facecolor='g',
                            arrowstyle='->', alpha=1))
ax1.annotate('II\n', xy=(1.3, 0.016), xytext=(0.5, 0.07), color='black',
            fontsize=11,
            arrowprops=dict(edgecolor='b', facecolor='g',
                            arrowstyle='->', alpha=1))
ax1.legend()

ax2.plot(x1, norm1, color='b', label='1')
ax2.plot(x2, norm2, color='r', label='2')

ax2.fill_between(x1, norm1, 0, where=np.any([x1<=-2.7, x1>=2.7], axis=0), color='blue')
ax2.fill_between(x2, norm2, 0, where=x2<=2.76, color='red')
ax2.set_ylim(0, 0.45)
ax2.set_yticks(np.linspace(0, 0.45, 10))

ax2.annotate('I\n', xy=(-3.5, 0.004), xytext=(-4, 0.07), color='black',
            fontsize=11,
            arrowprops=dict(edgecolor='b', facecolor='g',
                            arrowstyle='->', alpha=1))
ax2.annotate('II\n', xy=(1.3, 0.016), xytext=(0.5, 0.07), color='black',
            fontsize=11,
            arrowprops=dict(edgecolor='b', facecolor='g',
                            arrowstyle='->', alpha=1))
ax2.legend()
```
