### 弗罗贝尼乌斯范数

&emsp;&emsp;奇异值分解也是一种矩阵近似的方法,这个近似是在弗罗贝尼乌斯范数(Frobenius norm)意义下的近似.矩阵的弗罗贝尼乌斯范数是向量范数$L_2$的直接推广,对应着机器学习中的平方损失函数.

<font color='red' size=4>定义:(弗罗贝尼乌斯范数)</font>设矩阵$A \in \mathbb{R}^{m \times n}, A=[a_{ij}]_{m \times n}$,定义矩阵$A$的弗罗贝尼乌斯范数为

$$ ||A||_F = \left( \sum_{i=1}^m \sum_{j=1}^n (a_{ij})^2 \right)^{\frac{1}{2}} $$

<font color='red' size=4>引理:</font>设$A \in \mathbb{R}^{m \times n}$,$A$的奇异值分解为$U \Sigma V^T$,其中$ \Sigma = \mathrm{diag}( \sigma_1, \sigma_2, \cdots,\sigma_n ) $,则

$$ ||A||_F = (\sigma_1^2 + \sigma_2^2 + \cdots + \sigma_n^2)^{1/2} $$

&emsp;&emsp;<font color='red' size=4>证明:</font> 一般地,若$Q$是$m$阶正交矩阵,则有

$$ ||QA||_F = ||A||_F $$

因为

\begin{align}
||QA||_F^2 &= ||(Q_{a_1}, Q_{a_2}, \cdots, Q_{a_n})||_F^2  \\
		   &= \sum_{i=1}^n ||Q a_i ||_2^2 = \sum_{i=1}^n ||a_i||^2_2 = ||A||_F^2
\end{align}

同样,若$P$是$n$阶正交矩阵,则有

$$ ||AP^T||_F = ||A||_F $$

故

$$  ||A||_F = ||U \Sigma V^T|| = ||\Sigma||_F $$

即

$$ ||A||_F = (\sigma_1^2 + \sigma_2^2 + \cdots + \sigma_n^2)^{1/2} $$


### 矩阵的最优近似

&emsp;&emsp;奇异值分解是在平方损失(弗罗贝尼乌斯范数)意义下对矩阵的最优近似,即数据压缩.

<font color='red' size=4>定理:</font>设$A \in \mathbb{R}^{m \times n}$,矩阵的秩$\mathrm{rank} (A) = r$,并设$ \mathcal{M} $为$ \mathbb{R}^{m \times n} $中所有秩不超过$k$的集合,$0 < k < r$,则存在一个秩为$k$的矩阵$ X \in \mathcal{M} $,使得

$$||A - X||_F = \min_{s \in \mathcal{M}} ||A -S ||_F$$

称矩阵$X$为矩阵$A$在弗罗贝尼乌斯范数意义下的最优近似.

<font color='red' size=4>定理:</font>设$A \in \mathbb{R}^{m \times n}$,矩阵的秩$\mathrm{rank} (A) = r$,有奇异值分解$A=U \Sigma V^T$,并设$ \mathcal{M} $为$ \mathbb{R}^{m \times n} $中所有秩不超过$k$的矩阵的集合,$0 < k < r$,若秩为$k$的矩阵$ X \in \mathcal{M} $满足

$$ ||A - X||_F = \min_{s \in \mathcal{M}} ||A -S ||_F \tag{1} $$
则

$$ ||A - X||_F = (\sigma_{k+1}^2 + \sigma_{k+2}^2 + \cdots + \sigma_n^2)^{1/2} \tag{2} $$

特别地,若$A'=U \Sigma' V^T$,其中

\begin{equation}
\Sigma' =
\begin{bmatrix}
\sigma_1 &        &           &   	\\
       & \ddots &           & 	O	\\
	   &	    & \sigma_k  &	    \\
	   &   O    &           &    O
\end{bmatrix} =
\begin{bmatrix}
\Sigma_k & O  \\
   O  & O \\
\end{bmatrix}
\end{equation}

则

$$ ||A - A'||_F = (\sigma_{k+1}^2 + \sigma_{k+2}^2 + \cdots + \sigma_n^2)^{1/2} = \min_{s \in \mathcal{M}} ||A -S ||_F \tag{3} $$

&emsp;&emsp;<font color='red' size=4>证明:</font>令$X \in \mathcal{M}$为满足式(1)的一个矩阵.由于$ A' \in \mathcal{M} $ ,可得

$$ ||A - X||_F \leq || A - A' ||_F =(\sigma_{k+1}^2 + \sigma_{k+2}^2 + \cdots + \sigma_n^2)^{1/2} \tag{4} $$

下面证明

$$ ||A - X||_F \geq (\sigma_{k+1}^2 + \sigma_{k+2}^2 + \cdots + \sigma_n^2)^{1/2} $$

于是式(2)成立.

&emsp;&emsp;设$X$的奇异值分解为$Q \Omega P^T$,其中

\begin{equation}
\Omega =
\begin{bmatrix}
\omega_1 &        &           &   	\\
       & \ddots &           & 	O	\\
	   &	    & \omega_k  &	    \\
	   &   O    &           &    O
\end{bmatrix} =
\begin{bmatrix}
\Omega_k & O  \\
   O  & O \\
\end{bmatrix}
\end{equation}

若令矩阵$B=Q^TAP$,则$A= QBP^T$.由此得到

$$ ||A-X||_F = ||Q (B -\Omega) P^T ||_F = ||B - \Omega||_F \tag{5} $$

&emsp;&emsp;用$\Omega$分块方法对$B$分块

\begin{equation}
B =
\begin{bmatrix}
B_{11} & B_{12} \\
B_{21}      & B_{22} \\
\end{bmatrix}
\end{equation}

其中$B_{11}$是$k \times k$子矩阵,$ B_{12}$是$ k \times (n-k) $子矩阵,$B_{21}$是$  (m -k ) \times k$子矩阵,$ B_{22} $是$ (m -k ) \times (n-k) $子矩阵.可得

\begin{align}
||A -X||_F &= ||B - \Omega||_F^2    \\
		   &= || B_{11} - \Omega_k ||_F^2 + ||B_{12}||_F^2 + ||B_{21}||_F^2 + ||B_{22}||_F^2  \tag{6}
\end{align}

&emsp;&emsp;现证$B_{21}=0, B_{12} =0$.用反证法.若$B_{12} \neq 0$,令

\begin{equation}
Y = Q
\begin{bmatrix}
B_{11} & B_{12} \\
O      & O \\
\end{bmatrix}
P^T
\end{equation}

则$ Y \in \mathcal{M}$,且

$$ ||A - Y||_F^2 = ||B_{21} ||_F^2 + ||B_{22}||^2_F < ||A -X||_F^2 \tag{7} $$

这与$X$的定义式(4)矛盾,证明了$B_{12} =0$.同样可证$B_{21}=0$.于是

$$ ||A-X||_F^2 = ||B_{11} - \Omega_k||_F^2 + ||B_{22}||_F^2 \tag{8} $$

&emsp;&emsp;再证$ B_{11} = \Omega_k $.为此令

\begin{align}
Z =
\begin{bmatrix}
B_{11} & O \\
O      & O \\
\end{bmatrix}
\end{align}

则$Z \in \mathcal{M}$,且

$$ ||A-Z||_F^2 = ||B_{22}||_F^2 \leq ||B_{11} -\Omega_k||_F^2 + ||B_{22}||_F^2 = ||A-X||_F^2 \tag{9} $$

由式(4)知,$ ||B_{11} - \Omega_k||_F^2 = 0 $,即$B_{11}=\Omega_k$

&emsp;&emsp;最后看$B_{22}$.若$(m -k ) \times (n-k)$子矩阵$B_{22}$有奇异值分解$U_1 \Lambda V_1^T$,则

$$ ||A_X||_F = ||B_{22}||_F =|\Lambda|_F $$

&emsp;&emsp;证明$\Lambda$的对角线元素为$A$的奇异值.为此,令

\begin{equation}
U_1 =
\begin{bmatrix}
I_k & O \\
O      & U_1 \\
\end{bmatrix}
V_2 =
\begin{bmatrix}
I_k & O \\
O      & V_1 \\
\end{bmatrix}
\end{equation}

其中$I_k$是$k$阶单位矩阵,$U_2, V_2$的分块矩阵与$B$的分块一致.注意到$B$及$B_{22}$的奇异值分解,即得

\begin{equation}
U_2^T Q^T AP V_2 =
\begin{bmatrix}
\Omega_k & O \\
O      & \Lambda \\
\end{bmatrix} \tag{10}
\end{equation}

\begin{equation}
A=(QU_2)
\begin{bmatrix}
\Omega_k & O \\
O      & \Lambda \\
\end{bmatrix}
(PV_2)^T \tag{11}
\end{equation}

由此可知$\Lambda$的对角线元素为$A$的奇异值.故有

$$ ||A - X||_F = || \Lambda ||_F\geq (\sigma_{k+1}^2 + \sigma_{k+2}^2 + \cdots + \sigma_n^2)^{1/2} $$

于是证明了

$$ ||A - X||_F = (\sigma_{k+1}^2 + \sigma_{k+2}^2 + \cdots + \sigma_n^2)^{1/2} = ||A-A'||_F  $$

&emsp;&emsp;此定理表明,在秩序不超过$k$的$m \times n$矩阵的集合中,存在矩阵$A$的弗罗贝尼乌斯范数意义下的最优近似矩阵$X$.$A'=U\Sigma'V^T$是达到最优值的一个矩阵.

&emsp;&emsp;前面定义了矩阵的紧奇异值分解和截断奇异值分解.事实上紧奇异值分解是在弗罗贝尼乌斯范数意义下的无损压缩.截断奇异值分解时有损压缩.截断奇异值分解得到的矩阵的秩为$k$,通常远小于原始矩阵的秩$k$,所以是由低秩矩阵实现了对原始矩阵的压缩.
