```python
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark') 
fig = plt.figure(figsize=(10, 15))
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)

with sns.axes_style('whitegrid'):
    # ,with
    ax1 = plt.subplot(311)
    sinplot()
    
ax2 = plt.subplot(312)
sinplot()

ax3 = plt.subplot(313)
sinplot()
```
