```python
from PIL import Image
from torchvision import transforms
```

```python
img = Image.open('test_picture/1yuan.jpg').convert('RGB') # 读取图片(RGB模式)
img
```

```python
img.size # 图片大小
```

```python
trans1 = transforms.Resize((224, 224)) # Resize(缩放) the input PIL Image to the given size
trans_img1 = trans1(img)
trans_img1
```

```python
trans_img1.size
```

```python
trans2 = transforms.Resize((100, 1000))
trans_img2 = trans2(img)
trans_img2
```
