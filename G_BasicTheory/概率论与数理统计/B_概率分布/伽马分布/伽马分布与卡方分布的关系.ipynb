<font color='red' size=4>定义:</font>
&emsp;&emsp;已知伽马分布的概率密度函数为:    
$$ f(x|\alpha ,\beta) = \frac{1}{\Gamma(\alpha)\beta^{\alpha}} x^{\alpha -1} e^{-x/\beta}, 0<x<+\infty,\alpha > 0, \beta >0$$  
若令$ \alpha = p/2 $, 其中 $ p$ 为整数且$ \beta =2$,则伽马概率密度函数变为  
$$ f(x|p) = \frac{1}{\Gamma(p/2)2^{p/2}}x^{p/2 -1} e^{-x/2}, 0<x<\infty $$  
此即自由度为$ p $ 的$ \chi^2 $概率密度函数,记为$ \chi_p^2 $

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

# PLOTTING CONFIG
%matplotlib inline
style.use('fivethirtyeight')

x = np.linspace(0, 10, 100)
n = range(1, 11)

plt.figure(figsize=(14, 7), dpi=100)
ax = plt.subplot(111)
for i in n:
    show = r' n=' + str(i)
    ax.plot(x, stats.chi2.pdf(x, df=i), label=show)
    ax.legend()
    
```
