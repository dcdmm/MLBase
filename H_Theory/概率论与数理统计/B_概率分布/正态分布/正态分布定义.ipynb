<font color='red' size='4'>定义:</font>  

<center>

\begin{equation}
    f(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^{2}}}
    e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}
    \quad \mu 是均值,
    \sigma 是标准差  
\end{equation}   
</center>

* 期望和方差:$$ EX =u, VarX=\sigma^2$$
* 矩母函数:$$ M_X(t)=e^{\mu t + \sigma^2 t^2 /2 } $$

<font color='red' size='4'>性质:</font>  
1. 一般把正态分布记为:$N(\mu,\sigma^2)$,特别地,当$\mu=0,\sigma=1$时,此时的正态分布为标准正态分布,
记为$N(0,1).$ $N(0,1)$的密度函数通常用$ \phi(x) $表示,任何具有正态分布$N(\mu,\sigma^2)的随机变量X$都可以通
过简单的标准化而成为标准正态分布随机变量,即$z=\frac{X-\mu}{\sigma}$,标准正态随机变量也记为$z$统计量.  
2. 设随机变量$X \sim N(0,1),对于给定的\alpha,0<\alpha<1$称满足条件

<center>

\begin{equation}
    P\{X>Z_{\alpha}\} =\int_{Z_{\alpha}}^{+\infty} \psi(x)dx=\alpha
\end{equation}

</center>

的点$Z_{\alpha}$为标准正态分布上的$\alpha$分位点


```python
%matplotlib inline
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('classic')
fig = plt.figure(figsize=(14, 7), dpi=100)
ax  = plt.subplot(111)

# PDF(pdf):概率密度函数
ax.plot(np.linspace(-4, 4, 100), 
         stats.norm.pdf(np.linspace(-4, 4, 100)) / np.max(stats.norm.pdf(np.linspace(-3, 3, 100))),
        )
ax.fill_between(np.linspace(-4, 4, 100),
                 stats.norm.pdf(np.linspace(-4, 4, 100)) / np.max(stats.norm.pdf(np.linspace(-3, 3, 100))),
                 alpha=.15,
                )

# CDF(cdf):累计分布函数
ax.plot(np.linspace(-4, 4, 100), 
         stats.norm.cdf(np.linspace(-4, 4, 100)),
        )

ax.text(x=-1.5, y=.7, s="pdf (normed)", rotation=65, alpha=.75, fontweight="bold", color="#008fd5")
ax.text(x=-.4, y=.5, s="cdf", rotation=55, alpha=.75, fontweight="bold", color="#fc4f30")

ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'red', linewidth = 1.3, alpha = .7)

ax.text(x = -2, y = -0.1, s = "Normal Distribution - Overview",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = -5, y = 1.01, 
         s = 'Depicted below are the normed probability density function (pdf) and the cumulative density\nfunction (cdf) of a normally distributed random variable $ y \sim \mathcal{N}(\mu,\sigma) $, given $ \mu = 0 $ and $ \sigma = 1$.',
         fontsize = 19, alpha = .85)
```
