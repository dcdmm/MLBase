1. get_xaxis().set_major_locator,get_xaxis().set_minor_locator
2. get_xaxis().set_major_formatter,get_xaxis().set_minor_formatter

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

t = np.arange(0.0, 100, 1)
s = np.sin(0.1 * np.pi * t) * np.exp(-t * 0.01)
fig = plt.figure(figsize=(10, 7))
ax = plt.subplot(111)
ax.plot(t, s, marker='*', linestyle='--')

xmajorLocator = MultipleLocator(40)
ax.get_xaxis().set_major_locator(xmajorLocator)  # x40
xmajorFormatter = FormatStrFormatter('%.1f')
ax.get_xaxis().set_major_formatter(xmajorFormatter)  # x1

# xminorLocator = MultipleLocator(5) 
# ax.get_xaxis().set_minor_locator(xminorLocator) #x5
# xminorFormatter = FormatStrFormatter('%.2f') 
# ax.get_xaxis().set_minor_formatter(xminorFormatter) #x2

ax.grid(True, which='both', axis='x')

ymajorLocator = MultipleLocator(0.3)
ax.get_yaxis().set_major_locator(ymajorLocator)  # y0.3
ymajorFormatter = FormatStrFormatter('%.1f')
ax.get_yaxis().set_major_formatter(ymajorFormatter)  # y1

yminorLocator = MultipleLocator(0.1)
ax.get_yaxis().set_minor_locator(yminorLocator)  # y0.1
yminorFormatter = FormatStrFormatter('%.2f')
ax.get_yaxis().set_minor_formatter(yminorFormatter)  # y2

ax.grid(True, which='both', axis='y')



# ax.get_xaxis().set_major_locator(plt.NullLocator()) 
# (,,),:ax.get_xaxis().set_ticks([])ax.set_xticks([])

# ax.get_xaxis().set_major_formatter(plt.NullFormatter()) 
# ,:#ax.get_xaxis().set_ticklabels([])

# ax.get_xaxis().set_minor_locator(plt.NullLocator()) 
# (,,),:ax.get_xaxis().set_ticks([], monor=True)ax.set_xticks([], monor=True)
# ax.get_xaxis().set_minor_formatter(plt.NullFormatter()) 
# ,:#ax.get_xaxis().set_ticklabels([], monor=True)

# ax.get_xaxis().set_visible(False) # x
# plt.setp(frame.get_yaxis(), visible=False) #
```
