```python
from PIL import Image
from torchvision import transforms
```

```python
img = Image.open('../test_picture/1yuan.jpg').convert('RGB') # 读取图片(RGB模式)
img
```

```python
norm_mean = [0.4, 0.8, 0.99]
norm_std = [0.5, 0.80, 0.9]

trans_tensor = transforms.ToTensor()
tensor_img = trans_tensor(img) # 先转换为tensor
```

```python
trans_normal = transforms.Normalize(mean=norm_mean, std=norm_std)
normal_img = trans_normal(tensor_img) # output[channel] = (input[channel] - mean[channel]) / std[channel]
normal_img
```

```python
trans_img = transforms.ToPILImage()
trans_img(normal_img)
```
