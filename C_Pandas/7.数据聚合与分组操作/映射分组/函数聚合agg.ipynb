```python
import numpy as np
import pandas as pd
```

```python
tips = pd.read_csv('tips.csv')
# Add tip percentage of total bill
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips[:6]
```

```python
group_df = tips.groupby([tips['day'], tips['smoker']])
group_df  # DataFrameGroupBy对象
```

```python
for key, group in group_df:
    print(key)
    print(group)
```

```python
group_ser = group_df['tip_pct']  # 等价与:tips['tip_pct'].groupby([tips['day'], tips['smoker]])
group_ser  # SeriesGroupBy对象
```

```python
for m, n in group_ser:
    print(m)
    print(n)
```

```python
group_ser.agg('mean')  # Series
```

```python
# 如果传输的是函数或函数名的列表,则columns为:
# * 内置的聚合函数为其字符串名
# * 自定义函数为函数名
# * 匿名函数为'<lambda_0>', '<lambda_1>', '<lambda_2>'......
group_ser.agg(['mean'])  # DataFrame
```

```python
def peak_to_peak(arr):
    print(arr.name)  # 每个分组下每列为一个group(不含分组列)
    return arr.max() - arr.min()


group_ser.agg(['mean', 'std', peak_to_peak, lambda x: x.sum(), lambda x: x.mean()])
```

```python
# 指定列进行聚合(对tip列进行np.max聚合,size列进行sum聚合)
# 其他列不会参与聚合
group_df.agg({'tip': [np.max, peak_to_peak], 'size': 'sum', 'tip_pct': peak_to_peak})
```
