```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

# PLOTTING CONFIG
%matplotlib inline
style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(12, 6), dpi=100)
ax = plt.subplot(111)

# A = B = 1
ax.plot(np.linspace(0, 1, 200), 
         stats.beta.pdf(np.linspace(0, 1, 200), a=1, b=1))
ax.fill_between(np.linspace(0, 1, 200),
                 stats.beta.pdf(np.linspace(0, 1, 200), a=1, b=1),
                 alpha=.15)

# A = B = 10
ax.plot(np.linspace(0, 1, 200), 
         stats.beta.pdf(np.linspace(0, 1, 200), a=10, b=10))
ax.fill_between(np.linspace(0, 1, 200),
                 stats.beta.pdf(np.linspace(0, 1, 200), a=10, b=10),
                 alpha=.15)

# A = B = 100
ax.plot(np.linspace(0, 1, 200), 
         stats.beta.pdf(np.linspace(0, 1, 200), a=100, b=100))
ax.fill_between(np.linspace(0, 1, 200),
                 stats.beta.pdf(np.linspace(0, 1, 200), a=100, b=100),
                 alpha=.15)

# LEGEND
ax.text(x=0.1, y=1.45, s=r"$ \alpha = 1, \beta = 1$", alpha=1, weight="bold", color="#008fd5")
ax.text(x=0.325, y=3.5, s=r"$ \alpha = 10, \beta = 10$", rotation=35, alpha=1, weight="bold", color="#fc4f30")
ax.text(x=0.4125, y=8, s=r"$ \alpha = 100, \beta = 100$", rotation=80, alpha=1, weight="bold", color="#e5ae38")


# TICKS
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

# TITLE, SUBTITLE & FOOTER
ax.text(x = -.1, y = 14.39, s = r"Beta Distribution - constant $\frac{\alpha}{\beta}$, varying $\alpha + \beta$",
               fontsize = 26, weight = 'bold', alpha = 1)
ax.text(x = -.1, y = 12.3, 
         s = 'Depicted below are three beta distributed random variables with '+ r'equal $\frac{\alpha}{\beta} $ and varying $\alpha+\beta$'+'.\nAs one can see the sum of ' + r'$\alpha + \beta$ (mainly) sharpens the distribution (the bigger the sharper).',
         fontsize = 19, alpha = 1)
```

```python
plt.figure(figsize=(15, 8), dpi=100)
ax1 = plt.subplot(111)

# A / B = 1/3
ax1.plot(np.linspace(0, 1, 200), 
         stats.beta.pdf(np.linspace(0, 1, 200), a=25, b=75))
ax1.fill_between(np.linspace(0, 1, 200),
                 stats.beta.pdf(np.linspace(0, 1, 200), a=25, b=75),
                 alpha=.15,
                )

# A / B = 1
ax1.plot(np.linspace(0, 1, 200), 
         stats.beta.pdf(np.linspace(0, 1, 200), a=50, b=50))
ax1.fill_between(np.linspace(0, 1, 200),
                 stats.beta.pdf(np.linspace(0, 1, 200), a=50, b=50),
                 alpha=.15)

# A / B = 3
ax1.plot(np.linspace(0, 1, 200), 
         stats.beta.pdf(np.linspace(0, 1, 200), a=75, b=25))
ax1.fill_between(np.linspace(0, 1, 200),
                 stats.beta.pdf(np.linspace(0, 1, 200), a=75, b=25),
                 alpha=.15)

# LEGEND
ax1.text(x=0.15, y=5, s=r"$ \alpha = 25, \beta = 75$", rotation=80, alpha=1, weight="bold", color="#008fd5")
ax1.text(x=0.39, y=5, s=r"$ \alpha = 50, \beta = 50$", rotation=80, alpha=1, weight="bold", color="#fc4f30")
ax1.text(x=0.65, y=5, s=r"$ \alpha = 75, \beta = 25$", rotation=80, alpha=1, weight="bold", color="#e5ae38")


# TICKS
ax1.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax1.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

# TITLE, SUBTITLE & FOOTER
ax1.text(x = -.1, y = 11.75, s = r"Beta Distribution - constant $\alpha + \beta$, varying $\frac{\alpha}{\beta}$",
               fontsize = 26, weight = 'bold', alpha = .75)
ax1.text(x = -.1, y = 10, 
         s = 'Depicted below are three beta distributed random variables with '+ r'equal $\alpha+\beta$ and varying $\frac{\alpha}{\beta} $'+'.\nAs one can see the fraction of ' + r'$\frac{\alpha}{\beta} $ (mainly) shifts the distribution ' + r'($\alpha$ towards 1, $\beta$ towards 0).',
         fontsize = 19, alpha = .85)
```
