```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan.jpg').convert('RGB') # 读取图片(RGB模式)
img
```

```python
trans = transforms.CenterCrop(size=(224, 200)) # Crops the given PIL Image at the center.
'''
size (sequence or int) –
    Desired output size of the crop. If size is an int instead of sequence like (h, w), a square crop (size, size) is made.
'''
trans_img = trans(img)
trans_img
```

```python
trans_img.size
```

```python
# 新增区域默认填充为0(即为黑色)
trans1 = transforms.CenterCrop(size=500)
trans_img1 = trans1(img)
trans_img1
```
