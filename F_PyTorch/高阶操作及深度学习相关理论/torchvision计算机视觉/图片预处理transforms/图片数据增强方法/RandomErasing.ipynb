```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # 读取图片(RGB模式)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans_tensor = transforms.ToTensor()
img_tensor = trans_tensor(img_resize) # 先转换为tensor
```

```python
trans_img = transforms.ToPILImage()
```

```python
# Randomly selects a rectangle region in an image and erases its pixels
trans = transforms.RandomErasing(p=0.9, scale=(0.02, 0.33), # (遮挡区域面积/原始图像面积)的范围
                                 ratio=(0.3, 3.3)) # 遮挡区域长宽比的范围
'''
p – probability that the random erasing operation will be performed.
scale – range of proportion of erased area against input image.
ratio – range of aspect ratio of erased area.
'''
trans_img(trans(img_tensor)) # ★★★★★输入输出均为tensor
```

```python
trans_img(trans(img_tensor))
```

```python
trans1 = transforms.RandomErasing(p=0.98, scale=(0.02, 0.33), ratio=(0.3, 3.3),
                                  value=(1, 0, 0))
'''
value –
    erasing value. Default is 0.(黑色)
    If a single int, it is used to erase all pixels.(0~1)
    If a tuple of length 3, it is used to erase R, G, B channels respectively.(0~1)
    If a str of ‘random’, erasing each pixel with random values.
'''
trans_img(trans1(img_tensor))
```

```python
trans1 = transforms.RandomErasing(p=0.98, scale=(0.02, 0.33), ratio=(0.3, 3.3),
                                  value='random')
trans_img(trans1(img_tensor))
```
