如果变量$X$是均匀分布的，则密度函数可以表示为:  
<img src='../../../../Other/img/均匀分布.png' style="width:400px;height:300px;float:bottom">

* 期望和方差为:$$E(X) = \frac{a+b}{2} \quad Var(X) = \frac{(b-1)^2}{12}$$

```python
# IMPORTS
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('fivethirtyeight')
fig = plt.figure(figsize=(14, 7), dpi=100)
ax  = plt.subplot(111)


# PDF(pdf):概率密度函数
ax.plot(np.linspace(-4, 4, 100), 
         stats.uniform.pdf(np.linspace(-4, 4, 100)) )
ax.fill_between(np.linspace(-4, 4, 100),
                 stats.uniform.pdf(np.linspace(-4, 4, 100)), alpha=.15)

# CDF
ax.plot(np.linspace(-4, 4, 100), 
         stats.uniform.cdf(np.linspace(-4, 4, 100)))

ax.text(x=-1.5, y=.7, s="pdf (normed)", rotation=65, alpha=.75, weight="bold", color="#008fd5")
ax.text(x=-.4, y=.5, s="cdf", rotation=55, alpha=.75, weight="bold", color="#fc4f30")

ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = -5, y = 1.25, s = "Nniform Distribution - Overview",
               fontsize = 26, weight = 'bold', alpha = .75)
```
