### 神经网络结构

1. 单隐层神经网络
2. 激活函数为Sigmoid函数
3. 损失函数为均方误差损失函数
4. 输入层,隐层,输出层神经元个数为d,q,s


输入层输出:$\mathbf{x}.shape=(d+1,1)$

\begin{equation}
\mathbf{x} =
\begin{bmatrix}
x_{1} \\
x_{2} \\
\vdots \\
x_{d} \\
1
\end{bmatrix}
\end{equation}

输入层到隐层权重矩阵:$W^1.shape=(q,d+1)$

\begin{equation}
W^1=
\begin{bmatrix}
w^1_{11} & \dots  & w^1_{1d} & b^1_1    \\
         & \ddots & \vdots  &  \vdots \\
w^1_{q1} & \dots  & w^1_{qd} & b^1_q
\end{bmatrix}
\end{equation}

隐层输入:$\mathbf{m}.shape=(q,1)$

$$ \mathbf{m} = W^1\mathbf{x} $$

隐层输出:$\mathbf{n}.shape=(q+1,1)$

\begin{equation}
\mathbf{n} =
\begin{bmatrix}
\mathrm{sigma}(m_{1}) \\
\mathrm{sigma}(m_{2}) \\
\vdots \\
\mathrm{sigma}(m_{q}) \\
1
\end{bmatrix}
\end{equation}

隐层到输出层权重矩阵:$W^2.shape=(s,q+1)$

\begin{equation}
W^2=
\begin{bmatrix}
w^2_{11} & \dots  & w^2_{1q} & b^2_1 \\
         & \ddots & \vdots & \vdots \\
w^2_{s1} & \dots  & w^2_{sq} & b^2_s
\end{bmatrix}
\end{equation}

输出层输入:$\mathbf{z}.shape=(s,1)$

$$ \mathbf{z} = W^2 \mathbf{n} $$

输出层输出:$\mathbf{\hat{y}}.shape=(s,1)$

$$ \mathbf{\hat{y}} = \mathrm{sigmoid}(\mathbf{z}) $$

真实标签:$\mathbf{y}.shape=(s,1)$


### 根据公式计算梯度


\begin{align}
\frac{ \partial \mathcal{L}(y_i, \hat{y}_i )}{\partial W^2} &=  \boldsymbol{\delta}^2 \left(\mathbf{o}^{1}\right)^T  \\
		&=  \boldsymbol{\delta}^2 \mathbf{n}^T \\
		&= A^2(\mathbf{\hat{y} - \mathbf{y}})  \mathbf{n}^T \\
		&=
\begin{bmatrix}
\hat{y}_1(1-\hat{y}_1)(\hat{y_1} - y_1) \\
\hat{y}_2(1-\hat{y}_2)(\hat{y_2} - y_2) \\
\vdots \\
\hat{y}_i(1-\hat{y}_i)(\hat{y_i} - y_i) \\
\vdots \\
\hat{y}_1(1-\hat{y}_s)(\hat{y_s} - y_s) \\
\end{bmatrix}
\mathbf{n}^T
\end{align}



\begin{align}
\frac{ \partial \mathcal{L}(y_i, \hat{y}_i )}{\partial W^1} &=   \boldsymbol{\delta}^1 \left(\mathbf{o}^{0}\right)^T  \\
			&= \boldsymbol{\delta}^1 \mathbf{x}^T \\
			&=A^1 (W^2)^T \boldsymbol{\delta}^2 \mathbf{x}^T \\
			&=A^1 (W^2)^T  A^2(\mathbf{\hat{y} - \mathbf{y}}) \mathbf{x}^T \\
			&=A^1 (W^2)^T
\begin{bmatrix}
\hat{y}_1(1-\hat{y}_1)(\hat{y_1} - y_1) \\
\hat{y}_2(1-\hat{y}_2)(\hat{y_2} - y_2) \\
\vdots \\
\hat{y}_i(1-\hat{y}_i)(\hat{y_i} - y_i) \\
\vdots \\
\hat{y}_1(1-\hat{y}_s)(\hat{y_s} - y_s) \\
\end{bmatrix}
\mathbf{x}^T
\end{align}
