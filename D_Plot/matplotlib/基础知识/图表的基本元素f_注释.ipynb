```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.set_title(r'$\alpha_i > \beta_i$', fontsize=20)

'''
x, y : scalars
    The position to place the text. By default, this is in data coordinates. The coordinate system can be changed using the transform parameter.

s : str
    The text.
'''
text0 = ax.text(0.5, 0,  # 注释文本坐标指的是该点的(xticks, yticks)
                r'period is:$\mathcal{A}\mathrm{sin}(2 \omega t)$')
text0.set_color('red')
text0.set_fontsize(20)
text0.set_alpha(0.9)
text0.set_rotation(45)
text0.set_fontfamily('Impact')
text0.set_rotation_mode('anchor')
# 下面将注释文本的左上点的坐标设置为(0.5, 0)
text0.set_va('top')
text0.set_ha('left')

text1 = ax.text(0.5, 0,  # 注释点坐标指的是该点的(xticks, yticks)
                r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',
                rotation_mode=None,
                ha='right', va='bottom')
text1.set_color('g')

ax.set_xlabel('time (s)')
ax.set_ylabel('volts (mV)')
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax1.plot(t, s, lw=2)

'''
text:str
    The text of the annotation.

xy:(float, float)
    The point (x, y) to annotate. The coordinate system is determined by xycoords.

xytext:(float, float), default: xy
    The position (x, y) to place the text at. The coordinate system is determined by textcoords.

arrowprops : dict, optional
    The properties used to draw a FancyArrowPatch arrow between the positions xy and xytext.

    If arrowprops does not contain the key 'arrowstyle' the allowed keys are:
    Key	                Description
    width	            The width of the arrow in points
    headwidth	        The width of the base of the arrow head in points
    headlength	        The length of the arrow head in points
    shrink	            Fraction of total length to shrink from both ends
    ?	                Any key to matplotlib.patches.FancyArrowPatch

    If arrowprops contains the key 'arrowstyle' the above keys are forbidden. The allowed values of 'arrowstyle' are:
    Name	            Attrs
    '-'	                None
    '->'	            head_length=0.4,head_width=0.2
    '-['	            widthB=1.0,lengthB=0.2,angleB=None
    '|-|'	            widthA=1.0,widthB=1.0
    '-|>'	            head_length=0.4,head_width=0.2
    '<-'	            head_length=0.4,head_width=0.2
    '<->'	            head_length=0.4,head_width=0.2
    '<|-'	            head_length=0.4,head_width=0.2
    '<|-|>'	            head_length=0.4,head_width=0.2
    'fancy'	            head_length=0.4,head_width=0.4,tail_width=0.4
    'simple'	        head_length=0.5,head_width=0.5,tail_width=0.2
    'wedge'	            tail_width=0.3,shrink_factor=0.5

    color:Set both the edgecolor and the facecolor.
    facecolor:Set the patch face color.
    edgecolor:Set the patch edge color.
    alpha:透明度(0-1)
'''
annotate = ax1.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
                        arrowprops=dict(edgecolor='r', facecolor='g',
                                        arrowstyle='<|-', alpha=0.3))  # 注释点坐标指的是该点的(xticks, yticks)
annotate.set_color('red')
annotate.set_fontsize(20)
annotate.set_alpha(0.8)

ax1.set_ylim(-2, 2)
plt.show()
```
