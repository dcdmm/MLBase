```python
import tensorflow as tf
```

```python
# Creates a Dataset comprising lines from one or more text files.
titanic_lines = tf.data.TextLineDataset(['data/train.csv',
                                         'data/eval.csv'])
titanic_lines
```

```python
def data_func(line):
    line = tf.strings.split(line, sep=",")
    return line
```

```python
"""
执行过程解析:
1. train.csv执行skip(1).map(data_func)
2. train.csv执行完后,eval.csv执行skip(1).map(data_func)
"""
# 数据处理方法含义与tf.data.Dataset相同
titanic_data = titanic_lines.skip(1).map(data_func)
```

```python
for line in titanic_data.as_numpy_iterator():
    print(line)
```
