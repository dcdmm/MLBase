```python
import tensorflow as tf
```

### 参考torch.utils.data.TensorDataset

```python
x = tf.linspace(0.0, 0.1, 100)
y = tf.reshape(tf.linspace(1, 200, 200), (100, 2))
z = tf.range(1000, 1100)
```

```python
# Creates a Dataset whose elements are slices of the given tensors.
for i in tf.data.Dataset.from_tensor_slices(x).as_numpy_iterator():
    print(i)
```

```python
dataset = tf.data.Dataset.from_tensor_slices((x, y, z))
print(dataset)
print(len(dataset))
```

```python
dataset._tensors
```

```python
for i in dataset.as_numpy_iterator():
    print(i)
```

```python
dataset = tf.data.Dataset.from_tensor_slices({"x": x, "y": y, "z": z})
dataset
```

```python
dataset._tensors
```

```python
for i in dataset.as_numpy_iterator():
    print(i)
```
