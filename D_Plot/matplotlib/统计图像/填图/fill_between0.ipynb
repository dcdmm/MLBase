```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 7))
x = np.linspace(0, 100, 1000)
y1 = np.sin(x)
y2 = np.sin(2 * x)
y3 = np.abs(np.sin(x))

axes[0].fill_between(x, y1, y2, color='b', alpha=0.5, label='area')  # Fill the area between two horizontal curves
axes[0].legend(loc='best')

axes[1].fill_between(x, y3, 0.3, color='y')
axes[1].plot(x[0], color='r', linewidth=3)
axes[1].plot([x[0], x[-1]], [0.3, 0.3], c='r', alpha=0.8, linewidth=1)
```
