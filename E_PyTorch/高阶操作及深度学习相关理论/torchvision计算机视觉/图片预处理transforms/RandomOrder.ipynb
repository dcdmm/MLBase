```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('test_picture/1yuan_1.jpg').convert('RGB')  # (RGB)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans0 = transforms.RandomRotation(degrees=(20, 20))
trans1 = transforms.Pad(padding=52)
# Apply a list of transformations in a random order
trans_order = transforms.RandomOrder([trans0, trans1])  # transformations
```

```python
trans_order(img_resize)  # trans0 --> trans1
```

```python
trans_order(img_resize)  # trans1 --> trans0
```

```python
trans_order(img_resize)
```
