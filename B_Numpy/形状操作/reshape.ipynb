```python
import  numpy as np
```

```python
a = np.arange(64)
a
```

```python
a.shape # ndim=1
```

```python
a.reshape(2, 32)
```

```python
a # a不变
```

```python
b = np.reshape(a, (8, 8))
# b = a.reshape(8, 8) # 与上等价
b
```

```python
b.reshape(-1, 1) # 等价于b.reshape(64, 1);ndim=2
```

```python
b.reshape(-1, ) # 等价于b.reshape(64, );ndim=1
```

```python
b.flatten().reshape(8, 8) - b # 不变
```

```python
arr = np.random.randn(2, 7, 7)
arr
```

```python
arr1 = arr.reshape(2, 7*7) # 向量化数据
arr1
```

```python
arr2 = arr1.reshape(2, 7, 7) # 按原顺序恢复成原shape
arr2 - arr
```

```python
mul_arr = np.arange(120).reshape((2, 3, 4, 5))
mul_arr
```

```python
'''
reshape规则:
mul_arr[0, 0, 0, 0] -> mul_arr[0, 0, 0, 1] ...-> mul_arr[0, 0, 0, -1]
mul_arr[0, 0, 1, 0] -> mul_arr[0, 0, 1, 1] ...-> mul_arr[0, 0, 1, -1]
...  ... ...
mul_arr[0, 0, -1, 0] -> mul_arr[0, 0, -1, 1] ...-> mul_arr[0, 0, -1, -1]
mul_arr[0, 1, 0, 0] -> mul_arr[0, 1, 0, 1] ...-> mul_arr[0, 1, 0, -1]
... ... ...
mul_arr[0, -1, -1, 0] -> mul_arr[0, -1, -1, 1] ...-> mul_arr[0, -1, -1, -1]
mul_arr[1, 0, 0, 0] -> mul_arr[0, 0, 0, 1] ...-> mul_arr[0, 0, 0, -1]
... ... ...
mul_arr[-1, -1, -1, 0] -> mul_arr[-1, -1, -1, 1] ...-> mul_arr[-1, -1, -1, -1]
'''
mul_arr.reshape(10, 12)
```
