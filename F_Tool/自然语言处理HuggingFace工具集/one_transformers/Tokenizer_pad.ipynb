```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained(pretrained_model_name_or_path='bert-base-chinese')
```

```python
sents = [
    '选择珠江花园的原因就是方便。',
    '笔记本的键盘确实爽。',
    '房间太小。其他的都一般。',
    '今天才知道这书还有第6卷,真有点郁闷.',
    '机器背面似乎被撕了张什么标签，残胶还在。',
]
```

```python
r = tokenizer(sents)
print(r)  # type: Dict[str, List[EncodedInput]]
for i in r['input_ids']:
    print(len(i))
```

```python
rp1 = tokenizer.pad(r,
                    # 参数含义参考Tokenizer __call__方法
                    padding=True, max_length=None, padding_side=None)
print(rp1)
for i in rp1['input_ids']:
    print(len(i))
```

```python
r2 = []
for s in sents:
    r_ = tokenizer(s)
    r2.append(r_)
print(r2)  # type: List[Dict[str, EncodedInput]]
rp2 = tokenizer.pad(r2)
print(rp2)
for i in rp2['input_ids']:
    print(len(i))
```
