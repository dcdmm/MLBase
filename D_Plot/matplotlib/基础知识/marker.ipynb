## matplotlib.markers
包含marker参数的对象有:
* plot
* scatter

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 15)
y = np.sin(x)
```

```python
plt.scatter(x, y, marker='o')
```

```python
plt.scatter(x, y, marker='.')
```

```python
plt.scatter(x, y, marker='*')
```

```python
plt.scatter(x, y, marker='_')
```

```python
plt.scatter(x, y, marker='v')
```

```python
plt.figure(figsize=(20, 20))
# 还可以为markdown数学符号
for i in range(len(x)):
    plt.scatter(x[i], y[i], marker="$ sin %f $" % x[i], linewidths=0, s=4000, color='r', alpha=1)
```
