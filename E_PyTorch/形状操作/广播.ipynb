```python
import torch
import copy

x = torch.randn(4, 32, 32, 3)
x.shape
```

```python
# 完全类似numpy 广播
(x + 4).shape # 等价于x + torch.tensor(4).repeat(x.shape)
```

```python
(x * 4).shape # 等价于x * torch.tensor(4).repeat(x.shape)
```

```python
(x ** 3).shape
```

```python
(x == 4).shape
```

```python
(x + torch.randn(3)).shape # [4, 32, 32, 3] broadcasting [3, ] --> [4, 32, 32, 3]
```

```python
(x + torch.randn(1, 3)).shape # [4, 32, 32, 3] boradcasting [1, 3] --> [4, 32, 32, 3]
```

```python
(x + torch.randn(32, 32, 1)).shape # [4, 32, 32, 3] boradcasting [32, 32, 1] --> [4, 32, 32, 3]
```

```python
(x + torch.randn(4, 1, 1, 1)).shape # [4, 32, 32, 3] boradcasting [4, 1, 1, 1] --> [4, 32, 32, 3]
```

```python
a = torch.ones(   [1, 12, 8, 1, 4, 1])
b = torch.zeros([5, 4, 1, 1, 3, 1, 4])
# [1, 12, 8, 1, 4, 1] boradcasting [5, 4, 1, 1, 3, 1, 4] --> [5, 4, 12, 8, 3, 4, 4]

(a + b).shape
```

```python
# 显式的使用broadcast_tensors函数完成此操作(不推荐)
```

```python
a1, b1 = torch.broadcast_tensors(copy.deepcopy(a), copy.deepcopy(b))
(a + b) - (a1 + b1)
```
