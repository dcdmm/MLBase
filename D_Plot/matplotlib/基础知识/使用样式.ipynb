```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
import  matplotlib.style as psl #使用样式之后,整个ipynb文件都会使用该样式

print(psl.available) # 查看样式列表
psl.use('fivethirtyeight') # 使用样式(范围:整个ipynb文件)
fig, ax = plt.subplots()
df = pd.DataFrame(np.random.randn(50, 4), columns=list('abcd'))
df = df.cumsum()
for i in df.columns:
    ax.plot(df[i], label=i+'d', linewidth=9) # label:Set the label to s for auto legend
ax.legend(loc='best')
ax.grid()
ax.set_xticks([])
```
