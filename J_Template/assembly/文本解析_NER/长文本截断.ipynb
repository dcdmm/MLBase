```python
def split_sentence(text, spo_list, sign='。', max_len=510):
    """长句子根据标点符号(sign)截断"""
    asf = [0, 0]

    def sen_position(data):
        """句子长度之和不超过max_len的分组为同一组"""
        asf[1] += data
        if asf[1] > max_len:
            asf[1] = data
            asf[0] += 1
        return pd.Series([data, asf[0]])

    def sen_join(data):
        return pd.Series([''.join(data['text'].tolist()), data['txt_len'].sum()], index=['join_text', 'join_text_len'])

    text_s = text.split(sign)  # split函数结果不包含分割字符sign
    if text_s[-1] == '':
        text_ser = pd.Series(dict([(i + sign, len(i) + 1) for i in text_s[:-1]]))  # 最后一句以sign结尾时
    else:
        text_tup = [(i + sign, len(i) + 1) for i in text_s]
        text_tup[-1] = (text_tup[-1][0][:-1], text_tup[-1][1] - 1)  # 最后一句不以sign结尾时
        text_ser = pd.Series(dict(text_tup))
    text_df = text_ser.apply(sen_position)
    text_df = text_df.reset_index()
    text_df.columns = ['text', 'txt_len', 'group']
    result = text_df.groupby(text_df['group']).apply(sen_join)
    result['text_cumsum_len'] = result['join_text_len'].cumsum()

    spo_list_split = [[] for _ in range(result.shape[0])]
    for spo in spo_list:
        for i in result.iterrows():
            # 根据句子长度分组情况重新计算spo_list
            if spo['t']['pos'][-1] <= i[1]['text_cumsum_len']:
                if i[1]['text_cumsum_len'] == i[1]['join_text_len']:
                    spo_list_split[i[0]].append(
                        (spo['h']['name'], spo['h']['pos'], spo['t']['name'], spo['t']['pos'], spo['relation']))
                else:
                    cut_number = i[1]['text_cumsum_len'] - i[1]['join_text_len']
                    spo_h_pos = [spo['h']['pos'][0] - cut_number, spo['h']['pos'][1] - cut_number]  # 更新pos
                    spo_t_pos = [spo['t']['pos'][0] - cut_number, spo['t']['pos'][1] - cut_number]  # 更新pos
                    spo_list_split[i[0]].append(
                        (spo['h']['name'], spo_h_pos, spo['t']['name'], spo_t_pos, spo['relation']))
                break

    result['spo_list'] = spo_list_split
    return result


test_ss = {'ID': 'AT0058',
           'text': '诊断过程思路分析：造成此故障的原因一般有：1、ABS控制电脑或ABS泵出现故障；2、ABS传感器信号没有或信号强度不够；3、ABS线路故障。故障诊断过程：1、明锐是ABS灯亮，很多都是传感器信号不好造成的，因为ABS传感器很容易粘附铁屑等异物导致信号差，ABS故障灯报警。用5051进入车辆自诊断系统，然后进入制动系统查询，查询结果显示为：ABS泵-64电路电气故障。说明ABS传感器没有问题，排除此方面原因。2、由此来看，故障应该是线路问题。但是线路问题比较复杂，先拔下ABS泵插头看是否有进水短路现象，经检查发现无明显进水痕迹。然后用万用表ABS泵电源电压，发现电压为零，于是检查电源保险丝。ABS泵电源保险在发动机仓内保险盒上，共有SB4、SB16、SB28三个相关保险，检查发现SB4保险烧坏，其它正常。在驾驶室内为ABS提供电源的保险有两个SC1、SC2两个，经检查这两个保险正常。这说明SB4输出线路有过载或短路现象，用万用表测量SB4保险到ABS泵的线路，未发现短路情况。于是换上新保险，故障消失，用5051消除故障码，启动发动机检查故障消失。由此判断，可能是SB4相关线路出现过载情况，导致保险烧坏3、将车辆装好，在院内进行路试，一切正常。于是做好收尾工作，正打算完工交车时，忽然想起客户说车辆是在走过一段颠簸路时出现这种故障的，为了确保万无一失，应该到颠簸路面试车确认。想到这里，我将车辆开到外面一段路面破坏教严重的路试车，刚开始一切正常，大约走了500m左右，忽然间ABS和ASR指示灯亮起，故障又出现。4、将车辆开回维修车间检查，发现又是SB4保险烧坏。对SB4到ABS泵之间的线路重新检查测量，并未发现短路情况，顺着保险盒到ABS泵间线束检查外观，也未发现明显破损地方。5、既然烧保险，又未发现短路情况，考虑可能是地线接触不良，产生冲击电流，导致保险损坏。于是检查所有的接地线，ABS制动系统在右侧A柱下方有一接地点（因为车装有地板革，怀疑装地板革时工作人员动了地线），拆下饰板检查发现其接地良好。其它的接地线也都很好，未发现异常。由于ABS泵能够正常工作，只是在颠簸路面行驶时才出现此故障，ABS泵应该没有问题，考虑应该还是线路问题，但是线路又查不出故障，到此陷入困境。6、静下心来，重新梳理思路，从故障现象分析，只能是线路故障，不是元器件损坏引起的。这次将排查重点放在SB4保险到ABS泵之间的线路上，于是将电瓶和发动机仓内保险盒拆下，将线束包扎拆开仔细检查。通过仔细检查发现SB4输出线在保险盒座下面有一处很小的磨损，外面的绝缘皮已经磨透，露出里面的铜线，并且磨损处有电火花的烧痕，到此真相大白，保险损坏就是此处线路接地造成的。原来是线束缺少一个固定卡子，车辆行驶时线束窜动与保险盒下的支架摩擦，时间长了把电线绝缘皮磨破接地，从而引发了这起故障。由于磨损很小，检查时不仔细不容易发现，并且只有当车辆在颠簸路面行驶，振动比较大时候才出现接地烧保险情况。7、将线束进行包扎处理，并做好固定。之后试车，一切正常，故障彻底排除。7、将线束进行包扎处理，并做好固定。之后试车，一切正常，故障彻底排除',
           'spo_list': [{'h': {'name': '保险盒下的支架', 'pos': [1153, 1160]},
                         't': {'name': '摩擦', 'pos': [1160, 1162]},
                         'relation': '部件故障'},
                        {'h': {'name': '保险', 'pos': [1113, 1115]},
                         't': {'name': '损坏', 'pos': [1115, 1117]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS故障灯', 'pos': [127, 133]},
                         't': {'name': '报警', 'pos': [133, 135]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS控制电脑', 'pos': [23, 30]},
                         't': {'name': '故障', 'pos': [37, 39]},
                         'relation': '部件故障'},
                        {'h': {'name': 'SB4相关线路', 'pos': [487, 494]},
                         't': {'name': '过载', 'pos': [496, 498]},
                         'relation': '部件故障'},
                        {'h': {'name': '线束', 'pos': [1148, 1150]},
                         't': {'name': '摩擦', 'pos': [1160, 1162]},
                         'relation': '部件故障'},
                        {'h': {'name': 'SB4输出线', 'pos': [1052, 1058]},
                         't': {'name': '磨损', 'pos': [1071, 1073]},
                         'relation': '部件故障'},
                        {'h': {'name': '保险', 'pos': [503, 505]},
                         't': {'name': '烧坏', 'pos': [505, 507]},
                         'relation': '部件故障'},
                        {'h': {'name': 'SB4保险', 'pos': [680, 685]},
                         't': {'name': '烧坏', 'pos': [685, 687]},
                         'relation': '部件故障'},
                        {'h': {'name': '电线', 'pos': [1168, 1170]},
                         't': {'name': '绝缘皮磨破接地', 'pos': [1170, 1177]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS传感器', 'pos': [42, 48]},
                         't': {'name': '信号强度不够', 'pos': [53, 59]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ASR', 'pos': [647, 650]},
                         't': {'name': '指示灯亮起', 'pos': [650, 655]},
                         'relation': '部件故障'},
                        {'h': {'name': '地线', 'pos': [769, 771]},
                         't': {'name': '接触不良', 'pos': [771, 775]},
                         'relation': '部件故障'},
                        {'h': {'name': '线束', 'pos': [1132, 1134]},
                         't': {'name': '缺少一个固定卡子', 'pos': [1134, 1142]},
                         'relation': '部件故障'},
                        {'h': {'name': '保险', 'pos': [785, 787]},
                         't': {'name': '损坏', 'pos': [787, 789]},
                         'relation': '部件故障'},
                        {'h': {'name': 'SB4保险', 'pos': [343, 348]},
                         't': {'name': '烧坏', 'pos': [348, 350]},
                         'relation': '部件故障'},
                        {'h': {'name': 'SB4输出线路', 'pos': [399, 406]},
                         't': {'name': '过载', 'pos': [407, 409]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS泵', 'pos': [31, 35]},
                         't': {'name': '故障', 'pos': [37, 39]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS泵', 'pos': [170, 174]},
                         't': {'name': '64电路电气故障', 'pos': [175, 183]},
                         'relation': '部件故障'},
                        {'h': {'name': 'SB4输出线路', 'pos': [399, 406]},
                         't': {'name': '短路', 'pos': [410, 412]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS', 'pos': [62, 65]},
                         't': {'name': '线路故障', 'pos': [65, 69]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS传感器', 'pos': [42, 48]},
                         't': {'name': '信号没有', 'pos': [48, 52]},
                         'relation': '部件故障'},
                        {'h': {'name': 'ABS', 'pos': [643, 646]},
                         't': {'name': '指示灯亮起', 'pos': [650, 655]},
                         'relation': '部件故障'}]}
test_ss_text = test_ss['text'].replace('；', '。')  # 分号替换为句号(切分句子优先级相等)
result_ss = split_sentence(test_ss_text, test_ss['spo_list'])
result_ss
```
