```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # 读取图片(RGB模式)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans = transforms.TenCrop(100,
                           vertical_flip=True) # 是否进行垂直翻转;若vertical_flip=False,则进行水平翻转
# 图片分别为左上,右上,左下,右下,中心,左上翻转,右上翻转,左下翻转,右下翻转,中心翻转
ten_result = trans(img_resize)
```

```python
import matplotlib.pyplot as plt

for i in ten_result:
    plt.imshow(i)
    plt.show()
```
