&emsp;&emsp;条件随机场的概率计算问题是给定条件随机场$P(Y|X)$,输入序列$\mathbf{x}$和输出序列$\mathbf{y}$,
计算条件概率$P(Y_i=y_i|\mathbf{x}),P(Y_{i-1}=y_{i-1},Y_i=y_i|\mathbf{x})$以及相应的数学期望的问题.

#### 前向-后向算法

&emsp;&emsp;对每个指标$i=0,1,\cdots,n+1$,定义前向向量$\alpha_i(\mathbf{x})$:

$$
\alpha_{0}(\mathbf{y} \mid \mathbf{x})=\left\{\begin{array}{ll}
1, & y=\text { start } \\
0, & \text { 否则 }
\end{array}\right.
$$

递推公式为

$$ \alpha_{i}^{\mathrm{T}}\left(y_{i} \mid \mathbf{x}\right)=\alpha_{i-1}^{\mathrm{T}}\left(y_{i-1} \mid \mathbf{x}\right)\left[M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x}\right)\right], \quad i=1,2, \cdots, n+1 $$

又可以表示为

$$ \alpha_{i}^{\mathrm{T}}(\mathbf{x})=\alpha_{i-1}^{\mathrm{T}}(\mathbf{x}) M_{i}(\mathbf{x}) $$

$\alpha_{i}(y_i|\mathbf{x})$表示在位置$i$的标记是$y_i$并且从1到$i$的前部分标记序列的非规范化概率,$y_i$可能的取值有$m$个,所以$\alpha_i(\mathbf{x})$是$m$维列向量.

&emsp;&emsp;同样,对每个指标$i=0,1,\cdots,n+1$,定义后向变量$\beta_i(\mathbf{x})$:

$$
\beta_{n+1}\left(y_{n+1} \mid \mathbf{x} \right)=\left\{\begin{array}{ll}
1, & y_{n+1}=\operatorname{stop} \\
0, & \text { 否则 }
\end{array}\right.
$$

$$ \beta_{i}\left(y_{i} \mid \mathbf{x} \right)=\left[M_{i+1}\left(y_{i}, y_{i+1} \mid \mathbf{x} \right)\right] \beta_{i+1}\left(y_{i+1} \mid \mathbf{x} \right) $$

又可以表示为

$$ \beta_{i}(\mathbf{x})=M_{i+1}(\mathbf{x}) \beta_{i+1}(\mathbf{x}) $$

$\beta_{i}(y_i|\mathbf{x})$表示在位置$i$的标记为$y_i$并且从$i+1$到$n$的后部分标记序列的非规范化概率.

### 概率计算

&emsp;&emsp;根据前向-后向向量的定义,很容易计算标记序列在位置$i$是标记$y_i$的条件概率和在位置$i-1$与$i$是标记$y_{i-1}$和$y_i$的条件概率:

$$ P\left(Y_{i}=y_{i} \mid \mathbf{x} \right)=\frac{\alpha_{i}^{\mathrm{T}}\left(y_{i} \mid \mathbf{x}\right) \beta_{i}\left(y_{i} \mid \mathbf{x}\right)}{Z(\mathbf{x})} $$

$$ P\left(Y_{i-1}=y_{i-1}, Y_{i}=y_{i} \mid \mathbf{x}\right)=\frac{\alpha_{i-1}^{\mathrm{T}}\left(y_{i-1} \mid \mathbf{x}\right) M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x}\right) \beta_{i}\left(y_{i} \mid \mathbf{x}\right)}{Z(\mathbf{x})} $$

其中,

$$ Z(x)=\alpha_{n}^{\mathrm{T}}(\mathbf{x}) \mathbf{1}=\mathbf{1}^T \beta_{1}(\mathbf{x}) $$

$\mathbf{1}$是元素均为1的$m$维列向量.

上述证明如下:

$$
\begin{aligned}
Z(\mathbf{x}) &=\left[ M_{1}(\mathbf{x}) M_{2}(\mathbf{x}) \cdots M_{n+1}(\mathbf{x})\right]_{(\text {start, stop})} \\
&=\left[M_{1}(\mathbf{x}) M_{2}(\mathbf{x}) \cdots M_{n}(\mathbf{x}) \beta_{n+1}(\mathbf{x})\right]_{\text {start }} \\
&=\left[ M_1(\mathbf{x}) \beta_{2}(\mathbf{x})\right]_{\text {start }} \\
&=\mathbf{1}^T \beta_{1}(\mathbf{x})
\end{aligned}
$$

其中,$\beta_{n+1}(\mathbf{x})= [1, 1, \cdots, 1]_{1 \times m}^T $

$$
\begin{aligned}
Z(\mathbf{x}) &=\left[M_{1}(\mathbf{x}) M_{2}(\mathbf{x}) \cdots M_{n+1}(\mathbf{x})\right]_{(\text {start, stop })} \\
&=\alpha_{0}(\mathbf{x})^{T} M_{1}(\mathbf{x}) M_{2}(\mathbf{x}) \cdots M_{n}(\mathbf{x}) \cdot \mathbf{1} \\
&=\alpha_{n}(\mathbf{x})^{T} \cdot \mathbf{1}
\end{aligned}
$$

其中,$\alpha_0(\mathbf{x})=[1,1,\cdots,1]_{1 \times m}^T$

$$
\begin{aligned}
P\left(Y_{i}=y_{i} \mid \mathbf{x} \right) &= \frac{ \left[ M_1(\mathbf{x})M_2(\mathbf{x})\cdots M_{i-1}(\mathbf{x}) M_{i}(Y_{i-1},Y_{i}=y_i|\mathbf{x}) M_{i+1}(Y_{i}=y_i,Y_{i+1}|\mathbf{x}) \cdots M_{n+1}(\mathbf{x}) \right]_{ \mathrm{start, stop} }}{Z(\mathbf{x})}  \\
&=\frac{\alpha_{i}^{\mathrm{T}}\left(y_{i} \mid \mathbf{x}\right) \beta_{i}\left(y_{i} \mid \mathbf{x}\right)}{Z(\mathbf{x})}
\end{aligned}
$$

其中,$M_{i}(Y_{i-1},Y_{i}=y_i|\mathbf{x})$表示 $M_i(\mathbf{x})$的第$y_i$列(共$m$列).$ M_{i+1}(Y_{i}=y_i,Y_{i+1}|\mathbf{x})$表示$M_{i=1}(\mathbf{x})$的第$y_i$行(共$m$行),参见条件随机场的矩阵形式例子.

### 期望值的计算

&emsp;&emsp;利用前向-后向向量,可以计算特征函数关于联合分布$P(X,Y)$和条件分布$P(X|Y)$的数学期望.

&emsp;&emsp;特征函数$f_k$关于条件分布$P(Y|X)$的数学期望是

$$
\begin{aligned}
E_{P(Y \mid X)}\left[f_{k}\right] &=\sum_{\mathbf{y}} P(\mathbf{y} \mid \mathbf{y}) f_{k}(\mathbf{y}, \mathbf{x}) \\
&=\sum_{i=1}^{n+1} \sum_{y_{i-1} y_{i}} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right) \frac{\alpha_{i-1}^{\mathrm{T}}\left(y_{i-1} \mid \mathbf{x} \right) M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right) \beta_{i}\left(y_{i} \mid \mathbf{x}\right)}{Z(\mathbf{x})} \\
 & \qquad \qquad \qquad \qquad \qquad \qquad \qquad k=1,2,\cdots,K
\end{aligned}
\tag{1}
$$

&emsp;&emsp;假设经验分布为$\tilde{P}(X)$,特征函数$f_k$关于联合分布$P(X,Y)$的数学期望是

$$
\begin{aligned}
E_{P(X, Y)}\left[f_{k}\right] &=\sum_{\mathbf{x}, \mathbf{y}} P(\mathbf{x}, \mathbf{y}) \sum_{i=1}^{n+1} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right) \\
&=\sum_{\mathbf{x}} \tilde{P}(\mathbf{x}) \sum_{\mathbf{y}} P(\mathbf{y} \mid \mathbf{x}) \sum_{i=1}^{n+1} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right) \\
&=\sum_{\mathbf{x}} \tilde{P}(\mathbf{x}) \sum_{i=1}^{n+1} \sum_{y_{i-1} y_{i}} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right) \frac{\alpha_{i-1}^{\mathrm{T}}\left(y_{i-1} \mid \mathbf{x} \right) M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x}\right) \beta_{i}\left(y_{i} \mid \mathbf{x} \right)}{Z(\mathbf{x})}
\\
 & \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad k=1,2,\cdots,K
\end{aligned}
\tag{2}
$$

&emsp;&emsp;式(1)和式(2)是特征函数数学期望的一般计算公式.对于转移特征$t_k(y_{i-1},y_i,\mathbf{x},i),k=1,2,\cdots,K_1$可以将式中的$f_k$换成$t_k$;
对于状态特征,可以将$f_k$换成$s_i$,表示为$s_l(y_i,\mathbf{x},i),k=K_1+l,l=1,2,\cdots,K_2$.
