```python
import numpy as np
import matplotlib.pyplot as plt

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color='r', fontsize=40)
ax1.plot(t, data1, color='g')
ax1.tick_params(axis='y', labelcolor='pink')
ax1.grid()

ax2 = ax1.twinx()  
# Create a new Axes instance with an invisible x-axis and an independent y-axis positioned opposite to the original one (i.e. at right). 
ax2.set_ylabel('sin', color='#fac205', fontsize=40)
ax2.plot(t, data2, color='#040273')
ax2.tick_params(axis='y', labelcolor='#ed0dd9',
                color='red', width=5, length=14,
                direction='inout')
ax2.grid()
plt.show()

```
