```python
import seaborn as sns

sns.set(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Draw a single horizontal boxplot:
sns.boxplot(x="total_bill",
            data=tips)
```

```python
# Draw a vertical boxplot grouped by a categorical variable:
sns.boxplot(x="day",  # 分组列为"day"(自动识别)
            y="total_bill",
            data=tips,
            palette="Set1")  # 通过调色盘为不同分组绘制不同颜色
```

```python
# 单一颜色绘制所有列
sns.barplot(x="day",
            y="total_bill",
            data=tips,
            color='red')
```

```python
# Draw a boxplot with nested grouping by two categorical variables:
sns.boxplot(x="day", y="total_bill",
            hue="smoker",  # 指定第二个分组列为"smoker"
            data=tips, palette="Set1")
```

```python
# Control box order by passing an explicit order:
sns.boxplot(x="time", y="tip", data=tips,
            order=["Dinner", "Lunch"])  # 自定义分组顺序
```

```python
# Draw a boxplot for each numeric variable in a DataFrame:
iris = sns.load_dataset("iris")
sns.boxplot(data=iris, orient="v")  # 绘制DataFrame的所有列(未指定x,y)
```

```python
sns.boxplot(data=iris,
            orient="h")  # 水平方向
```
