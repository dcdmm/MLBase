```python
import tensorflow as tf
```

```python
x = tf.linspace(0.0, 0.1, 100)
y = tf.reshape(tf.linspace(1, 200, 200), (100, 2))
z = tf.range(1000, 1100)
```

```python
# Creates a Dataset with a single element, comprising the given tensors.
for i in tf.data.Dataset.from_tensors(x).as_numpy_iterator():
    print(i)  # 输出x
```

```python
dataset = tf.data.Dataset.from_tensors((x, y, z))
print(dataset)
print(len(dataset))
```

```python
dataset._tensors
```

```python
for i in dataset.as_numpy_iterator():
    print(i)  # 输出(x, y, z)
```

```python
dataset = tf.data.Dataset.from_tensors({"x": x, "y": y, "z": z})
dataset
```

```python
dataset._tensors
```

```python
for i in dataset.as_numpy_iterator():
    print(i)  # 输出{"x": x, "y": y, "z": z}
```
