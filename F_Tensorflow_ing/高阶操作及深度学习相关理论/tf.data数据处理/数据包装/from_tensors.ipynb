```python
import tensorflow as tf
```

```python
x = tf.linspace(0.0, 0.1, 100)
y = tf.reshape(tf.linspace(1, 400, 400), (100, 2, 2))
z = tf.range(1000, 1100)
```

```python
# Creates a Dataset with a single element, comprising the given tensors.
for i in tf.data.Dataset.from_tensors(x):
    print(i)  # 输出x
```

```python
dataset = tf.data.Dataset.from_tensors((x, y, z))  # 使用元组打包x,y,z

print(len(dataset))
print(dataset)
print(len(dataset._tensors))  # 数据集[x, y, z]
print(dataset._tensors)
```

```python
for i in dataset:
    print(i)  # 输出数据集(x, y, z)
```

```python
dataset = tf.data.Dataset.from_tensors({"x": x, "y": y, "z": z})  # 使用字典打包x,y
dataset
```

```python
dataset._tensors  # 数据集[x, y]
```

```python
for i in dataset:
    print(i)  # 输出{"x": x, "y": y, "z": z}
```
