```python
import pandas as pd
import numpy as np

np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 2] = np.nan
```

```python
df.style
```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color # 颜色设置

s = df.style.applymap(color_negative_red) # 按元素运行
s
```

```python
df.style.applymap(color_negative_red,
                  subset=pd.IndexSlice[2:5, ['B','D']]) # 切片
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max] # 背景色设置

df.style.apply(highlight_max, axis=1) # 一次传递一列
```

```python
df.style.apply(highlight_max, subset=pd.IndexSlice[:, 'B':'D']) # 切片
```

```python
df.style.\
    applymap(color_negative_red).\
    apply(highlight_max) # 同时进行多个设置
```

```python
def highlight_all_max(s, all_max, color='yellow'):
    '''
    highlight the maximum in a Series or DataFrame
    '''
    is_max = s == all_max
    attr = 'background-color: {}'.format(color)
    return attr if is_max  else ''

all_max = df.max().max()
v = df.style.applymap(highlight_all_max, color='blue', all_max=all_max)
v
```
