```python
import  numpy as np
import  matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

fig = plt.figure(figsize=(10, 7))
ax1 = plt.subplot(111)
x = np.random.randn(1000)
y = np.random.randn(1000)
cm_bright = ListedColormap(['red', 'yellow', 'green', 'pink', 'blue', 'black', 'orange']) 
s = ax1.scatter(x, y, marker='x', alpha=0.9,
            s=np.abs(np.random.randn(1000)*100),
            c=x**2 + y, cmap=cm_bright, # A sequence of n numbers to be mapped to colors using cmap and norm.
                lw=10, edgecolor='red') 
'''
x, y : array_like, shape (n, )
    The data positions.

s : scalar or array_like, shape (n, ), optional
    The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.

c : color, sequence, or sequence of color, optional
    The marker color. Possible values:

    A single color format string.
    A sequence of color specifications of length n.
    A sequence of n numbers to be mapped to colors using cmap and norm.
    A 2-D array in which the rows are RGB or RGBA.
    
marker : MarkerStyle, optional
    The marker style. marker can be either an instance of the class or the text 
    shorthand for a particular marker. Defaults to None, in which case it takes 
    the value of rcParams["scatter.marker"] = 'o'. See markers for more information 
    about marker styles.

cmap : Colormap, optional, default: None
    A Colormap instance or registered colormap name. cmap is only used if c is an 
    array of floats. If None, defaults to rc image.cmap.

lw : scalar or array_like, optional, default: None
    The linewidth of the marker edges. Note: The default edgecolors is 'face'. You may want to change this as well.

edgecolors : color or sequence of color, optional, default: 'face'
    The edge color of the marker. Possible values:
        'face': The edge color will always be the same as the face color.
        'none': No patch boundary will be drawn.
        A matplotib color.
    For non-filled markers, the edgecolors kwarg is ignored and forced to 'face' internally.
'''
ax1.grid()
fig.colorbar(s) # Add a colorbar to a plot.
```
