```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sinc(x)

fig, ax = plt.subplots()

ax.plot(x, y)
```

```python
# The default margin around the data limits is 5%:
ax.margins()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

# Specific margin values for the x-axis and y-axis
ax.margins(x=1., y=2.)  # 指定x轴边距长/图形x轴区域长=1;y轴边距/图形y轴区域长=2
```
