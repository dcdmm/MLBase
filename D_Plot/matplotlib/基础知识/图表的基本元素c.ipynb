1. get_xaxis().set_major_locator,get_xaxis().set_minor_locator
2. get_xaxis().set_major_formatter,get_xaxis().set_minor_formatter

```python
import matplotlib.pyplot as plt
import  numpy as np
from  matplotlib.ticker import  MultipleLocator, FormatStrFormatter

t = np.arange(0.0, 100, 1)
s = np.sin(0.1*np.pi*t)*np.exp(-t*0.01)
fig = plt.figure(figsize=(10, 7))
ax = plt.subplot(111)
ax.plot(t, s , marker='*', linestyle='--')

xmajorLocator = MultipleLocator(40) 
ax.get_xaxis().set_major_locator(xmajorLocator) # 将x主刻度间隔设置为40的倍数
xmajorFormatter = FormatStrFormatter('%.1f') 
ax.get_xaxis().set_major_formatter(xmajorFormatter) # 将x主刻度刻度值设置保留1位小数

# xminorLocator = MultipleLocator(5) 
# ax.get_xaxis().set_minor_locator(xminorLocator) #将x次刻度间隔设置为5的倍数
# xminorFormatter = FormatStrFormatter('%.2f') 
# ax.get_xaxis().set_minor_formatter(xminorFormatter) #将x次刻度刻度值设置保留2位小数

ax.grid(True, which='both', axis='x')

ymajorLocator = MultipleLocator(0.3) 
ax.get_yaxis().set_major_locator(ymajorLocator) # 将y主刻度间隔设置为0.3的倍数
ymajorFormatter = FormatStrFormatter('%.1f') 
ax.get_yaxis().set_major_formatter(ymajorFormatter) # 将y主刻度刻度值设置保留1位小数

yminorLocator = MultipleLocator(0.1) 
ax.get_yaxis().set_minor_locator(yminorLocator) # 将y次刻度标签设置为0.1的倍数
yminorFormatter = FormatStrFormatter('%.2f') 
ax.get_yaxis().set_minor_formatter(yminorFormatter) # 将y次刻度刻度值设置保留2位小数

ax.grid(True, which='both', axis='y')



# ax.get_xaxis().set_major_locator(plt.NullLocator()) 
# 主刻度(刻度点,刻度值,网格线)均不可见,等价于:ax.get_xaxis().set_ticks([])或ax.set_xticks([])

# ax.get_xaxis().set_major_formatter(plt.NullFormatter()) 
# 只主刻度刻度值不可见,等价于:#ax.get_xaxis().set_ticklabels([])

# ax.get_xaxis().set_minor_locator(plt.NullLocator()) 
# 次刻度(刻度点,刻度值,网格线)均不可见,等价于:ax.get_xaxis().set_ticks([], monor=True)或ax.set_xticks([], monor=True)
# ax.get_xaxis().set_minor_formatter(plt.NullFormatter()) 
# 只次刻度刻度值不可见,等价于:#ax.get_xaxis().set_ticklabels([], monor=True)

# ax.get_xaxis().set_visible(False) # 整个x轴不可见
# plt.setp(frame.get_yaxis(), visible=False) #与上等价
```
