```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # 读取图片(RGB模式)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
trans = transforms.RandomGrayscale(p=0.5) #  probability that image should be converted to grayscale.
#  If input image is 1 channel: grayscale version is 1 channel - If input image is 3 channel: grayscale version is 3 channel with r == g == b
gray_img = trans(img_resize)
gray_img
```

```python
trans_tensor = transforms.ToTensor()
img_tensor = trans_tensor(gray_img)
img_tensor[0, :, :] - img_tensor[1, :, :] # 每个通道数值相等



```
