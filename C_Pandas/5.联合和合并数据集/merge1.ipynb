## merge(数据库风格的DataFrame连接)

```python
import pandas as pd
import numpy as np

left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lva1': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'lva1': [4, 5, 6, 7]})
```

```python
print(left, '\n\n', right)
```

```python
# 连接键为多个时,传入一个用作连接键的列名的列表
pd.merge(left, right=right, on=['key1', 'key2'], how='outer')
```

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})

right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
print(left1, '\n\n', right1)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)  # 使用right DataFrame的行索引作为它的连接键
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
lefth = pd.DataFrame({'key1': ['ohio', 'ohio', 'ohio', 'nevada', 'nevada'],
                      'key2': [2000, 2001, 2002, 2001, 2002],
                      'data': np.arange(5.0)})

righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['nevada', 'nevada', 'ohio', 'ohio', 'ohio',
                              'ohio'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['event1', 'event2'])
print(lefth, '\n\n', righth)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')  # 多层索引数据合并
```

```python
left1 = pd.DataFrame([[1, 2], [3, 4], [5, 6]], index=['a', 'b', 'c'],
                     columns=['ohio', 'nevada'])

right1 = pd.DataFrame([[7, 8], [9, 10], [11, 12], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['missouri', 'alabama'])

print(left1, '\n\n', right1)
```

```python
# 使用left DataFrame的行索引作为它的连接键
# 使用right DataFrame的行索引作为它的连接键
pd.merge(left1, right1, how='outer', left_index=True, right_index=True)
```
