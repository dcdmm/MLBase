```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('../test_picture/1yuan_1.jpg').convert('RGB') # (RGB)
trans_resize = transforms.Resize((224, 224))
img_resize = trans_resize(img)
img_resize
```

```python
# 0()
trans0 = transforms.Pad(padding=54) # 54
trans0(img_resize)
```

```python
trans0_1 = transforms.Pad(padding=(12, 59)) #  # 12,59
trans0_1(img_resize)
```

```python
# 12,24,36,48
trans0_2 = transforms.Pad(padding=(12, 24, 36, 48))
trans0_2(img_resize)
```

```python
trans1 = transforms.Pad(padding=(12, 24, 36, 48), fill=(244, 1, 1)) # (padding_mode='constant')
trans1(img_resize)
```

```python
# edge: pads with the last value at the edge of the image
trans2 = transforms.Pad(70, padding_mode='edge') # padding_mode='constant'
trans2(img_resize)
```

```python
trans2_1 = transforms.Pad(520, padding_mode='reflect')
'''
reflect: pads with reflection of image without repeating the last value on the edge
    For example, padding [1, 2, 3, 4] with 2 elements on both sides in reflect mode will result in [3, 2, 1, 2, 3, 4, 3, 2]
'''
trans2_1(img_resize)
```

```python
trans2_2 = transforms.Pad(520, padding_mode='symmetric')
'''
symmetric: pads with reflection of image repeating the last value on the edge
    For example, padding [1, 2, 3, 4] with 2 elements on both sides in symmetric mode will result in [2, 1, 1, 2, 3, 4, 4, 3]
'''
last = trans2_2(img_resize)
last
```
