```python
import  seaborn as sns
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
```

```python
v = sns.color_palette(palette=['pink', 'g', '#fac205'], n_colors=15, desat=0.9) # 自定义颜色列表
print(len(v))
print(v) # RGB或RGBA颜色列表
sns.palplot(v, size=1)
```

```python
s = sns.color_palette(palette='hsv', n_colors=25, desat=0.9) # matplotlib.plt.cm
sns.palplot(s, size=.2) # Plot the values in a color palette as a horizontal array
```

```python
sns.palplot(sns.color_palette(palette=s, n_colors=16, desat=0.9)) # seaborn.color_paletter
'''
seaborn.color_palette(palette=None, n_colors=None, desat=None)
Return a list of colors defining a color palette
    
    Parameters:	
        palette: None, string, or sequence, optional
            Name of palette or None to return current palette. 
    
        n_colors : int, optional
            Number of colors in the palette. 

        desat : float, optional
            Proportion to desaturate each color by.

    Returns:	
        palette : list of RGB tuples.
            Color palette. 
'''
```

```python
# 简单应用
fig, axes = plt.subplots(3, 5, figsize=(40, 15))
df = pd.DataFrame(np.random.rand(50, 15))
for i in df.columns:
    j = i//5 # 地板除法
    k = i%5
    axes[j][k].plot(df[i], color=v[i], linewidth=4)
```
