### 条件随机场的参数化形式

&emsp;&emsp;根据Hammersley-Clifford定理,可以给出线性链条件随机场$P(Y|X)$的因子分解式,各因子是定义在相邻两个结点(最大团)上的势函数.

&emsp;&emsp;<font color='red'>定理:(线性条件随机场的参数化形式)</font>设$P(Y|X)$为线性链条件随机场,
则在随机变量$X$取值为$\mathbf{x}$(为序列)的情况下,
随机变量$Y$取值为$\mathbf{y}$(为序列)的条件概率具有如下形式(没有线性条件链中的$y_{i+1}$,因为$i+1$时为$t_k \left( y_{i}, y_{i+1}, \mathbf{x}, i \right)$,故可以不带$y_{t+1}$:

$$ P(\mathbf{y} \mid \mathbf{x})=\frac{1}{Z(\mathbf{x})} \exp \left(\sum_{i, k} \lambda_{k} t_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)+\sum_{i, l} \mu_{l} s_{l}\left(y_{i}, \mathbf{x}, i\right)\right) \tag{1} $$

其中,

$$ Z(\mathbf{x})=\sum_{\mathbf{y}} \exp \left(\sum_{i, k} \lambda_{k} t_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)+\sum_{i, l} \mu_{l} s_{l}\left(y_{i}, \mathbf{x}, i\right)\right) \tag{2} $$

其中,$t_k$和$s_l$是特征函数,$\lambda_k$和$\mu_l$是对应的权值.$Z(\mathbf{x})$是规范化因子,求和是在所有可能的输出序列上进行的(注意:$\mathbf{x}表示的为全局信息$,可以根据需求为任意位置的$x_i$或任意多个$x_i, i=1,2,\cdots,n$).

&emsp;&emsp;式(1)和式(2)是线性链条件随机场模型的基本形式,表示给定输入序列$\mathbf{x}$,对输出序列$y$预测的条件概率.式(1)和式(2)中,$t_k$是定义在边上的特征函数,称为转移特征,依赖于当前和前一个位置;$s_l$是定义在结点上的特征函数,称为状态特征,依赖于当前位置.$t_k$和$s_l$都依赖于位置,是局部特征函数.通常,特征函数$t_k$和$s_l$取值为1或0;当满足特征条件时取值为1,否则为0.条件随机场完全由特征函数$t_k,s_l$和对应的权值$\lambda_k,\mu_l$确定.

&emsp;&emsp;线性链条件随机场也是对数线性模型(log linear model)

**特征函数含义示例:** 设有一标注问题:输入观测序列为$X=(X_1,X_2,X_3, X_4)=( \mathrm{Dog \quad  caught  \quad the \quad cat})$ ,输出标记序列为$Y=(Y_1,Y_2,Y_3, Y_4)$,取值与$\mathcal{Y}= (1, 2, 3) =(\mathrm{verb, article, noun}) $.

假设特征$t_k,s_l$和对应的权值$\lambda_k, \mu_l$如下

\begin{equation}
t_{1}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)=\left\{\begin{array}{ll}
1, & y_{i-1}=3, y_{i}=2, \mathbf{x}, i,(i=2) \\
0, & \text { 其他 }
\end{array}\right.
\end{equation}

下面只注明特征为1的条件,取值为0的条件省略,即

特征(转移)函数$t_1$的含义:对于该句子($\mathbf{x}$为全局信息),若第一个单词为名词,第二个单词为动词,则特征函数$t_k$设置为1,并设置权值为1.

$$ t_{2}=t_{2}\left(y_{i-1}=1, y_{i}=2, \mathbf{x}, i\right), \quad i=3, \quad \lambda_{1}=0.5 $$

\begin{array}{ll}
s_{1}=s_{1}\left(y_{2}=1, \mathbf{x}_2, 2\right), & \mu_{1}=1 \\
s_{2}=s_{2}\left(y_{i}=3, \mathbf{x}_2, 3\right), & \mu_{2}=0.5

特征(状态)函数$s_1$的含义:对于第2个单词,且这个单词词性为动词,则特征函数$s_l$设置为1,并设置权值为1.

特征(状态)函数$s_2$的含义:对于第3个单词,若第二个的单词为caught,这个单词为冠词.则特征函数$s_l$设置为1,并设置权值为0.5.

### 条件随机场的简化形式
&emsp;&emsp;条件随机场还可以由简化形式表示.注意到条件随机场式(1)中同一个特征函数在各个位置求和,将局部特征函数转化为一个全局特征函数,这样就可以将条件随机场写成权值向量和特征向量的内积形式,即条件随机场的简化形式.

&emsp;&emsp;为了简便起见,首先将转移特征和状态特征及其权值用统一的符号表示.设有$K_1$个转移特,$K_2$个状态特征,$K=K_1+K_2$,记

\begin{equation}
f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)=\left\{\begin{array}{ll}
t_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right), & k=1,2, \cdots, K_{1} \\
s_{l}\left(y_{i}, \mathbf{x}, i\right), & k=K_{1}+l ; \quad l=1,2, \cdots, K_{2}
\end{array}\right.
\end{equation}

然后,对转移与状态特征在各个位置$i$求和,记作

$$ f_{k}(\mathbf{y}, \mathbf{x})=\sum_{i=1}^{n} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right), \quad k=1,2, \cdots, K $$

用$w_k$表示特征$f_k(\mathbf{y}, \mathbf{x})$的权值,即

\begin{equation}
w_{k}=\left\{\begin{array}{ll}
\lambda_{k}, & k=1,2, \cdots, K_{1} \\
\mu_{l}, & k=K_{1}+l ; l=1,2, \cdots, K_{2}
\end{array}\right.
\end{equation}

即:

\begin{align}
\sum_{i, k} \lambda_{k} t_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)+\sum_{i, l} \mu_{l} s_{l}\left(y_{i}, \mathbf{x}, i\right) &= \sum_{k=1}^{K_1}\lambda_k \left[\sum_it_k(y_{i-1},y_i,\mathbf{x},i) \right] + \sum_{l=1}^{K_2}\mu_l \left[\sum_is_l(y_i,\mathbf{x},i)\right]\\
&=\left[ \lambda_1,..,\lambda_{K_1},\mu_1,..,\mu_{K_2} \right] \left[ \sum_it_1(y_{i-1},y_i,\mathbf{x},i),...,\sum_it_{K_1}(y_{i-1},y_i,\mathbf{x},i),\sum_is_1(y_i,\mathbf{x},i),...,\sum_is_{K_2}(y_i,\mathbf{x},i) \right]^T\\
&=[w_1,..,w_K][f_1(\mathbf{y},\mathbf{x}),...,f_K(\mathbf{y},\mathbf{x})]^T\\
&=\mathbf{w}^TF(y,\mathbf{x})
\end{align}

其中,$\mathbf{w}$表示权值向量,即

$$ \mathbf{w}=\left(w_{1}, w_{2}, \cdots, w_{K}\right)^{\mathrm{T}} $$

$F(\mathbf{y},\mathbf{x})$表示全局特征向量,即

$$ F(\mathbf{y}, \mathbf{x})=\left(f_{1}(\mathbf{y}, \mathbf{x}), f_{2}(\mathbf{y}, \mathbf{x}), \cdots, f_{K}(\mathbf{y}, \mathbf{x})\right)^{\mathrm{T}} $$

于是,条件随机场式(1)式(2)可表示为

$$ P_{\mathbf{w}}(\mathbf{y} \mid \mathbf{x})=\frac{\exp (\mathbf{w} \cdot F(\mathbf{y}, \mathbf{x}))}{Z_{\mathbf{w}}(\mathbf{x})} $$

其中,

$$ Z_{\mathbf{w}}(\mathbf{x})=\sum_{\mathbf{y}} \exp (\mathbf{w} \cdot F(\mathbf{y}, \mathbf{x})) $$

### 条件随机场的矩阵形式

&emsp;&emsp;条件随机场还可以由矩阵表示.对每个标记序列引进特殊的起点和终点状态标记$y_0=\mathrm{start}$和$y_{n+1}=\mathrm{stop}$,这时标记序列的概率$P_{\mathbf{w}}(\mathbf{y}|\mathbf{x})$可以通过矩阵形式表示并有效计算.

&emsp;&emsp;对观测序列$\mathbf{x} $的每一个位置$i=1,2,\cdots,n+1$,由于$y_{i-1}$和$y_i$在$m$个标记中取值,可以定义一个$m$阶矩阵随机变量

$$ M_{i}(\mathbf{x})=\left[M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right)\right]_{m \times m} $$

矩阵随机变量元素为

\begin{array}{l}
M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right)=\exp \left(W_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right)\right) \\
W_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right)=\sum_{k=1}^{K} w_{k} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)
\end{array}

其中,$y_{i-1}$和$y_i$是标记随机变量$Y_{i-1}$和$Y_i$的取值.

&emsp;&emsp;这样,给定观测序列$\mathbf{x}$,对应的标记序列$\mathbf{y}$的非规范概率可以通过该序列$n+1$个矩阵的<font color='red'>适当元素</font>乘积$\prod_{i=1}^{n+1} M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right)$表示.于是,条件概率$P_{\mathbf{w}}(\mathbf{y}|\mathbf{x})$是

$$ P_{\mathbf{w}}(\mathbf{y}|\mathbf{x})=\frac{1}{Z_{\mathbf{w}}(\mathbf{x})} \prod_{i=1}^{n+1} M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x} \right) $$

其中,$Z_{\mathbf{w}}$为规范化因子,是$n+1$个矩阵的乘积(start,stop)元素,即(矩阵乘法运行法则运行易知)

$$ Z_{\mathbf{w}}(\mathbf{x})=\left[M_{1}(\mathbf{x}) M_{2}(\mathbf{x}) \cdots M_{n+1}(\mathbf{x})\right]_{\text {start,stop }} \tag{1} $$

注意,$y_0=\mathrm{start}$和$y_{n+1}=\mathrm{stop}$表示开始状态与结束状态,规范化因子$Z_{\mathbf{w}}$是以start为起点stop为终点通过状态的所有路径$y_1,y_2,\cdots,y_n$的非规范化概率$\prod_{i=1}^{n+1} M_{i}\left(y_{i-1}, y_{i} \mid \mathbf{x}\right)$.这样做的目的是提高$Z_{\mathbf{w}}$的计算效率.

**矩阵形式举例:**给定一个由下图所示的线性链条件随机场,观测序列$\mathbf{x}$,状态序列$\mathbf{y},i=1,2,3,n=3$,标记$y_i \in \{ 1,2 \}$,假设$y_0=\mathrm{start}=1,y_4=\mathrm{stop}=1$各个位置的随机矩阵$M_{1}(\mathbf{x}), M_{2}(\mathbf{x}), M_{3}(\mathbf{x}), M_{4}(\mathbf{x})$分别是

\begin{equation}
M_{1}(\mathbf{x})=
\left[\begin{array}{cc}
a_{01} & a_{02} \\
0 & 0
\end{array}\right], \quad M_{2}(\mathbf{x})=
\left[\begin{array}{ll}
b_{11} & b_{12} \\
b_{21} & b_{22}
\end{array}\right] \\
M_{3}(\mathbf{x})=
\left[\begin{array}{cc}
c_{11} & c_{12} \\
c_{21} & c_{22}
\end{array}\right], \quad M_{4}(\mathbf{x})=
\left[\begin{array}{cc}
1 & 0 \\
1 & 0
\end{array}\right]
\end{equation}

试求状态序列$\mathbf{y}$以start为起点stop为终点的所有路径的非规范化概率及规范化因子.

<img src="../../../Other/img/条件随机场的矩阵形式.png">


**解**&emsp;4个随机矩阵元素的意义为:

\begin{equation}
M_{1}(\mathbf{x})=\left[\begin{array}{cc}
(y_0=1<-->y_1=1)a_{01} & (y_0=1<-->y_1=2)a_{02} \\
0 & 0
\end{array}\right]
\end{equation}

\begin{equation}
M_{2}(\mathbf{x})=\left[\begin{array}{cc}
(y_1=1<-->y_2=1)b_{11} & (y_1=1<-->y_2=2)b_{12} \\
(y_1=2<-->y_2=1)b_{21} & (y_1=2<-->y_2=2)b_{22} \\
\end{array}\right]
\end{equation}

\begin{equation}
M_{3}(\mathbf{x})=\left[\begin{array}{cc}
(y_2=1<-->y_3=1)c_{11} & (y_2=1<-->y_3=2)c_{12} \\
(y_2=1<-->y_3=1)c_{21} & (y_2=2<-->y_3=2)c_{22} \\
\end{array}\right]
\end{equation}

\begin{equation}
M_{4}(\mathbf{x})=\left[\begin{array}{cc}
(y_3=1<-->y_4=1)1& 0\\
(y_3=2<-->y_4=1)1 & 0
\end{array}\right]
\end{equation}

故从start到stop对应于$\mathbf{y}=(1,1,1),\mathbf{y}=(1,1,2),\cdots,\mathbf{y}=(2,2,2)$各路径的非规范化概率分别是

\begin{aligned}
&a_{01} b_{11} c_{11}, \quad a_{01} b_{11} c_{12}, \quad a_{01} b_{12} c_{21}, \quad a_{01} b_{12} c_{22}\\
&a_{02} b_{21} c_{11}, \quad a_{02} b_{21} c_{12}, \quad a_{02} b_{22} c_{21}, \quad a_{02} b_{22} c_{22}
\end{aligned}

然后根据式(1)求规范化因子.通过计算矩阵乘积$M_{1}(\mathbf{x}) M_{2}(\mathbf{x}) M_{3}(\mathbf{x}) M_{4}(\mathbf{x})$,其第1行第1列的元素为(其他位置元素易知为0):

$$ Z_{\mathbf{w}} =a_{01} b_{11} c_{11} +  a_{01} b_{11} c_{12} + a_{01} b_{12} c_{21}, + a_{01} b_{12} c_{22} + a_{02} b_{21} c_{11} +  a_{02} b_{21} c_{12} + a_{02} b_{22} c_{21}+ a_{02} b_{22} c_{22} $$

这恰好送上面等于从start到stop所有路径的非规范化概率之和
