```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_axies(i):  # 展示每次生成的图片
    t = np.arange(0.0, 2.0, 0.01)
    s = np.sin(i * np.pi * t)
    fig1, axes = plt.subplots()
    axes.plot(t, s)
    axes.set_title(i, color='red', size=23)
    fig1.savefig("img/" + str(i) + '.jpg')  # 通过Figure对象保存图片;保存每次生成的图片


def plot_axieso_cos(i):
    t = np.arange(0.0, 2.0, 0.01)
    s = np.cos(i * np.pi * t)
    fig1, axes = plt.subplots()
    axes.plot(t, s)
    axes.set_title(i, color='red', size=23)
    fig1.savefig("img/" + str(i) + 'cos_.jpg')


def plot_plt(i):  #  展示所有图片的融合,图片相关设置为最后一次设置的结果
    t = np.arange(0.0, 2.0, 0.01)
    s = np.sin(i * np.pi * t)
    plt.plot(t, s)

    plt.title(i, color='red', size=23)

    plt.savefig("img/" + str(i) + '_.png')  # 保存每次生成的图片和之前所有图片的融合


def plot_plt_show(i):  # 展示每次生成的图片
    t = np.arange(0.0, 2.0, 0.01)
    s = np.sin(i * np.pi * t)
    plt.plot(t, s)
    plt.title(i, color='red', size=23)
    plt.savefig("img/" + str(i) + ".png")  # 必须在plt.show()语句前;保存每次生成的图片
    plt.show()  # Display all open figures.
```

```python
lst = [1, 2, 3]
```

```python
# 推荐统一使用此方式进行绘图★★★★★
for i in lst:
    plot_axies(i)
    plot_axieso_cos(i)
```

```python
for j in lst:
    plot_plt_show(j)
```

```python
for k in lst:
    plot_plt(k)
```
