## 常见凸函数
1. 仿射函数
2. 指数函数.对任意$ a \in R $,函数$e^{ax}$在$R$上是凸的
3. 幂函数.当$a \geq 1或 a \leq 0 $时,$x^a$在$R^+$上是凸函数,当$0 \leq a \leq 1$时,$x^a$在$R^+$上是凹函数
4. 绝对幂函数.当$p \geq 1$时,函数$|x|^p$在$R$上是凸函数
5. 负对数函数.函数$- \log x$在$R^+$上是凸函数
6. 负熵.函数$x \log x$在其定义域上是凸函数
7. 范数($L_0$范数除外)
8. 任意个凸函数的逐点最大或上确界仍然是凸函数
9. 指数和的对数.函数$f(x)=\log(e^{x_1} + \dots + e^{x_n})$在$R^n$上是凸函数 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
x = np.linspace(0, 3, 100)
y = np.exp(2 * x)  # 指数函数
plt.plot(x, y)
plt.plot([x[10], x[80]], [y[10], y[80]], color='red')
```

```python
fig1 = plt.figure()
x = np.linspace(0.01, 3, 100)
y = np.power(x, -1)  # 幂函数
plt.plot(x, y)
plt.plot([x[2], x[80]], [y[2], y[80]], color='red')
```

```python
fig2 = plt.figure()
x = np.linspace(0.001, 3, 100)
y = -np.log(x)  # 负对数函数
plt.plot(x, y)
plt.plot([x[2], x[80]], [y[2], y[80]], color='red')
```

```python
fig3 = plt.figure()
x = np.linspace(-3, 3, 100)
y = np.power(np.abs(x), 1)  # 绝对幂函数
plt.plot(x, y)
plt.plot([x[2], x[80]], [y[2], y[80]], color='red')
```

```python
fig4 = plt.figure()
x = np.linspace(0.001, 3, 100)
y = x * np.log(x)  # 负熵
plt.plot(x, y)
plt.plot([x[2], x[80]], [y[2], y[80]], color='red')
```

```python
fig5 = plt.figure()
x = np.linspace(-3, 3, 100)
y = np.log(1 + np.exp(2 * x))  # 指数和的对数
plt.plot(x, y)
plt.plot([x[2], x[80]], [y[2], y[80]], color='red')
```

```python
fig6 = plt.figure()
x = np.linspace(0, 5, 100)
y1 = np.exp(x).reshape(-1, 1)  # 指数函数
y2 = np.power(10 * x, 1).reshape(-1, 1)  # 幂函数
conbine = np.concatenate((y1, y2), axis=1)
y = np.max(conbine, axis=1)  # 任意个凸函数的逐点最大或上确界仍然是凸函数
plt.plot(x, y)
plt.plot([x[2], x[80]], [y[2], y[80]], color='red')
```
