&emsp;&emsp;条件随机场的预测问题是给定条件随机场$P(Y|X)$和输入序列(观察序列)$\mathbf{x}$,求条件概率最大的输出序列(标记序列)$\mathbf{y}^*$,即对观测序列进行标注.条件随机场的预测算法是著名的维特比算法.

&emsp;&emsp;由

$$ P_{\mathbf{w}}(\mathbf{y} \mid \mathbf{x})=\frac{\exp (\mathbf{w} \cdot F(\mathbf{y}, \mathbf{x}))}{Z_{\mathbf{w}}(\mathbf{x})} $$

可得

$$
\begin{aligned}
\mathbf{y}^{*} &=\arg \max _{\mathbf{y}} P_{\mathbf{w}}(\mathbf{y} \mid \mathbf{x}) \\
&=\arg \max _{\mathbf{y}} \frac{\exp (\mathbf{w} \cdot F(\mathbf{y}, \mathbf{x}))}{Z_{\mathbf{w}}(\mathbf{x})} \\
&=\arg \max _{\mathbf{y}} \exp (\mathbf{w}  \cdot F(\mathbf{y}, \mathbf{x})) \\
&=\arg \max _{\mathbf{y}}(\mathbf{w} \cdot F(\mathbf{y}, \mathbf{x}))
\end{aligned}
$$

于是,条件随机场的预测问题成为求非规范化概率最大的最优路径问题

$$ \max_{\mathbf{y}} ( \mathbf{w} \cdot F(\mathbf{y}, \mathbf{x})) \tag{1} $$

这里,路径表示标记序列.其中,

$$
\begin{aligned}
\mathbf{w} &=\left(w_{1}, w_{2}, \cdots, w_{K}\right)^{\mathrm{T}} \\
F(\mathbf{y}, \mathbf{x}) &=\left(f_{1}(\mathbf{y}, x), f_{2}(\mathbf{y}, \mathbf{x}), \cdots, f_{K}(\mathbf{y}, \mathbf{x})\right)^{\mathrm{T}} \\
f_{k}(\mathbf{y}, \mathbf{x}) &=\sum_{i=1}^{n} f_{k}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right), \quad k=1,2, \cdots, K
\end{aligned}
$$

注意,这时只需计算非规范化概率,而不必计算概率,可以大大提高效率.为了求解最优路径,将式(1)写成如下形式($\mathbf{w}$与$i$无关)

$$ \max_{\mathbf{y}} \sum_{i=1}^n \mathbf{w}  \cdot F_i(y_{i-1}, y_i, \mathbf{x}) \tag{2} $$

其中,

$$ F_{i}\left(y_{i-1}, y_{i}, \mathbf{x} \right)=\left(f_{1}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right), f_{2}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right), \cdots, f_{K}\left(y_{i-1}, y_{i}, \mathbf{x}, i\right)\right)^{\mathrm{T}} $$
是局部特征向量.

<font color='red' size=4>算法:(维特比算法)</font>

输入:模型特征向量$F(\mathbf{y},\mathbf{x})$和权值向量$\mathbf{w}$,观测序列$\mathbf{x}=(x_1,x_2,\cdots,x_n)$

输出:最优路径$\mathbf{y}^* = (y^*_1,y^*_2,\cdots,y^*_n)$

1. 初始化

   $$ \delta_{1}(j)= \mathbf{w} \cdot F_1(y_0=\mathrm{start},y_1=j,\mathbf{x}), \qquad j=1,2,\cdots m $$

2. 递推,对$i=2,3,\cdots,n$

   $$
   \begin{array}{c}
   \delta_{i}(l)=\max _{1 \leqslant j \leqslant m}\left\{\delta_{i-1}(j)+ \mathbf{w} \cdot F_{i}\left(y_{i-1}=j, y_{i}=l, \mathbf{x} \right)\right\}, \quad l=1,2, \cdots, m \\
   \Psi_{i}(l)=\arg \max _{1 \leqslant j \leqslant m}\left\{\delta_{i-1}(j)+ \mathbf{w} \cdot F_{i}\left(y_{i-1}=j, y_{i}=l, \mathbf{x} \right)\right\}, \quad l=1,2, \cdots, m
   \end{array}
   $$

3. 终止
   
   $$
   \begin{array}{c}
   \max _{}(\mathbf{w} \cdot F(\mathbf{y}, \mathbf{x}))=\max _{1 \leqslant j \leqslant m} \delta_{n}(j) \\
   y_{n}^{*}=\arg \max _{1 \leqslant j \leqslant m} \delta_{n}(j)
   \end{array}
   $$

4. 返回路径

   $$ \mathbf{y}^*_i = \Psi_{t+1}(\mathbf{y}^*_{i+1}) \qquad i=n-1,n-2,\cdots,1 $$

求得最优路径$\mathbf{y}^* = (y^*_1,y^*_2,\cdots,y^*_n)$


**例**设有一标注问题:输入观测序列为$X=(X_1,X_2,X_3)$,输出标记序列为$Y=(Y_1,Y_2,Y_3),Y_1,Y_2,Y_3$取值于$\mathcal{Y}=\{1, 2 \}$.

&emsp;&emsp;假设特征$t_k,s_l$和对应的权值$\lambda_k,\mu_l$如下(这里只注明特征取值为1的条件,取值为0的条件省略):

$$
\begin{array}{ll}
t_{1}=t_{1}\left(y_{i-1}=1, y_{i}=2, \mathbf{x}, i\right), i=2,3 & \lambda_{1}=1 \\
t_{2}=t_{2}\left(y_{1}=1, y_{2}=1, \mathbf{x}, 2\right) & \lambda_{2}=0.6 \\
t_{3}=t_{3}\left(y_{2}=2, y_{3}=1, \mathbf{x}, 3\right) & \lambda_{3}=1 \\
t_{4}=t_{4}\left(y_{1}=2, y_{2}=1, \mathbf{x}, 2\right), & \lambda_{4}=1 \\
t_{5}=t_{5}\left(y_{2}=2, y_{3}=2, \mathbf{x}, 3\right), & \lambda_{5}=0.2 \\
s_{1}=s_{1}\left(y_{1}=1, \mathbf{x}, 1\right), & \mu_{1}=1 \\
s_{2}=s_{2}\left(y_{i}=2, \mathbf{x}, i\right), i=1,2 & \mu_{2}=0.5 \\
s_{3}=s_{3}\left(y_{i}=1, \mathbf{x}, i\right), i=2,3 & \mu_{3}=0.8 \\
s_{4}=s_{4}\left(y_{3}=2, \mathbf{x}, 3\right), & \mu_{4}=0.5
\end{array}
$$

用维特比算法求给定的输入序列(观测序列)$\mathbf{x}$对应的最优输出序列(标记序列)$\mathbf{y}^*=(y_1^*,y_2^*,y_3^*)$

**解**  由式(2)可得

$$ \max \sum_{i=1}^{3} \mathbf{w} \cdot F_{i}\left(y_{i-1}, y_{i}, x\right) $$

1. 初始化

   $$ \delta_{1}(j)=\mathbf{w} \cdot F_{1}\left(y_{0}=\operatorname{start}, y_{1}=j, \mathbf{x} \right), \quad j=1,2 $$

   $ i=1, \qquad  \delta_{1}(1)= \mu_1 \times s_1 = 1, \quad \delta_{1}(2)= \mu_2 \times s_2 = 0.5$

2. 递推


   $$
   \begin{align}
   i=2 \qquad  & \delta_{2}(l)=\max _{j}\left\{\delta_{1}(j)+w \cdot F_{2}(j, l, \mathbf{x})\right\} \\
   & \delta_{2}(1)=\max \left\{1+\lambda_{2} t_{2}+\mu_{3} s_{3}, 0.5+\lambda_{4} t_{4}+\mu_{3} s_{3}\right\}=2.4, \quad \Psi_{2}(1)=1 \\
    & \delta_{2}(2)=\max \left\{1+\lambda_{1} t_{1}+\mu_{2} s_{2}, 0.5+\mu_{2} s_{2}\right\}=2.5, \quad \Psi_{2}(2)=1 \\
   i=3 \qquad& \delta_{3}(l)=\max _{j}\left\{\delta_{2}(j)+w \cdot F_{3}(j, l, \mathbf{x})\right\} \\
   & \delta_{3}(1)=\max \left\{2.4+\mu_{3} s_{3}, 2.5+\lambda_{3} t_{3}+\mu_{3} s_{3}\right\}=4.3, \quad \Psi_{3}(1)=2 \\
   &\delta_{3}(2)=\max \left\{2.4+ \lambda_1 t_1 + \mu_{4} s_{4}, 2.5+\lambda_{5} t_{5}+\mu_{4} s_{4}\right\}=3.9, \quad \Psi_{3}(2)=1
   \end{align}
   $$

3. 终止
   
   $$
   \begin{array}{c}
   \max _{\mathbf{y}}(\mathbf{y} \cdot F(\mathbf{y}, \mathbf{x}))=\max \delta_{3}(l)=\delta_{3}(1)=4.3 \\
   y_{3}^{*}=\arg \max _{l} \delta_{3}(l)=1
   \end{array}
   $$

4. 返回

   $$
   \begin{array}{l}
   y_{2}^{*}=\Psi_{3}\left(y_{3}^{*}\right)=\Psi_{3}(1)=2 \\
   y_{1}^{*}=\Psi_{2}\left(y_{2}^{*}\right)=\Psi_{2}(2)=1
   \end{array}
   $$
   
最优标记序列

$$ \mathbf{y}^{*}=\left(y_{1}^{*}, y_{2}^{*}, y_{3}^{*}\right)=(1,2,1) $$
