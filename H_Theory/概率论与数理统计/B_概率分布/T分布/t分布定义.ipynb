<font color='red' size=4>定义:</font>   
&emsp;&emsp;设随机变量$X\sim N(0,1), Y\sim \chi_{n}^2,$且$X$与$Y$相互独立,则随机变量
$T=\frac{X}{\sqrt{Y/n}}$的分布称为自由度为n的t分布,记为$T\sim t_{n}$   
<font color='red' size=4>性质:</font>  
&emsp;&emsp;设$T \sim t_{n}$对于给定的$\alpha,0<\alpha<1$称满足条件  
$$ P \{T>t_{n}(\alpha)\} =\int_{t_{n}(\alpha)}^{+\infty} f(x)dx=\alpha $$     
的点$ t_{n}(\alpha)$为$t_{n}$分布上的$\alpha $分位点

```python
# IMPORTS
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('classic')
fig = plt.figure(figsize=(14, 7), dpi=100)
ax  = plt.subplot(111)

# PDF
ax.plot(np.linspace(-20, 20, 1000), 
         stats.t.pdf(np.linspace(-20, 20, 1000), df=4) ,)
ax.fill_between(np.linspace(-20, 20, 1000),
                 stats.t.pdf(np.linspace(-20, 20, 1000), df=4) ,
                 alpha=.15,)

# CDF
ax.plot(np.linspace(-20, 20, 1000), 
         stats.t.cdf(np.linspace(-20, 20, 1000), df=4))

ax.text(x=1, y=.25, s="pdf (normed)", alpha=.95, weight="bold", color="#008fd5")
ax.text(x=3, y=.85, s="cdf", alpha=.95, weight="bold", color="#fc4f30")

ax.set_xticks(np.arange(-20, 21, 2))
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = -20, y = 1.25, s = r"$ t $ Distribution - Overview",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = -20, y = 1.1, 
         s = 'Depicted below are the normed probability density function (pdf) and the cumulative density\nfunction (cdf) of a t distributed random variable $ T \sim t(4) $, given $df$=4.',
         fontsize = 19, alpha = .85)
```
