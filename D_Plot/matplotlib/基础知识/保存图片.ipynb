```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
fig1, axes = plt.subplots()
axes.plot(t, s)
axes.set_title(r'$\alpha_i > \beta_i$', fontsize=20)
axes.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
axes.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',
          fontsize=20)
axes.set_xlabel('time (s)')
axes.set_ylabel('volts (mV)')
axes.grid()
fig1.savefig("img/picture_0.jpg",
             dpi=500, facecolor='yellow')
'''
dpi : [ *None* | scalar > 0 | 'figure']
    The resolution in dots per inch.  If *None* it will default to
    the value ``savefig.dpi`` in the matplotlibrc file. If 'figure'
    it will set the dpi to be the value of the figure.
    
facecolor : color spec or None, optional
    the facecolor of the figure; if None, defaults to savefig.facecolor
'''

```
