## $ \Gamma $ 函数  
$$ \Gamma = \int_{0}^{+\infty} t_{\alpha-1}e^{-t} dt $$   
<font color='red' size=4>$\Gamma$函数性质:</font>
$ \Gamma(\alpha + 1) = \alpha \Gamma(\alpha) $  
$ \Gamma(1) = 1 $  
$ \Gamma(1/2) = \pi $  
易知:$ \Gamma(n) = (n-1)! $   


## $ \Gamma $分布
<font color='red' size=4>定义:</font>  
&emsp;&emsp;参数为$ (\alpha, \beta) $的伽马分布,其概率密度函数为:  
$$ f(x|\alpha ,\beta) = \frac{1}{\Gamma(\alpha)\beta^{\alpha}} x^{\alpha -1} e^{-x/\beta}, 0<x<+\infty,\alpha > 0, \beta >0$$  
其中参数$ \alpha $主要影响分布的峰起状态,成为形状参数;参数$ \beta$主要影响分布的散度情况,称为尺度参数.记该分布为$ gamma(\alpha, \beta) $  

* 期望和方差: $$ EX=\alpha \beta , VarX=\alpha \beta^2 $$   
* 矩母函数:$$ M_X(t)= (\frac{1}{1-\beta t})^\alpha, t<\frac{1}{\beta} $$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
from scipy.special import gamma

# PLOTTING CONFIG
%matplotlib inline
style.use('fivethirtyeight')

class Complete_gamma:
    """手工实现伽马分布"""
    
    def __init__(self, x, a, b):
        self.x = x
        self.a = a
        self.b = b
        
    def pdf(self):
        part0 = 1 / (gamma(self.a) * (self.b ** self.a))
        part1 = self.x ** (self.a - 1)
        part2 = np.exp(-self.x / self.b)
        return part0 * part1 * part2
```

## 参数 $ \alpha $ 对伽马分布的影响

```python
x = np.linspace(0, 10, 100)
a = np.linspace(0.5, 5, num=10)
b = range(1, 11)

plt.figure(figsize=(14, 7), dpi=100)
ax = plt.subplot(111)
for i in a:
    gam = Complete_gamma(x, i, 2) # \beta=2,即为卡方分布
    show = r' $ \alpha $ =' + str(i) + r',$ \beta  $=2'
    ax.plot(x, gam.pdf(), label=show)
    ax.legend()
```

## 参数$ \beta $对伽马分布的影响

```python
plt.figure(figsize=(14, 7), dpi=100)
ax1 = plt.subplot(111)
for i in np.array(b)/2:
    gam = Complete_gamma(x, 1, i)
    show = r' $ \alpha =1, \beta  $=' + str(i)
    ax1.plot(x, gam.pdf(), label=show)
    ax1.legend()
```
