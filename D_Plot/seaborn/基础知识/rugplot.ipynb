```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.subplot(111)
iris = sns.load_dataset("iris")
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
sns.kdeplot(x=setosa.sepal_width, y=setosa.sepal_length,
            cmap="Reds", cbar=True, shade=True, thresh=0.1)
sns.kdeplot(x=virginica.sepal_width, y=virginica.sepal_length,
            cmap="Blues", cbar=True, shade=True, thresh=0.2,
            cbar_kws={'orientation': 'horizontal'})

# Plot marginal distributions by drawing ticks along the x and y axes.
'''
x, yvectors or keys in data
    Variables that specify positions on the x and y axes.

height : scalar, optional
    Height of ticks as proportion of the axis.

axis : {‘x’ | ‘y’}, optional
    Axis to draw rugplot on.

color:Set the color of the line
'''
sns.rugplot(x=virginica.sepal_width, height=0.04, color='b')
sns.rugplot(y=virginica.sepal_length, height=0.04, color='b')
```
