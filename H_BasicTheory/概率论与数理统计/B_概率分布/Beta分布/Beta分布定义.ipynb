<font color='red' size=4>定义</font>  
&emsp;&emsp;贝塔分布是$ (0,1) $区间上含两个参数的一类连续分布,参数为$ (\alpha, \beta ) $的贝塔概率密度函数为:   
$$ f(x|\alpha,\beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)}, 0<x<1, \alpha>0, \beta>0$$  
其中$ B(\alpha, \beta) $表示$B$函数(贝塔函数):  
$$ B(\alpha, \beta) =  \int_0^1 x^{\alpha -1} (1-x)^{\beta -1} dx$$ &emsp;(正则化常量,用来保证该分布的积分为1)   
$ B $ 函数与$ \Gamma $函数之间存在下列关系:  
$$ B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}$$      

* 期望和方差:$$ E(X)=\frac{\alpha}{\alpha+\beta} \quad  Var(X)=\frac{\alpha \beta}{(\alpha + \beta)^2(\alpha+\beta+1)} $$    
* 矩母函数:$$ M_X(t) =  1+ \sum_{k=1}^{\infty} \big(  \prod_{r=0}^{k-1} \frac{\alpha + r}{ \alpha + \beta + r }  \big) \frac{k!}{t^k} $$
