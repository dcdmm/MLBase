```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Assign a variable to x to plot a univariate distribution along the x axis:
sns.histplot(data=penguins, x="flipper_length_mm",
             color='green',  # 直方图bin颜色(使用分组列时通过palette参数设置)
             edgecolor='b')  # 直方图bin边框颜色
```

```python
# Flip the plot by assigning the data variable to the y axis:
sns.histplot(data=penguins, y="flipper_length_mm")  # 水平方向
```

```python
sns.histplot(data=penguins, x="flipper_length_mm",
             bins=100)  # total number of bins
```

```python
# Add a kernel density estimate to smooth the histogram, providing complementary information about the shape of the distribution:
sns.histplot(data=penguins, x="flipper_length_mm",
             kde=True)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm",
             # 使用列"species"进行分组
             hue="species")  # 默认multiple='layer',分组结果遮盖(较低的bins被高的bins遮盖)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm",
             hue="species", multiple="stack")  # 分组结果堆叠
```

```python
sns.histplot(data=penguins, x="flipper_length_mm",
             hue="species", multiple="dodge")  # 分组结果不同列展示
```

```python
sns.histplot(data=penguins, x="flipper_length_mm",
             hue="species", multiple="dodge",
             palette='Set1')  # 通过调色盘为不同分组绘制不同颜色
```
