```python
import pandas as pd
import numpy as np

np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 2] = np.nan
```

```python
df.style
```

```python
import  matplotlib.pyplot as plt
s = df.style.background_gradient(cmap=plt.cm.cool)
s
```

```python
df.loc[:4].style.background_gradient(cmap=plt.cm.Blues) # 先切片再进行样式
```

```python
df.style.background_gradient(cmap=plt.cm.Blues, subset=pd.IndexSlice[0:4, :])  # 在样式中进行切片
```

```python
(df.loc[:4]
    .style
    .background_gradient(cmap=plt.cm.Set1)
    .highlight_null('red')).to_excel('styled.xlsx', engine='openpyxl')
```
