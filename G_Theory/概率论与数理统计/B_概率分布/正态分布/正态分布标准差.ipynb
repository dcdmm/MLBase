```python
import  numpy as np
import  matplotlib.pyplot as plt
from scipy import  stats
import matplotlib.style as style

style.use('classic')
fig = plt.figure(figsize=(14, 7), dpi=100)
ax = plt.subplot(111)

#\sigma = 1
ax.plot(np.linspace(-4, 4, 100), 
         stats.norm.pdf(np.linspace(-4, 4, 100), scale=1),
        )
'''
scale : array_like, optional
            scale parameter (default=1)
'''
ax.fill_between(np.linspace(-4, 4, 100),
                 stats.norm.pdf(np.linspace(-4, 4, 100), scale=1),
                 alpha=.15,
                )

#\sigma = 2
ax.plot(np.linspace(-4, 4, 100), 
         stats.norm.pdf(np.linspace(-4, 4, 100), scale=2),
        )
ax.fill_between(np.linspace(-4, 4, 100),
                 stats.norm.pdf(np.linspace(-4, 4, 100), scale=2),
                 alpha=.15,
                )

#\sigma = 0.5
ax.plot(np.linspace(-4, 4, 100), 
         stats.norm.pdf(np.linspace(-4, 4, 100), scale=0.5),
        )
ax.fill_between(np.linspace(-4, 4, 100),
                 stats.norm.pdf(np.linspace(-4, 4, 100), scale=0.5),
                 alpha=.15,
                )

ax.text(x=-1.25, y=.3, s="$ \sigma = 1$", rotation=51, alpha=1, weight="bold", color="red")
ax.text(x=-2.5, y=.13, s="$ \sigma = 2$", rotation=11, alpha=1, weight="bold", color="blue")
ax.text(x=-0.75, y=.55, s="$ \sigma = 0.5$", rotation=75, alpha=1, weight="bold", color="green")

ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = -5, y = 0.98, s = "Normal Distribution - $\sigma $",
               fontsize = 26, weight = 'bold', alpha = .75)
ax.text(x = -5, y = 0.87, 
         s = 'Depicted below are three normally distributed random variables with varying $\sigma $. As one can easily\nsee the parameter $\sigma$ "sharpens" the distribution (the smaller $ \sigma $ the sharper the function).',
         fontsize = 19, alpha = .85)
```
