```python
import numpy as np
import  matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import  norm

fig1 = plt.figure(figsize=(14, 7), dpi=100)
ax = plt.subplot(111)
mu_real = 10
sigma_real = 2

np.random.seed(42)
sample = stats.norm.rvs(loc=mu_real, scale=sigma_real, size=1000)

# ESTIMATE MU AND SIGMA
mu_est = np.mean(sample)
sigma_est = np.std(sample)
print("Estimated MU: {}\nEstimated SIGMA: {}".format(mu_est, sigma_est))

ax.hist(sample, bins=50, density=True, alpha=.25)

# TRUE CURVE
ax.plot(np.linspace(2, 18, 1000), 
        norm.pdf(np.linspace(2, 18, 1000), loc=mu_real, scale=sigma_real), color='red')

# ESTIMATED CURVE
ax.plot(np.linspace(2, 18, 1000), 
        norm.pdf(np.linspace(2, 18, 1000), loc=np.mean(sample), scale=np.std(sample)), color='g')

ax.text(x=9.5, y=.1, s="sample", alpha=.75, weight="bold", color="#008fd5")
ax.text(x=7, y=.2, s="true distrubtion", rotation=55, alpha=.75, weight="bold", color="red")
ax.text(x=5, y=.12, s="estimated distribution", rotation=55, alpha=.75, weight="bold", color="g")

ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

ax.text(x = 5, y = 0.27, s = "Normal Distribution", color='y',
               fontsize = 36, weight = 'bold', alpha = .95)
```
